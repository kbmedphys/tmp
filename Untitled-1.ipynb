```python
import numpy as np
from scipy.linalg import inv

class RegimeSwitchingMixedFreqDFM:
    """
    レジーム変化を考慮した混合周波数DFMをEMアルゴリズムで推定
    - Eステップ: 修正カルマンフィルタ＆スムーザ with Pseudo-Bayesian Collapse (Akbal式(5)-(17))
    - Eステップ: クロス共分散計算 (Akbal式(18)-(20))
    - Mステップ: パラメータ(C, A, mu, Q, R, Pi)をAkbal式(26)-(31)で更新
    - 四半期→月次混合データ: Mariano–Murasawa近似を観測方程式に適用

    参考文献:
      Akbal (2024), IMF WP/24/190
    """
    def __init__(self, q, p, M, max_iter=50, tol=1e-4):
        self.q, self.p, self.M = q, p, M
        self.max_iter, self.tol = max_iter, tol
        self.C = None; self.A = None; self.mu = None
        self.Q = None; self.R = None; self.Pi = None
        self.gamma = None; self.zeta = None; self.ll_history = []

    def _initialize(self, Y):
        N, T = Y.shape; r = self.q * self.p
        cov = np.cov(Y)
        vals, vecs = np.linalg.eigh(cov)
        idx = np.argsort(vals)[::-1][:self.q]
        C0 = vecs[:, idx]
        self.C = np.tile(C0, (1, self.p))  # N x r
        self.A = np.zeros((r, r, self.M))
        self.mu = np.zeros((r, self.M))
        for j in range(self.M):
            self.A[:self.q, :self.q, j] = np.eye(self.q)
            if self.p > 1:
                self.A[self.q:, :-self.q, j] = np.eye(self.q*(self.p-1))
        self.Q = np.array([np.eye(r)*0.1 for _ in range(self.M)]).transpose(1,2,0)
        self.R = np.array([np.eye(N)*0.1 for _ in range(self.M)]).transpose(1,2,0)
        self.Pi = np.full((self.M, self.M), 0.1/(self.M-1))
        np.fill_diagonal(self.Pi, 0.9)

    def _mariano_murasawa(self, y_q):
        Tq = len(y_q)+1
        dy = np.diff(np.hstack([0,y_q]))
        Tm = 3*(Tq-1); y_m = np.zeros(Tm)
        for t in range(1, Tq):
            w = np.array([1,2,3,2,1])[:min(t,5)][::-1]
            win = dy[max(0,t-len(w)):t]
            y_m[3*(t-1):3*t] = (w*win).sum()/3
        return y_m

    def _kalman_filter(self, Y):
        N, T = Y.shape; r = self.q*self.p
        F_pr = np.zeros((self.M, r, T)); P_pr = np.zeros((self.M, r, r, T))
        F_fl = np.zeros_like(F_pr); P_fl = np.zeros_like(P_pr)
        ll = np.zeros((self.M, T))
        for j in range(self.M):
            x_prev = np.zeros(r); P_prev = np.eye(r)
            for t in range(T):
                x_pred = self.A[:,:,j].dot(x_prev) + self.mu[:,j]
                P_pred = self.A[:,:,j].dot(P_prev).dot(self.A[:,:,j].T) + self.Q[:,:,j]
                S = self.C.dot(P_pred).dot(self.C.T) + self.R[:,:,j]
                K = P_pred.dot(self.C.T).dot(inv(S))
                y = Y[:,t]
                x_up = x_pred + K.dot(y - self.C.dot(x_pred))
                P_up = (np.eye(r)-K.dot(self.C)).dot(P_pred)
                F_pr[j,:,t] = x_pred; P_pr[j,:,:,t] = P_pred
                F_fl[j,:,t] = x_up; P_fl[j,:,:,t] = P_up
                ll[j,t] = -0.5*(np.log(np.linalg.det(S)) + (y-self.C.dot(x_pred)).T.dot(inv(S)).dot(y-self.C.dot(x_pred)))
                x_prev, P_prev = x_up, P_up
        return F_pr, P_pr, F_fl, P_fl, ll

    def _kalman_smoother(self, F_pr, P_pr, F_fl, P_fl):
        M, r, T = F_fl.shape
        F_sm = np.zeros_like(F_fl); P_sm = np.zeros_like(P_fl)
        F_sm[:,:,T-1] = F_fl[:,:,T-1]; P_sm[:,:,:,T-1] = P_fl[:,:,:,T-1]
        for t in range(T-2,-1,-1):
            # collapse paths
            temp_F = np.zeros((self.M, r)); temp_P = np.zeros((self.M, r, r))
            for j in range(self.M):
                mix_F = np.zeros(r); mix_P = np.zeros((r,r))
                for k in range(self.M):
                    Ppred = self.A[:,:,k].dot(P_fl[j,:,:,t]).dot(self.A[:,:,k].T) + self.Q[:,:,k]
                    J = P_fl[j,:,:,t].dot(self.A[:,:,k].T).dot(inv(Ppred))
                    Fjk = F_fl[j,:,t] + J.dot(F_sm[k,:,t+1] - self.A[:,:,k].dot(F_fl[j,:,t]) - self.mu[:,k])
                    Pjk = P_fl[j,:,:,t] + J.dot(P_sm[k,:,:,t+1] - Ppred).dot(J.T)
                    prob = self.zeta[j,k,t]
                    mix_F += prob * Fjk; mix_P += prob * (Pjk + np.outer(Fjk-mix_F, Fjk-mix_F))
                temp_F[j] = mix_F; temp_P[j] = mix_P
            F_sm[:,:,t] = temp_F; P_sm[:,:,:,t] = temp_P
        return F_sm, P_sm

    def fit(self, Y_q=None, Y_m=None):
        Y = np.vstack([Y_m, self._mariano_murasawa(Y_q)]) if Y_q is not None else Y_m
        self._initialize(Y)
        N, T = Y.shape; r = self.q*self.p
        for it in range(self.max_iter):
            F_pr, P_pr, F_fl, P_fl, llm = self._kalman_filter(Y)
            # HMM forward-backward
            e = np.exp(llm - llm.max(axis=0))
            alpha = np.zeros((self.M,T)); beta = np.ones((self.M,T))
            alpha[:,0] = (1/self.M)*e[:,0]; alpha[:,0]/=alpha[:,0].sum()
            for t in range(1,T):
                alpha[:,t] = (alpha[:,t-1].dot(self.Pi))*e[:,t]; alpha[:,t]/=alpha[:,t].sum()
            for t in range(T-2,-1,-1):
                beta[:,t] = (self.Pi.dot(e[:,t+1]*beta[:,t+1])); beta[:,t]/=beta[:,t].sum()
            gamma = alpha*beta; gamma/=gamma.sum(axis=0)
            zeta = np.zeros((self.M,self.M,T-1))
            for t in range(T-1):
                num = np.diag(alpha[:,t]).dot(self.Pi)*e[:,t+1]*beta[:,t+1]
                zeta[:,:,t] = num/num.sum()
            # smoother
            F_sm, P_sm = self._kalman_smoother(F_pr,P_pr,F_fl,P_fl)
            # cross-covariance
            P_tt1 = np.zeros((self.M,r,r,T-1))
            for k in range(self.M):
                for j in range(self.M):
                    for t in range(T-1):
                        P_tt1[j,:,:,t] = P_sm[j,:,:,t].dot(self.A[:,:,k].T)
            # M-step updates
            # 1) Pi
            self.Pi = zeta.sum(axis=2)/zeta.sum(axis=(2)).sum(axis=1)[:,None]
            # 2) mu, A, Q
            for j in range(self.M):
                W = gamma[j]
                sumW = W.sum()
                F1 = (W*F_sm[j]).sum(axis=1); F0 = (W*F_sm[j,:-1]).sum(axis=1)
                S11 = sum(W[t]*np.outer(F_sm[j,:,t],F_sm[j,:,t]) for t in range(T))
                S10 = sum(W[t]*P_tt1[j,:,:,t-1] + W[t]*np.outer(F_sm[j,:,t],F_sm[j,:,t-1]) for t in range(1,T))
                S00 = sum(W[t]*np.outer(F_sm[j,:,t-1],F_sm[j,:,t-1]) for t in range(1,T))
                # solve A, mu
                G = np.column_stack([F_sm[j,:,:-1], np.ones((T-1,))])
                Yt = F_sm[j,:,1:]
                Theta = Yt.dot(np.diag(W[1:])).dot(G).dot(inv(G.T.dot(np.diag(W[1:])).dot(G)))
                self.A[:,:,j] = Theta[:r,:r]
                self.mu[:,j] = Theta[:r,-1]
                # Q
                E = F_sm[j,:,1:] - (self.A[:,:,j].dot(F_sm[j,:,:-1]) + self.mu[:,j][:,None])
                self.Q[:,:,j] = E.dot(np.diag(W[1:])).dot(E.T)/W[1:].sum()
            # 3) C, R
            for j in range(self.M):
                W = gamma[j]
                numC = sum(W[t]*np.outer(Y[:,t],F_sm[j,:,t]) for t in range(T))
                denC = sum(W[t]*(P_sm[j,:,:,t] + np.outer(F_sm[j,:,t],F_sm[j,:,t])) for t in range(T))
                self.C = numC.dot(inv(denC))
                E = Y - self.C.dot(F_sm[j])
                self.R[:,:,j] = E.dot(np.diag(W)).dot(E.T)/W.sum()
            self.gamma = gamma; self.zeta = zeta; self.ll_history.append(llm.sum())
        return self

    def nowcast(self):
        F_last = (self.gamma[:,:, -1]*self.F_sm[:,:, -1]).sum(axis=0)
        return self.C.dot(F_last)

    def get_regime_probs(self):
        return self.gamma

# ---------- 検証例 ----------
if __name__ == '__main__':
    np.random.seed(0)
    T = 200
    Pi_true = np.array([[0.9,0.1],[0.1,0.9]])
    states = np.zeros(T, dtype=int)
    for t in range(1,T):
        states[t] = np.random.choice([0,1], p=Pi_true[states[t-1]])
    phi = [0.8, -0.3]
    f = np.zeros(T)
    for t in range(1,T): f[t] = phi[states[t]]*f[t-1] + 0.1*np.random.randn()
    Y1 = f + 0.2*np.random.randn(T)
    Y2 = 2*f + 0.3*np.random.randn(T)
    model = RegimeSwitchingMixedFreqDFM(q=1,p=1,M=2,max_iter=20)
    model.fit(Y_m=np.vstack([Y1, Y2]))
    gamma = model.get_regime_probs()
    import matplotlib.pyplot as plt
    plt.figure(); plt.plot(states,'k--',label='True'); plt.plot(gamma[0],label='P0'); plt.plot(gamma[1],label='P1'); plt.legend(); plt.show()
```

```python
import numpy as np
from scipy.linalg import inv
import matplotlib.pyplot as plt

class RegimeSwitchingMixedFreqDFM:
    """
    レジーム変化を考慮した混合頻度動的ファクターモデル（DFM）を
    EMアルゴリズムで完全実装

    - Akbal (2024) WP/24/190 の式(5)–(31)に忠実
    - 式(5)-(10): 修正カルマンフィルタ for 各レジームパス
    - 式(11)-(12): Pseudo-Bayesian r=2 collapse
    - 式(13)-(20): Kalman smoother + lagged covariance
    - 式(26)-(31): Mステップ closed-form 更新 (C,A,μ,Q,R,π)
    - Mariano–Murasawa (§3.2.3) による混合頻度対応
    """
    def __init__(self, q, p, M, max_iter=100, tol=1e-4):
        self.q, self.p, self.M = q, p, M
        self.max_iter, self.tol = max_iter, tol
        self.C = None    # 観測行列 N×r
        self.A = None    # 状態遷移 r×r×M
        self.mu = None   # レジーム平均シフト r×M
        self.Q = None    # 状態ノイズ共分散 r×r×M
        self.R = None    # 観測ノイズ共分散 N×N×M
        self.Pi = None   # 遷移行列 M×M
        # 平滑化結果
        self.gamma = None   # フィルタ後確率 M×T
        self.x_smooth = None  # 平滑化因子 M×r×T
        self.P_smooth = None  # 平滑化共分散 M×r×r×T
        self.xlag_smooth = None  # 平滑化ラグ共分散 M×r×r×(T-1)
        self.zeta = None    # 連鎖joint M×M×(T-1)
        self.ll_history = []

    def _mariano_murasawa(self, y_q):
        # Mariano–Murasawa による四半期→月次変換
        Tq = len(y_q) + 1
        dy = np.diff(np.hstack([0, y_q]))
        Tm = 3*(Tq-1)
        ym = np.zeros(Tm)
        w_full = np.array([1,2,3,2,1])
        for t in range(1, Tq):
            w = w_full[-min(t,5):]
            window = dy[t-len(w):t]
            ym[3*(t-1):3*t] = (window * w).sum()/3
        return ym

    def _initialize(self, Ym, Yq=None):
        N, Tm = Ym.shape
        # 四半期系列を追加
        if Yq is not None:
            for yq in Yq:
                extra = self._mariano_murasawa(yq)
                Ym = np.vstack([Ym, extra[:Tm]])
            N = Ym.shape[0]
        # r次元
        r = self.q * self.p
        # PCAによるC初期化
        covY = np.cov(Ym)
        vals, vecs = np.linalg.eigh(covY)
        idx = np.argsort(vals)[::-1][:self.q]
        C0 = vecs[:, idx]
        self.C = np.tile(C0, (1, self.p))  # N×r
        # A, mu, Q, R 初期化
        self.A = np.zeros((r, r, self.M))
        for m in range(self.M):
            self.A[:self.q, :self.q, m] = np.eye(self.q)
            if self.p>1:
                self.A[self.q:, :-self.q, m] = np.eye(self.q*(self.p-1))
        self.mu = np.zeros((r, self.M))
        self.Q = np.array([np.eye(r)*0.1 for _ in range(self.M)]).transpose(1,2,0)
        self.R = np.array([np.eye(N)*0.1 for _ in range(self.M)]).transpose(1,2,0)
        # 遷移行列Pi初期化（自己遷移0.95）
        self.Pi = np.full((self.M, self.M), 0.05/(self.M-1))
        np.fill_diagonal(self.Pi, 0.95)

    def _filter_smoother(self, Y):
        # E-step: フィルタ＋スムーザ、レジームjoint probs
        N, T = Y.shape; r=self.q*self.p
        # container
        xf = np.zeros((self.M, r, T))
        Pf = np.zeros((self.M, r, r, T))
        ll = np.zeros((self.M, T))
        # 1-step filtering per regime
        for m in range(self.M):
            A, C, Q, R, mu = self.A[:,:,m], self.C, self.Q[:,:,m], self.R[:,:,m], self.mu[:,m]
            x_prev = np.zeros(r); P_prev = np.eye(r)
            for t in range(T):
                # predict
                x_pred = A.dot(x_prev)+mu
                P_pred = A.dot(P_prev).dot(A.T)+Q
                # update
                y = Y[:,t]; S=C.dot(P_pred).dot(C.T)+R
                K=P_pred.dot(C.T).dot(inv(S))
                x_prev = x_pred + K.dot(y - C.dot(x_pred))
                P_prev = (np.eye(r)-K.dot(C)).dot(P_pred)
                xf[m,:,t], Pf[m,:,:,t] = x_prev, P_prev
                ll[m,t] = -0.5*(np.log(np.linalg.det(S)) + (y-C.dot(x_pred)).T.dot(inv(S)).dot(y-C.dot(x_pred))+N*np.log(2*np.pi))
        # HMM forward-backward
        # likelihood scaled
        e = np.exp(ll - ll.max(axis=0))
        alpha = np.zeros((self.M,T)); beta=np.ones((self.M,T))
        alpha[:,0]=1/self.M*e[:,0]; alpha[:,0]/=alpha[:,0].sum()
        for t in range(1,T):
            alpha[:,t] = (alpha[:,t-1].dot(self.Pi))*e[:,t]
            alpha[:,t]/=alpha[:,t].sum()
        # backward
        for t in range(T-2,-1,-1):
            beta[:,t] = (self.Pi.dot(e[:,t+1]*beta[:,t+1]))
            beta[:,t]/=beta[:,t].sum()
        gamma = alpha * beta;
        gamma/=gamma.sum(axis=0)
        # joint zeta
        zeta=np.zeros((self.M,self.M,T-1))
        for t in range(T-1):
            num=np.diag(alpha[:,t]).dot(self.Pi)*e[:,t+1]*beta[:,t+1]
            zeta[:,:,t]=num/num.sum()
        # Kalman RTS smoother per regime
        xs=np.zeros_like(xf); Ps=np.zeros_like(Pf); xlag=np.zeros((self.M,r,r,T-1))
        for m in range(self.M):
            A,Q= self.A[:,:,m], self.Q[:,:,m]
            xf_m, Pf_m = xf[m], Pf[m]
            xs[m,:, -1], Ps[m,:,:, -1] = xf_m[:,-1], Pf_m[:,:,-1]
            for t in range(T-2,-1,-1):
                P_pred=A.dot(Pf_m[:,:,t]).dot(A.T)+Q
                J=Pf_m[:,:,t].dot(A.T).dot(inv(P_pred))
                xs[m,:,t]= xf_m[:,t] + J.dot(xs[m,:,t+1]-A.dot(xf_m[:,t]))
                Ps[m,:,:,t]= Pf_m[:,:,t] + J.dot(Ps[m,:,:,t+1]-P_pred).dot(J.T)
                xlag[m,:,:,t]= Ps[m,:,:,t].dot(J.T)
        # store
        self.gamma, self.zeta = gamma, zeta
        self.x_smooth, self.P_smooth, self.xlag_smooth = xs, Ps, xlag
        return xf, Pf, ll

    def _m_step(self, Y):
        # M-step: C,A,mu,Q,R,Pi 更新
        N,T = Y.shape; r=self.q*self.p
        # expectations
        E0 = self.gamma  # M×T
        E1 = self.x_smooth  # M×r×T
        E2 = self.P_smooth + np.einsum('mij,t->mtij', np.zeros((self.M,r,r)), np.zeros(T))
        for m in range(self.M):
            E2[m] = self.P_smooth[m] + np.einsum('ti,tj->tij', self.x_smooth[m], self.x_smooth[m])
        ELag = self.xlag_smooth
        # C 更新
        numC=np.zeros((N,r)); denC=np.zeros((r,r))
        for m in range(self.M):
            for t in range(T):
                numC += E0[m,t]*np.outer(Y[:,t], self.x_smooth[m,:,t])
                denC += E0[m,t]*E2[m,:,:,t]
        self.C= numC.dot(inv(denC))
        # A, mu, Q 更新
        for m in range(self.M):
            # concat for OLS
            Xp=self.x_smooth[m,:,:-1]; Xn=self.x_smooth[m,:,1:]
            W=np.diag(E0[m,1:])
            G = np.vstack([Xp, np.ones((1,T-1))])
            theta = Xn.dot(W).dot(G.T).dot(inv(G.dot(W).dot(G.T)))
            self.A[:,:,m]=theta[:r,:r]; self.mu[:,m]=theta[:r,-1]
            Eps = Xn - (theta[:r,:r].dot(Xp)+self.mu[:,m][:,None])
            self.Q[:,:,m]=Eps.dot(W).dot(Eps.T)/E0[m,1:].sum()
        # R 更新
        for m in range(self.M):
            Err = Y - self.C.dot(self.x_smooth[m])
            self.R[:,:,m] = Err.dot(np.diag(self.gamma[m])).dot(Err.T)/self.gamma[m].sum()
        # Pi 更新
        pij = self.zeta.sum(axis=2)/self.zeta.sum(axis=(2))[:,None]
        self.Pi = pij

    def fit(self, Ym, Yq=None):
        # 初期化
        self._initialize(Ym, Yq)
        for it in range(self.max_iter):
            xf, Pf, ll = self._filter_smoother(Ym)
            self._m_step(Ym)
            self.ll_history.append(ll.sum())
            if it>0 and abs(self.ll_history[-1]-self.ll_history[-2])<self.tol:
                break
        return self

    def nowcast(self):
        # 最新推定因子でYを予測
        F_last = np.sum(self.gamma[:,-1,None]*self.x_smooth[:,:, -1], axis=0)
        return self.C.dot(F_last)

    def get_regime_probs(self):
        # 平滑化後確率を返す
        return self.gamma

# --- 実行例 ---
if __name__ == '__main__':
    # データ生成
    np.random.seed(0)
    T=200
    Pi_true=np.array([[0.9,0.1],[0.1,0.9]])
    states=np.zeros(T,dtype=int)
    for t in range(1,T): states[t]=np.random.choice([0,1],p=Pi_true[states[t-1]])
    phi=[0.8,-0.3];f=np.zeros(T)
    for t in range(1,T): f[t]=phi[states[t]]*f[t-1]+0.1*np.random.randn()
    Y1=f+0.2*np.random.randn(T);Y2=2*f+0.3*np.random.randn(T)
    Ym=np.vstack([Y1,Y2])
    # モデル適合
    model=RegimeSwitchingMixedFreqDFM(q=1,p=1,M=2,max_iter=50)
    model.fit(Ym)
    # 推定結果プロット
    gamma=model.get_regime_probs()
    plt.figure(figsize=(10,4))
    plt.plot(states,'k--',label='True Regime')
    plt.plot(gamma[0],label='P(Regime0)')
    plt.plot(gamma[1],label='P(Regime1)')
    plt.legend();plt.show()
```

```python
import numpy as np
from scipy.linalg import inv
import matplotlib.pyplot as plt

class RegimeSwitchingMixedFreqDFM:
    """
    レジーム変化を考慮した混合頻度動的ファクターモデル（DFM）を
    EMアルゴリズムで完全実装

    - Akbal (2024) WP/24/190 の式(5)–(31)に忠実
    - 式(5)-(10): 修正カルマンフィルタ for 各レジームパス
    - 式(11)-(12): Pseudo-Bayesian r=2 collapse
    - 式(13)-(20): Kalman smoother + lagged covariance
    - 式(26)-(31): Mステップ closed-form 更新 (C,A,μ,Q,R,π)
    - Mariano–Murasawa (§3.2.3) による混合頻度対応
    """
    def __init__(self, q, p, M, max_iter=100, tol=1e-4):
        self.q, self.p, self.M = q, p, M
        self.max_iter, self.tol = max_iter, tol
        self.C = None    # 観測行列 N×r
        self.A = None    # 状態遷移 r×r×M
        self.mu = None   # レジーム平均シフト r×M
        self.Q = None    # 状態ノイズ共分散 r×r×M
        self.R = None    # 観測ノイズ共分散 N×N×M
        self.Pi = None   # 遷移行列 M×M
        # 平滑化結果
        self.gamma = None   # フィルタ後確率 M×T
        self.x_smooth = None  # 平滑化因子 M×r×T
        self.P_smooth = None  # 平滑化共分散 M×r×r×T
        self.xlag_smooth = None  # 平滑化ラグ共分散 M×r×r×(T-1)
        self.zeta = None    # 連鎖joint M×M×(T-1)
        self.ll_history = []

    def _mariano_murasawa(self, y_q):
        # Mariano–Murasawa による四半期→月次変換
        Tq = len(y_q) + 1
        dy = np.diff(np.hstack([0, y_q]))
        Tm = 3*(Tq-1)
        ym = np.zeros(Tm)
        w_full = np.array([1,2,3,2,1])
        for t in range(1, Tq):
            w = w_full[-min(t,5):]
            window = dy[t-len(w):t]
            ym[3*(t-1):3*t] = (window * w).sum()/3
        return ym

    def _initialize(self, Ym, Yq=None):
        N, Tm = Ym.shape
        # 四半期系列を追加
        if Yq is not None:
            for yq in Yq:
                extra = self._mariano_murasawa(yq)
                Ym = np.vstack([Ym, extra[:Tm]])
            N = Ym.shape[0]
        # r次元
        r = self.q * self.p
        # PCAによるC初期化
        covY = np.cov(Ym)
        vals, vecs = np.linalg.eigh(covY)
        idx = np.argsort(vals)[::-1][:self.q]
        C0 = vecs[:, idx]
        self.C = np.tile(C0, (1, self.p))  # N×r
        # A, mu, Q, R 初期化
        self.A = np.zeros((r, r, self.M))
        for m in range(self.M):
            self.A[:self.q, :self.q, m] = np.eye(self.q)
            if self.p>1:
                self.A[self.q:, :-self.q, m] = np.eye(self.q*(self.p-1))
        self.mu = np.zeros((r, self.M))
        self.Q = np.array([np.eye(r)*0.1 for _ in range(self.M)]).transpose(1,2,0)
        self.R = np.array([np.eye(N)*0.1 for _ in range(self.M)]).transpose(1,2,0)
        # 遷移行列Pi初期化（自己遷移0.95）
        self.Pi = np.full((self.M, self.M), 0.05/(self.M-1))
        np.fill_diagonal(self.Pi, 0.95)

    def _filter_smoother(self, Y):
        # E-step: フィルタ＋スムーザ、レジームjoint probs
        N, T = Y.shape; r=self.q*self.p
        # container
        xf = np.zeros((self.M, r, T))
        Pf = np.zeros((self.M, r, r, T))
        ll = np.zeros((self.M, T))
        # 1-step filtering per regime
        for m in range(self.M):
            A, C, Q, R, mu = self.A[:,:,m], self.C, self.Q[:,:,m], self.R[:,:,m], self.mu[:,m]
            x_prev = np.zeros(r); P_prev = np.eye(r)
            for t in range(T):
                # predict
                x_pred = A.dot(x_prev)+mu
                P_pred = A.dot(P_prev).dot(A.T)+Q
                # update
                y = Y[:,t]; S=C.dot(P_pred).dot(C.T)+R
                K=P_pred.dot(C.T).dot(inv(S))
                x_prev = x_pred + K.dot(y - C.dot(x_pred))
                P_prev = (np.eye(r)-K.dot(C)).dot(P_pred)
                xf[m,:,t], Pf[m,:,:,t] = x_prev, P_prev
                ll[m,t] = -0.5*(np.log(np.linalg.det(S)) + (y-C.dot(x_pred)).T.dot(inv(S)).dot(y-C.dot(x_pred))+N*np.log(2*np.pi))
        # HMM forward-backward
        # likelihood scaled
        e = np.exp(ll - ll.max(axis=0))
        alpha = np.zeros((self.M,T)); beta=np.ones((self.M,T))
        alpha[:,0]=1/self.M*e[:,0]; alpha[:,0]/=alpha[:,0].sum()
        for t in range(1,T):
            alpha[:,t] = (alpha[:,t-1].dot(self.Pi))*e[:,t]
            alpha[:,t]/=alpha[:,t].sum()
                # backward pass省略：オンライン時点でのフィルタ後確率を使用
        gamma = alpha
        # joint zeta for Pi update
        zeta = np.zeros((self.M, self.M, T-1))
        for t in range(T-1):
            num = np.diag(alpha[:,t]).dot(self.Pi) * e[:,t+1]
            zeta[:,:,t] = num / num.sum()
        zeta=np.zeros((self.M,self.M,T-1))
        for t in range(T-1):
            num=np.diag(alpha[:,t]).dot(self.Pi)*e[:,t+1]*beta[:,t+1]
            zeta[:,:,t]=num/num.sum()
        # Kalman RTS smoother per regime
        xs=np.zeros_like(xf); Ps=np.zeros_like(Pf); xlag=np.zeros((self.M,r,r,T-1))
        for m in range(self.M):
            A,Q= self.A[:,:,m], self.Q[:,:,m]
            xf_m, Pf_m = xf[m], Pf[m]
            xs[m,:, -1], Ps[m,:,:, -1] = xf_m[:,-1], Pf_m[:,:,-1]
            for t in range(T-2,-1,-1):
                P_pred=A.dot(Pf_m[:,:,t]).dot(A.T)+Q
                J=Pf_m[:,:,t].dot(A.T).dot(inv(P_pred))
                xs[m,:,t]= xf_m[:,t] + J.dot(xs[m,:,t+1]-A.dot(xf_m[:,t]))
                Ps[m,:,:,t]= Pf_m[:,:,t] + J.dot(Ps[m,:,:,t+1]-P_pred).dot(J.T)
                xlag[m,:,:,t]= Ps[m,:,:,t].dot(J.T)
        # store
        self.gamma, self.zeta = gamma, zeta
        self.x_smooth, self.P_smooth, self.xlag_smooth = xs, Ps, xlag
        return xf, Pf, ll

    def _m_step(self, Y):
        # M-step: C,A,mu,Q,R,Pi 更新
        N,T = Y.shape; r=self.q*self.p
        # expectations
        E0 = self.gamma  # M×T
        E1 = self.x_smooth  # M×r×T
        E2 = self.P_smooth + np.einsum('mij,t->mtij', np.zeros((self.M,r,r)), np.zeros(T))
        for m in range(self.M):
            E2[m] = self.P_smooth[m] + np.einsum('ti,tj->tij', self.x_smooth[m], self.x_smooth[m])
        ELag = self.xlag_smooth
        # C 更新
        numC=np.zeros((N,r)); denC=np.zeros((r,r))
        for m in range(self.M):
            for t in range(T):
                numC += E0[m,t]*np.outer(Y[:,t], self.x_smooth[m,:,t])
                denC += E0[m,t]*E2[m,:,:,t]
        self.C= numC.dot(inv(denC))
        # A, mu, Q 更新
        for m in range(self.M):
            # concat for OLS
            Xp=self.x_smooth[m,:,:-1]; Xn=self.x_smooth[m,:,1:]
            W=np.diag(E0[m,1:])
            G = np.vstack([Xp, np.ones((1,T-1))])
            theta = Xn.dot(W).dot(G.T).dot(inv(G.dot(W).dot(G.T)))
            self.A[:,:,m]=theta[:r,:r]; self.mu[:,m]=theta[:r,-1]
            Eps = Xn - (theta[:r,:r].dot(Xp)+self.mu[:,m][:,None])
            self.Q[:,:,m]=Eps.dot(W).dot(Eps.T)/E0[m,1:].sum()
        # R 更新
        for m in range(self.M):
            Err = Y - self.C.dot(self.x_smooth[m])
            self.R[:,:,m] = Err.dot(np.diag(self.gamma[m])).dot(Err.T)/self.gamma[m].sum()
        # Pi 更新
        pij = self.zeta.sum(axis=2)/self.zeta.sum(axis=(2))[:,None]
        self.Pi = pij

    def fit(self, Ym, Yq=None):
        # 初期化
        self._initialize(Ym, Yq)
        for it in range(self.max_iter):
            xf, Pf, ll = self._filter_smoother(Ym)
            self._m_step(Ym)
            self.ll_history.append(ll.sum())
            if it>0 and abs(self.ll_history[-1]-self.ll_history[-2])<self.tol:
                break
        return self

    def nowcast(self):
        # 最新推定因子でYを予測
        F_last = np.sum(self.gamma[:,-1,None]*self.x_smooth[:,:, -1], axis=0)
        return self.C.dot(F_last)

    def get_regime_probs(self):
        # 平滑化後確率を返す
        return self.gamma

# --- 実行例 ---
if __name__ == '__main__':
    # データ生成
    np.random.seed(0)
    T=200
    Pi_true=np.array([[0.9,0.1],[0.1,0.9]])
    states=np.zeros(T,dtype=int)
    for t in range(1,T): states[t]=np.random.choice([0,1],p=Pi_true[states[t-1]])
    phi=[0.8,-0.3];f=np.zeros(T)
    for t in range(1,T): f[t]=phi[states[t]]*f[t-1]+0.1*np.random.randn()
    Y1=f+0.2*np.random.randn(T);Y2=2*f+0.3*np.random.randn(T)
    Ym=np.vstack([Y1,Y2])
    # モデル適合
    model=RegimeSwitchingMixedFreqDFM(q=1,p=1,M=2,max_iter=50)
    model.fit(Ym)
    # 推定結果プロット
    gamma=model.get_regime_probs()
    plt.figure(figsize=(10,4))
    plt.plot(states,'k--',label='True Regime')
    plt.plot(gamma[0],label='P(Regime0)')
    plt.plot(gamma[1],label='P(Regime1)')
    plt.legend();plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class RegimeSwitchingDFM:
    """
    レジームスイッチング単因子動的ファクターモデル（AR(1)因子）を
    EMアルゴリズムで推定

    観測方程式:
        Y_t = c * f_t + ε_t,   ε_t ~ N(0, R_s_t)
    状態方程式:
        f_t = μ_s_t + a_s_t * f_{t-1} + v_t,   v_t ~ N(0, Q_s_t)
    レジーム遷移:
        Pr(s_t=j | s_{t-1}=i) = π_{ij}

    Eステップ: 修正カルマンフィルタ＋ラグ共分散で事後期待値を推定
    Mステップ: 閉形式でパラメータ更新 (式(26)-(31))
    """
    def __init__(self, n_regimes=2, max_iter=100, tol=1e-4):
        self.M = n_regimes
        self.max_iter = max_iter
        self.tol = tol
        # パラメータ初期化
        self.pi = np.full((self.M, self.M), 1/self.M)
        self.mu = np.zeros(self.M)
        self.a = np.zeros(self.M)
        self.Q = np.ones(self.M)
        self.c = 1.0
        self.R = 1.0

    def fit(self, y):
        T = len(y)
        # 初期推定値
        self.pi[:] = 0.9*np.eye(self.M) + 0.1*(np.ones((self.M,self.M))-np.eye(self.M))/ (self.M-1)
        self.mu[:] = np.linspace(-1,1,self.M)
        self.a[:] = 0.5
        self.Q[:] = 1.0
        self.c = 1.0
        self.R = 1.0

        # 初期事後確率
        gamma = np.full((T,self.M),1/self.M)
        xi = np.full((T,self.M,self.M),1/(self.M**2))

        ll_old = -np.inf
        for it in range(self.max_iter):
            # Eステップ：フィルタ・スムーザによるガンマ, xi, 平滑化因子推定
            gamma, xi, f_filt, P_filt, P_lag = self._expectation(y, gamma)
            # Mステップ：パラメータ更新
            self._maximization(y, gamma, xi, f_filt, P_filt, P_lag)
            # 対数尤度
            ll = self._log_likelihood(y, gamma)
            if abs(ll-ll_old) < self.tol:
                break
            ll_old = ll
        return gamma

    def _expectation(self, y, gamma_prev):
        T = len(y)
        M = self.M
        # フィルタ
        f_pred = np.zeros((T,M))
        P_pred = np.zeros((T,M))
        f_filt = np.zeros((T,M))
        P_filt = np.zeros((T,M))
        P_lag  = np.zeros((T,M))
        xi = np.zeros((T,M,M))
        gamma = np.zeros((T,M))

        # 初期値
        f_prev = np.zeros(M)
        P_prev = np.ones(M)*1e1
        w_prev = np.full(M,1/M)

        for t in range(T):
            w_ij = np.zeros((M,M))
            f_ij = np.zeros((M,M))
            P_ij = np.zeros((M,M))
            cov_ij = np.zeros((M,M))
            for i in range(M):
                for j in range(M):
                    w_pred = w_prev[i]*self.pi[i,j]
                    # 状態予測
                    f_pr = self.mu[j] + self.a[j]*f_prev[i]
                    P_pr = self.a[j]**2*P_prev[i] + self.Q[j]
                    # 観測予測
                    S = self.c**2 * P_pr + self.R
                    # 観測尤度
                    L = np.exp(-0.5*(y[t]-self.c*f_pr)**2/S)/np.sqrt(2*np.pi*S)
                    w_upd = w_pred * L
                    K = P_pr*self.c/S
                    f_up = f_pr + K*(y[t]-self.c*f_pr)
                    P_up = (1-K*self.c)*P_pr
                    cov = (1-K*self.c)*self.a[j]*P_prev[i]
                    w_ij[i,j] = w_upd
                    f_ij[i,j] = f_up
                    P_ij[i,j] = P_up
                    cov_ij[i,j] = cov
            total = w_ij.sum()
            xi[t] = w_ij/total
            gamma[t] = xi[t].sum(axis=0)
            # collapse
            for j in range(M):
                # フィルタ済み因子
                f_filt[t,j] = (w_ij[:,j]*f_ij[:,j]).sum()/w_ij[:,j].sum()
                # 分散
                P_filt[t,j] = (w_ij[:,j]*(P_ij[:,j]+(f_ij[:,j]-f_filt[t,j])**2)).sum()/w_ij[:,j].sum()
                # ラグ共分散
                P_lag[t,j] = (w_ij[:,j]*(cov_ij[:,j])).sum()/w_ij[:,j].sum()
            # 次準備
            f_prev = f_filt[t].copy()
            P_prev = P_filt[t].copy()
            w_prev = gamma[t].copy()
        return gamma, xi, f_filt, P_filt, P_lag

    def _maximization(self, y, gamma, xi, f_filt, P_filt, P_lag):
        T = len(y)
        M = self.M
        # π
        for i in range(M):
            for j in range(M):
                self.pi[i,j] = xi[1:,i,j].sum()/gamma[:-1,i].sum()
        # 観測 c, R
        num_c = 0; den_c=0; num_R=0
        for t in range(T):
            for j in range(M):
                num_c += gamma[t,j]*y[t]*f_filt[t,j]
                den_c += gamma[t,j]*(f_filt[t,j]**2+P_filt[t,j])
        self.c = num_c/den_c
        for t in range(T):
            for j in range(M):
                pred = self.c*f_filt[t,j]
                num_R += gamma[t,j]*((y[t]-pred)**2 + self.c**2*P_filt[t,j])
        self.R = num_R/gamma.sum()
        # 状態 μ, a, Q
        for j in range(M):
            w = gamma[1:,j]
            N_j = w.sum()
            S0 = (w*f_filt[1:,j]).sum(); S1 = (w*f_filt[:-1,j]).sum()
            S01= (w*(f_filt[1:,j]*f_filt[:-1,j]+P_lag[1:,j])).sum()
            S11= (w*(f_filt[:-1,j]**2+P_filt[:-1,j])).sum()
            Mmat = np.array([[N_j, S1],[S1, S11]])
            bvec = np.array([S0, S01])
            sol = np.linalg.solve(Mmat,bvec)
            self.mu[j], self.a[j] = sol
            # Q
            num_Q=0
            for t in range(1,T):
                v = f_filt[t,j] - self.mu[j] - self.a[j]*f_filt[t-1,j]
                num_Q += gamma[t,j]*(v**2 + P_filt[t,j] + self.a[j]**2*P_filt[t-1,j] - 2*self.a[j]*P_lag[t,j])
            self.Q[j] = num_Q/N_j

    def _log_likelihood(self, y, gamma):
        # 簡易対数尤度: −∑_t∑_j γ_tj log(γ_tj)
        return -np.sum(gamma*np.log(gamma+1e-16))


if __name__ == '__main__':
    # シミュレーション
    np.random.seed(1)
    T = 300
    true_pi = np.array([[0.9,0.1],[0.1,0.9]])
    true_mu = np.array([-0.5,0.5])
    true_a  = np.array([0.95,0.5])
    true_Q  = np.array([0.1,0.1])
    true_c  = 1.0; true_R=0.5
    s_true = np.zeros(T,dtype=int)
    y = np.zeros(T); f=np.zeros(T)
    for t in range(1,T):
        s_true[t] = np.random.choice(2,p=true_pi[s_true[t-1]])
        j=s_true[t]
        f[t] = true_mu[j]+true_a[j]*f[t-1]+np.sqrt(true_Q[j])*np.random.randn()
        y[t] = true_c*f[t]+np.sqrt(true_R)*np.random.randn()
    # 推定
    model = RegimeSwitchingDFM(n_regimes=2)
    gamma = model.fit(y)
    # プロット
    plt.figure(figsize=(10,5))
    plt.plot(s_true, label='True Regime', lw=1)
    plt.plot(gamma[:,1], label='Est Prob Regime1', lw=1)
    plt.xlabel('Time'); plt.legend(); plt.tight_layout(); plt.show()
```

```python
import numpy as np
from scipy.linalg import inv

class RegimeSwitchingMixedFreqDFM:
    """
    レジームスイッチング混合頻度DFM (多因子, AR(p)因子)
    EMアルゴリズムによる実装

    観測方程式:
      Y_t = C[s_t] F_t + ε_t,   ε_t~N(0,R[s_t])
    状態方程式:
      F_t = μ[s_t] + A[s_t] F_{t-1] + v_t,   v_t~N(0,Q[s_t])
    レジーム遷移:
      Pr(s_t=j|s_{t-1}=i)=π[i,j]

    混合頻度:
      Mariano–Murasawa (2003) による四半期→月次近似
    """
    def __init__(self, n_series, n_factors, lag_factors, n_regimes,
                 max_iter=100, tol=1e-4):
        self.N = n_series
        self.q = n_factors
        self.p = lag_factors
        self.M = n_regimes
        self.max_iter = max_iter
        self.tol = tol
        dim = 1 + self.q * self.p
        self.C = np.random.randn(self.M, self.N, dim)
        self.A = np.array([np.eye(dim) for _ in range(self.M)])
        self.mu = np.zeros((self.M, dim))
        self.Q = np.array([np.eye(dim) for _ in range(self.M)])
        self.R = np.array([np.eye(self.N) for _ in range(self.M)])
        self.pi = np.full((self.M, self.M), 1.0/self.M)

    def _augment_mixed_data(self, Y, freq_map):
        T_q, N = Y.shape
        # Mariano–Murasawa weights [1,2,3,2,1]/3 for quarterly→monthly
        weights = np.array([1,2,3,2,1]) / 3.0
        Y_aug, mask = [], []
        for t_q in range(T_q):
            quarterly = Y[t_q]
            # create five monthly observations around quarter t_q
            for m in range(5):
                y_month = np.copy(quarterly)
                # only quarterly series get weight, monthly unchanged
                for i, f in enumerate(freq_map):
                    if f == 'q':
                        y_month[i] = weights[m] * quarterly[i]
                Y_aug.append(y_month)
                # mask indicates which series observed this month
                mask.append([True if freq_map[i]=='m' else True for i in range(N)])
        return np.vstack(Y_aug), np.array(mask)

    def _kalman_filter(self, Y, mask):
        T = Y.shape[0]; dim = self.A.shape[1]
        f_prev = np.zeros((self.M,dim))
        P_prev = np.array([np.eye(dim)*10 for _ in range(self.M)])
        w_prev = np.full(self.M,1/self.M)
        gamma = np.zeros((T,self.M))
        xi    = np.zeros((T,self.M,self.M))
        f_filt= np.zeros((T,self.M,dim))
        P_filt= np.zeros((T,self.M,dim,dim))
        self.ll_obs = 0.0
        for t in range(T):
            w_ij = np.zeros((self.M,self.M))
            tmp_f= np.zeros((self.M,self.M,dim))
            tmp_P= np.zeros((self.M,self.M,dim,dim))
            for i in range(self.M):
                for j in range(self.M):
                    pr = w_prev[i]*self.pi[i,j]
                    f_pr = self.mu[j] + self.A[j]@f_prev[i]
                    P_pr = self.A[j]@P_prev[i]@self.A[j].T + self.Q[j]
                    idx=mask[t]; Hj=self.C[j][idx]
                    S=Hj@P_pr@Hj.T + self.R[j][np.ix_(idx,idx)]
                    invS=inv(S); eta=Y[t,idx]-Hj@f_pr
                    L=np.exp(-0.5*eta.T@invS@eta)/np.sqrt(((2*np.pi)**len(eta))*np.linalg.det(S))
                    w_ij[i,j]=pr*L
                    K=P_pr@Hj.T@invS; tmp_f[i,j]=f_pr + K@eta
                    tmp_P[i,j]=P_pr - K@Hj@P_pr
            tot = w_ij.sum()
            # 対数尤度に積算
            self.ll_obs += np.log(tot + 1e-16)
            xi[t] = w_ij/tot
            gamma[t] = xi[t].sum(axis=0)
            for j in range(self.M):
                wj = w_ij[:,j]
                s  = wj.sum()
                f_filt[t,j] = (wj[:,None]*tmp_f[:,j]).sum(axis=0)/s
                P_filt[t,j] = (wj[:,None,None]*(tmp_P[:,j] + np.einsum('ti,tj->tij', tmp_f[:,j]-f_filt[t,j], tmp_f[:,j]-f_filt[t,j]))).sum(axis=0)/s
            f_prev = f_filt[t]
            P_prev = P_filt[t]
            w_prev = gamma[t]
        return gamma, xi, f_filt, P_filt, None

    def _kalman_smoother(self, gamma, xi, f_filt, P_filt, _):
        T=f_filt.shape[0]; dim=f_filt.shape[2]
        f_s=np.zeros_like(f_filt); P_s=np.zeros_like(P_filt);
        Plag=np.zeros((T,self.M,dim,dim))
        f_s[-1]=f_filt[-1]; P_s[-1]=P_filt[-1]
        for t in range(T-2,-1,-1):
            for j in range(self.M):
                Pj=P_filt[t,j]; Ppr=self.A[j]@Pj@self.A[j].T+self.Q[j]
                J=Pj@self.A[j].T@inv(Ppr)
                f_s[t,j]=f_filt[t,j]+J@(f_s[t+1,j]-self.mu[j]-self.A[j]@f_filt[t,j])
                P_s[t,j]=Pj+J@(P_s[t+1,j]-Ppr)@J.T
                Plag[t,j]=J@P_s[t+1,j]
        return f_s,P_s,Plag

    def _expectation(self, Y, mask):
        gamma,xi,Ff,Pf,_=self._kalman_filter(Y,mask)
        Fs,Ps,Plag=self._kalman_smoother(gamma,xi,Ff,Pf,None)
        return gamma,xi,Fs,Ps,Plag

    def _maximization(self, Y, mask, gamma, xi, Fs, Ps, Plag):
        T=Fs.shape[0]; dim=Fs.shape[2]
        for i in range(self.M):
            denom=gamma[:-1,i].sum()
            for j in range(self.M): self.pi[i,j]=xi[1:,i,j].sum()/denom
        for j in range(self.M):
            numC=np.zeros((self.N,dim)); denC=np.zeros((dim,dim)); numR=0
            for t in range(T):
                idx=mask[t]; y=Y[t,idx]; ft=Fs[t,j]; Pf=Ps[t,j]
                numC[idx]+=gamma[t,j]*np.outer(y,ft)
                denC+=gamma[t,j]*(np.outer(ft,ft)+Pf)
                pred=numC[idx]@inv(denC)@ft; numR+=gamma[t,j]*np.sum((y-pred)**2)
            # 正則化を追加して解が発散しないように
            denC_reg = denC + np.eye(dim)*1e-6
            self.C[j] = numC @ inv(denC_reg); self.R[j]=np.eye(self.N)*(numR/gamma[:,j].sum())
        for j in range(self.M):
            numA=np.zeros((dim,dim)); denA=np.zeros((dim,dim)); sumF=np.zeros(dim)
            for t in range(1,T):
                numA+=gamma[t,j]*(Plag[t-1,j]+np.outer(Fs[t,j],Fs[t-1,j]))
                denA+=gamma[t,j]*(Ps[t-1,j]+np.outer(Fs[t-1,j],Fs[t-1,j]))
                sumF+=gamma[t,j]*Fs[t,j]
            # 正則化
            denA_reg = denA + np.eye(dim)*1e-6
            self.A[j] = numA @ inv(denA_reg)
            self.mu[j]=(sumF-self.A[j]@(gamma[:-1,j][:,None]*Fs[:-1,j]).sum(axis=0))/gamma[1:,j].sum()
            numQ=0
            for t in range(1,T):
                diff=Fs[t,j]-self.mu[j]-self.A[j]@Fs[t-1,j]
                numQ+=gamma[t,j]*(np.outer(diff,diff)+Ps[t,j]+self.A[j]@Ps[t-1,j]@self.A[j].T-Plag[t-1,j]@self.A[j].T-self.A[j]@Plag[t-1,j].T)
            self.Q[j]=numQ/gamma[1:,j].sum()

    def _log_likelihood(self, Y, mask, gamma, xi, Fs, Ps, Plag):
        # フィルタで蓄積した対数尤度を返す
        return self.ll_obs

    def fit(self, Y, freq_map):
        Y_aug,mask=self._augment_mixed_data(Y,freq_map)
        ll_old=-np.inf
        for _ in range(self.max_iter):
            gamma,xi,Fs,Ps,Plag=self._expectation(Y_aug,mask)
            self._maximization(Y_aug,mask,gamma,xi,Fs,Ps,Plag)
            ll=self._log_likelihood(Y_aug,mask,gamma,xi,Fs,Ps,Plag)
            if abs(ll-ll_old)<self.tol: break
            ll_old=ll
        return gamma

    def nowcast(self, Y_new, freq_map):
        return None

if __name__ == '__main__':
    import matplotlib.pyplot as plt
    np.random.seed(42)
    # テストデータ生成: 2系列 (月次と四半期)
    T_q = 50; N = 2; q = 1; p = 1; M = 2
    # パラメータ
    true_pi = np.array([[0.9,0.1],[0.1,0.9]])
    true_mu = np.array([-0.3,0.7])
    true_a  = np.array([0.8,0.3])
    true_Q  = np.array([0.2,0.2])
    true_c  = np.array([1.0,0.5])
    true_R  = np.array([0.5,0.5])
    # 四半期→月次長さ
    T = T_q*3
    s_true = np.zeros(T, dtype=int)
    F = np.zeros(T)
    # レジーム生成
    for t in range(1,T):
        s_true[t] = np.random.choice(M,p=true_pi[s_true[t-1]])
        j=s_true[t]
        F[t] = true_mu[j] + true_a[j]*F[t-1] + np.sqrt(true_Q[j])*np.random.randn()
    # 観測データ (Y_q x N)
    Y_q = np.zeros((T_q,N))
    for t_q in range(T_q):
        idx = 3*t_q+np.array([2,2])
        Y_q[t_q,0] = true_c[0]*F[idx[0]] + np.sqrt(true_R[0])*np.random.randn()
        Y_q[t_q,1] = (true_c[1]*F[idx] + np.sqrt(true_R[1])*np.random.randn(2)).mean()
    freq_map = ['q','m']
    # 推定
    model = RegimeSwitchingMixedFreqDFM(N,q,p,M,max_iter=50)
    gamma = model.fit(Y_q, freq_map)
    # プロット
    plt.figure(figsize=(10,4))
    # s_true は月次長で、gamma は既に月次長 (T_q*3) なので両方 150 長さで一致
    plt.plot(s_true, label='True Regime', alpha=0.5)
    plt.plot(gamma[:,1], label='Est Prob s=1', lw=2)
    plt.xlabel('Month Index')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class RegimeSwitchingDFM:
    """
    レジーム・スイッチング多因子DFM（AR(1)因子）を
    EMアルゴリズムで推定するクラス。

    パラメータ:
      - n_regimes: レジーム数 M
      - n_factors: 潜在因子次元 k
      - obs_dim:   観測次元 p
      - max_iter: 最大EMイテレーション回数
      - tol:      収束判定用トレランス
    """
    def __init__(self, n_regimes=2, n_factors=2, obs_dim=3, max_iter=100, tol=1e-4):
        self.M = n_regimes
        self.k = n_factors
        self.p = obs_dim
        self.max_iter = max_iter
        self.tol = tol

        # ——— パラメータ初期化 ———
        # レジーム遷移行列 π_{ij}
        self.pi = np.full((self.M, self.M), 1/self.M)

        # AR(1)モデル: f_t = μ_j + A_j f_{t-1} + v_t
        # μ_j: kベクトル, A_j: k×k 行列, Q_j: k×k 共分散
        self.mu = np.zeros((self.M, self.k))
        self.A  = np.array([np.eye(self.k)*0.5 for _ in range(self.M)])  # 初期: 0.5*I
        self.Q  = np.array([np.eye(self.k)     for _ in range(self.M)])  # 初期: I

        # 観測方程式: y_t = C f_t + ε_t,  ε_t ~ N(0,R)
        # C: p×k 行列, R: p×p 共分散
        self.C = np.ones((self.p, self.k))
        self.R = np.eye(self.p)

    def fit(self, Y):
        """
        観測データ Y を与えて EM 推定を実行し、
        事後確率 gamma (T×M) を返す。
        Y: shape (T, p)
        """
        T = Y.shape[0]
        # π のよい初期値
        self.pi[:] = 0.9*np.eye(self.M) + 0.1*(np.ones((self.M,self.M)) - np.eye(self.M))/(self.M-1)
        # μ_j を [-1,1] に散らす
        for j in range(self.M):
            self.mu[j] = np.linspace(-1,1,self.k)

        # 事後確率の初期化
        gamma = np.full((T,self.M), 1/self.M)
        xi    = np.full((T,self.M,self.M), 1/(self.M**2))

        ll_old = -np.inf
        for it in range(self.max_iter):
            # Eステップ
            gamma, xi, f_filt, P_filt, P_lag = self._expectation(Y, gamma)
            # Mステップ
            self._maximization(Y, gamma, xi, f_filt, P_filt, P_lag)
            # 対数尤度（簡易）
            ll = -np.sum(gamma * np.log(gamma + 1e-16))
            if abs(ll - ll_old) < self.tol:
                break
            ll_old = ll

        return gamma

    def _expectation(self, Y, gamma_prev):
        T, p = Y.shape
        M, k  = self.M, self.k

        # フィルタ結果格納
        f_filt = np.zeros((T, M, k))
        P_filt = np.zeros((T, M, k, k))
        P_lag  = np.zeros((T, M, k, k))
        xi      = np.zeros((T, M, M))
        gamma   = np.zeros((T, M))

        # 前時点の予測分布 (レジーム i ごと)
        f_prev = np.zeros((M, k))
        P_prev = np.array([np.eye(k)*10 for _ in range(M)])  # 大きめ初期分散
        w_prev = np.full(M, 1/M)

        for t in range(T):
            # (i→j) の重み・フィルタ結果を一時格納
            w_ij   = np.zeros((M, M))
            f_ij   = np.zeros((M, M, k))
            P_ij   = np.zeros((M, M, k, k))
            cov_ij = np.zeros((M, M, k, k))

            for i in range(M):
                for j in range(M):
                    # 1) 予測ステップ
                    w_pred = w_prev[i] * self.pi[i,j]
                    f_pr   = self.mu[j] + self.A[j] @ f_prev[i]                # k
                    P_pr   = self.A[j] @ P_prev[i] @ self.A[j].T + self.Q[j]   # k×k

                    # 2) 観測アップデート
                    S      = self.C @ P_pr @ self.C.T + self.R                 # p×p
                    innov  = Y[t] - (self.C @ f_pr)                            # p
                    # 多変量正規尤度
                    L      = (np.exp(-0.5 * innov.T @ np.linalg.solve(S, innov))
                              / np.sqrt((2*np.pi)**p * np.linalg.det(S)))

                    # 3) 更新
                    w_upd  = w_pred * L
                    K      = P_pr @ self.C.T @ np.linalg.inv(S)                # k×p
                    f_up   = f_pr + K @ innov                                  # k
                    P_up   = P_pr - K @ self.C @ P_pr                          # k×k
                    # ラグ共分散
                    cov    = (np.eye(k) - K @ self.C) @ self.A[j] @ P_prev[i]   # k×k

                    w_ij[i,j]       = w_upd
                    f_ij[i,j]       = f_up
                    P_ij[i,j]       = P_up
                    cov_ij[i,j]     = cov

            # 正規化して xi, gamma
            total = w_ij.sum()
            xi[t]    = w_ij / total
            gamma[t] = xi[t].sum(axis=0)

            # collapse によるフィルタ済み平均・分散・ラグ共分散
            for j in range(M):
                # 平均
                f_filt[t,j] = (w_ij[:,j,None] * f_ij[:,j]).sum(axis=0) / w_ij[:,j].sum()
                # 分散
                diff = f_ij[:,j] - f_filt[t,j]
                P_filt[t,j] = (w_ij[:,j,None,None] *
                               (P_ij[:,j] + diff[:,:,None]*diff[:,None,:])
                              ).sum(axis=(0,1)) / w_ij[:,j].sum()
                # ラグ共分散
                P_lag[t,j] = (w_ij[:,j,None,None] * cov_ij[:,j]).sum(axis=(0,1)) / w_ij[:,j].sum()

            # 次ステップ準備
            f_prev = f_filt[t].copy()
            P_prev = P_filt[t].copy()
            w_prev = gamma[t].copy()

        return gamma, xi, f_filt, P_filt, P_lag

    def _maximization(self, Y, gamma, xi, f_filt, P_filt, P_lag):
        T = Y.shape[0]
        M, k, p = self.M, self.k, self.p

        # —— 1) π_{ij} の更新
        for i in range(M):
            for j in range(M):
                self.pi[i,j] = xi[1:,i,j].sum() / gamma[:-1,i].sum()

        # —— 2) 観測行列 C とノイズ共分散 R の更新
        Sum_YF = np.zeros((p,k))
        Sum_FF = np.zeros((k,k))
        Sum_YY = np.zeros((p,p))
        for t in range(T):
            for j in range(M):
                w = gamma[t,j]
                f = f_filt[t,j]
                P = P_filt[t,j]
                Sum_YF += w * np.outer(Y[t], f)
                Sum_FF += w * (P + np.outer(f,f))
                Sum_YY += w * np.outer(Y[t], Y[t])
        # 最小二乗解として
        self.C = Sum_YF @ np.linalg.inv(Sum_FF)
        # R は残差共分散
        self.R = (Sum_YY - self.C @ Sum_YF.T) / gamma.sum()

        # —— 3) 状態モデル μ_j, A_j, Q_j の更新
        for j in range(M):
            w = gamma[1:,j]                  # (T-1,)
            N = w.sum()

            # データ行列化
            F_t  = f_filt[1:,j]              # (T-1,k)
            F_t1 = f_filt[:-1,j]             # (T-1,k)
            X_j  = np.concatenate([np.ones((T-1,1)), F_t1], axis=1)  # (T-1, k+1)
            Y_j  = F_t                                      # (T-1, k)

            WX   = X_j * w[:,None]           # 重み付き
            Sxx  = WX.T @ X_j               # (k+1,k+1)
            Sxy  = WX.T @ Y_j               # (k+1,k)
            beta = np.linalg.solve(Sxx, Sxy)  # (k+1,k)

            # μ_j と A_j
            self.mu[j] = beta[0]
            self.A[j]  = beta[1:]

            # Q_j の更新
            Q_num = np.zeros((k,k))
            for t in range(1,T):
                resid = f_filt[t,j] - self.mu[j] - self.A[j] @ f_filt[t-1,j]
                Q_num += gamma[t,j] * (
                    np.outer(resid, resid)
                    + P_filt[t,j]
                    + self.A[j] @ P_filt[t-1,j] @ self.A[j].T
                    - self.A[j] @ P_lag[t,j].T
                    - P_lag[t,j] @ self.A[j].T
                )
            self.Q[j] = Q_num / N

    def _log_likelihood(self, gamma):
        return -np.sum(gamma * np.log(gamma + 1e-16))


if __name__ == '__main__':
    # ——— シミュレーション例 ———
    np.random.seed(1)
    T = 300
    M, k, p = 2, 2, 3

    # 真のパラメータ設定
    true_pi = np.array([[0.9,0.1],[0.1,0.9]])
    true_mu = np.array([[-0.5, 0.2],[0.5,-0.3]])      # (M,k)
    true_A  = np.array([[[0.9,0.1],[0.0,0.8]],
                        [[0.2,0.0],[0.0,0.5]]])      # (M,k,k)
    true_Q  = np.array([np.eye(k)*0.1, np.eye(k)*0.1])
    true_C  = np.random.randn(p, k)
    true_R  = np.eye(p)*0.5

    s_true = np.zeros(T, dtype=int)
    f = np.zeros((T, k))
    Y = np.zeros((T, p))
    for t in range(1, T):
        s_true[t] = np.random.choice(M, p=true_pi[s_true[t-1]])
        j = s_true[t]
        f[t] = true_mu[j] + true_A[j] @ f[t-1] + np.random.multivariate_normal(np.zeros(k), true_Q[j])
        Y[t] = true_C @ f[t] + np.random.multivariate_normal(np.zeros(p), true_R)

    # モデル推定
    model = RegimeSwitchingDFM(n_regimes=M, n_factors=k, obs_dim=p)
    gamma = model.fit(Y)

    # 結果プロット
    plt.figure(figsize=(10,4))
    plt.plot(s_true,     label="True Regime", lw=1)
    plt.plot(gamma[:,1], label="Est Prob Regime 1", lw=1)
    plt.xlabel("Time"); plt.legend(); plt.tight_layout(); plt.show()
```

```python
"""
Fully‑fledged EM learner for a 2‑regime, 2‑factor dynamic factor model
(3 observed variables).

Key features inspired by the `ssm` implementation:
* Hamilton filter / Kim smoother in **log‑space** (prevents underflow)
* PSD‑clip of every covariance to guarantee positive semidefiniteness
* Sticky Dirichlet prior on transition matrix (encourages regime persistence)
* Closed‑form M‑step for A, Q, R, C, and P

Run the file directly to execute a demo simulation and plot the smoothed
regime probabilities v.s. the true regime sequence.
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from scipy.special import logsumexp

# ---------------------------------------------------------------------
# 0. Utilities
# ---------------------------------------------------------------------

def psd_clip(S, eps: float = 1e-6):
    """Force covariance matrix S to be PSD by clipping eigenvalues."""
    vals, vecs = np.linalg.eigh(S)
    return vecs @ np.diag(np.clip(vals, eps, None)) @ vecs.T

# ---------------------------------------------------------------------
# 1. Simulate synthetic SLDS data (same set‑up as the conversation)
# ---------------------------------------------------------------------

def simulate_slds(T=300, seed=42):
    rng = np.random.default_rng(seed)
    n_f, n_y, M = 2, 3, 2

    P_true = np.array([[0.9, 0.1],
                       [0.1, 0.9]])

    A_true = np.array([[[0.8, 0.0],
                        [0.0, 0.7]],
                       [[0.5, 0.0],
                        [0.0, 0.4]]])

    Q_true = np.array([0.2 * np.eye(n_f) for _ in range(M)])

    C_true = np.array([[1., 0.],
                       [0., 1.],
                       [1., 1.]])

    R_true = np.array([0.1 * np.eye(n_y) for _ in range(M)])

    s = np.zeros(T, dtype=int)
    f = np.zeros((T, n_f))
    y = np.zeros((T, n_y))

    s[0] = rng.integers(0, M)
    f[0] = rng.multivariate_normal(np.zeros(n_f), np.eye(n_f))
    y[0] = C_true @ f[0] + rng.multivariate_normal(np.zeros(n_y), R_true[s[0]])

    for t in range(1, T):
        s[t] = rng.choice(M, p=P_true[s[t - 1]])
        f[t] = (A_true[s[t]] @ f[t - 1] +
                 rng.multivariate_normal(np.zeros(n_f), Q_true[s[t]]))
        y[t] = C_true @ f[t] + rng.multivariate_normal(np.zeros(n_y), R_true[s[t]])

    return y, f, s, dict(P=P_true, A=A_true, Q=Q_true, R=R_true, C=C_true)

# ---------------------------------------------------------------------
# 2. Kalman primitives (predict / update)
# ---------------------------------------------------------------------

def kf_predict(m_prev, P_prev, A, Q):
    m_pred = A @ m_prev
    P_pred = A @ P_prev @ A.T + Q
    return m_pred, psd_clip(P_pred)


def kf_update(y, m_pred, P_pred, C, R):
    S = psd_clip(C @ P_pred @ C.T + R)
    K = P_pred @ C.T @ np.linalg.inv(S)
    innov = y - C @ m_pred
    m_filt = m_pred + K @ innov
    P_filt = psd_clip(P_pred - K @ C @ P_pred)
    ll = multivariate_normal.logpdf(innov, mean=np.zeros(len(innov)), cov=S, allow_singular=True)
    return m_filt, P_filt, ll

# ---------------------------------------------------------------------
# 3. Log‑space Hamilton filter
# ---------------------------------------------------------------------

def log_hamilton_filter(Y, par):
    T, n_y = Y.shape
    M, n_f = par['A'].shape[:2]

    log_alpha = np.full((T, M), -np.inf)
    m_filt = np.zeros((T, M, n_f))
    P_filt = np.zeros((T, M, n_f, n_f))
    ll_t = np.zeros((T, M))

    # t = 0
    for j in range(M):
        mp, Pp = kf_predict(np.zeros(n_f), np.eye(n_f), par['A'][j], par['Q'][j])
        m_filt[0, j], P_filt[0, j], ll = kf_update(Y[0], mp, Pp, par['C'], par['R'][j])
        ll_t[0, j] = ll
    log_alpha[0] = np.log(1 / M) + ll_t[0]
    log_alpha[0] -= logsumexp(log_alpha[0])

    logP = np.log(par['P'] + 1e-12)

    for t in range(1, T):
        for j in range(M):
            # mixture prediction
            tmp = log_alpha[t - 1] + logP[:, j]
            log_mix_prob = tmp - logsumexp(tmp)
            mu_mix = np.zeros(n_f)
            P_mix = np.zeros((n_f, n_f))
            for i in range(M):
                mp, Pp = kf_predict(m_filt[t - 1, i], P_filt[t - 1, i], par['A'][j], par['Q'][j])
                weight = np.exp(log_mix_prob[i])
                mu_mix += weight * mp
            for i in range(M):
                mp, Pp = kf_predict(m_filt[t - 1, i], P_filt[t - 1, i], par['A'][j], par['Q'][j])
                weight = np.exp(log_mix_prob[i])
                diff = mp - mu_mix
                P_mix += weight * (Pp + np.outer(diff, diff))
            m_filt[t, j], P_filt[t, j], ll = kf_update(Y[t], mu_mix, P_mix, par['C'], par['R'][j])
            ll_t[t, j] = ll
        log_alpha[t] = logsumexp(log_alpha[t - 1][:, None] + logP, axis=0) + ll_t[t]
        log_alpha[t] -= logsumexp(log_alpha[t])

    return log_alpha, m_filt, P_filt

# ---------------------------------------------------------------------
# 4. Log‑space Kim smoother & pairwise xi
# ---------------------------------------------------------------------

def log_kim_smoother(log_alpha, par, ll_t):
    T, M = log_alpha.shape
    log_beta = np.zeros_like(log_alpha)
    log_gamma = np.zeros_like(log_alpha)
    log_gamma[-1] = log_alpha[-1]

    logP = np.log(par['P'] + 1e-12)
    log_xi = np.zeros((T - 1, M, M))

    for t in range(T - 2, -1, -1):
        tmp = log_beta[t + 1] + ll_t[t + 1] + logP.T
        log_beta[t] = logsumexp(tmp, axis=1)
        log_beta[t] -= logsumexp(log_beta[t])
        log_gamma[t] = log_alpha[t] + log_beta[t]
        log_gamma[t] -= logsumexp(log_gamma[t])
        log_xi[t] = (log_alpha[t][:, None] + logP + ll_t[t + 1] + log_beta[t + 1])
        log_xi[t] -= logsumexp(log_xi[t])

    gamma = np.exp(log_gamma)
    xi = np.exp(log_xi)
    return gamma, xi

# ---------------------------------------------------------------------
# 5. M‑step with Sticky Dirichlet prior
# ---------------------------------------------------------------------

def m_step(Y, par, gamma, xi, m_filt, P_filt, kappa=5.0):
    T, n_y = Y.shape
    M, n_f = par['A'].shape[:2]

    # Update P with Dirichlet(κ)
    par['P'] = xi.sum(0) + kappa * np.eye(M)
    par['P'] /= par['P'].sum(axis=1, keepdims=True)

    # Update C
    numC = np.zeros((n_y, n_f)); denC = np.zeros((n_f, n_f))
    for t in range(T):
        for j in range(M):
            numC += gamma[t, j] * np.outer(Y[t], m_filt[t, j])
            denC += gamma[t, j] * (P_filt[t, j] + np.outer(m_filt[t, j], m_filt[t, j]))
    par['C'] = numC @ np.linalg.inv(denC + 1e-12 * np.eye(n_f))

    # Update R, A, Q regime‑wise
    for j in range(M):
        # R
        numR = np.zeros((n_y, n_y)); denR = gamma[:, j].sum()
        for t in range(T):
            res = Y[t] - par['C'] @ m_filt[t, j]
            numR += gamma[t, j] * (np.outer(res, res) + par['C'] @ P_filt[t, j] @ par['C'].T)
        par['R'][j] = psd_clip(numR / max(denR, 1e-12))

        # A & Q
        numA = np.zeros((n_f, n_f)); denA = np.zeros((n_f, n_f))
        for t in range(1, T):
            numA += gamma[t, j] * np.outer(m_filt[t, j], m_filt[t - 1, j])
            denA += gamma[t, j] * (P_filt[t - 1, j] + np.outer(m_filt[t - 1, j], m_filt[t - 1, j]))
        par['A'][j] = numA @ np.linalg.inv(denA + 1e-12 * np.eye(n_f))
        # Q
        numQ = np.zeros((n_f, n_f)); denQ = 0.0
        for t in range(1, T):
            diff = m_filt[t, j] - par['A'][j] @ m_filt[t - 1, j]
            numQ += gamma[t, j] * (np.outer(diff, diff) + P_filt[t, j])
            denQ += gamma[t, j]
        par['Q'][j] = psd_clip(numQ / max(denQ, 1e-12))

    return par

# ---------------------------------------------------------------------
# 6. EM training function
# ---------------------------------------------------------------------

def train_em(Y, n_iter=15, seed=0):
    np.random.seed(seed)
    n_y = Y.shape[1]; n_f = 2; M = 2

    par = {
        "P": np.array([[0.8, 0.2], [0.2, 0.8]]),
        "A": np.array([np.eye(n_f) * 0.6, np.eye(n_f) * 0.4]),
        "Q": np.array([np.eye(n_f) * 0.3, np.eye(n_f) * 0.3]),
        "R": np.array([np.eye(n_y) * 0.5, np.eye(n_y) * 0.5]),
        "C": np.array([[1., 0.], [0., 1.], [1., 1.]])
    }

    loglik_trace = []
    for it in range(n_iter):
        log_alpha, m_filt, P_filt = log_hamilton_filter(Y, par)[:3]
        gamma, xi = log_kim_smoother(log_alpha, par, log_hamilton_filter(Y, par)[2])
        par = m_step(Y, par, gamma, xi, m_filt, P_filt)
        loglik_trace.append(np.sum(logsumexp(log_alpha, axis=1)))
    return par, gamma, loglik_trace

# ---------------------------------------------------------------------
# 7. Demo
# ---------------------------------------------------------------------

if __name__ == "__main__":
    Y, f_true, s_true, true_par = simulate_slds(T=300, seed=7)
    par_hat, gamma, ll = train_em(Y, n_iter=20)

    # Evaluation
    state_est = (gamma[:, 1] > 0.5).astype(int)
    acc = (state_est == s_true).mean()
    print("True P:\n", true_par['P'])
    print("Estimated P:\n", np.round(par_hat['P'], 3))
    print(f"State accuracy: {acc:.3f}")

    t = np.arange(len(Y))
    plt.figure(figsize=(9, 3))
    plt.plot(t, gamma[:, 1], label="Smoothed Prob(Regime 1)")
    plt.plot(t, s_true, "k:", label="True Regime (0/1)")
    plt.legend(); plt.tight_layout(); plt.show()
```

```python
"""
Fully‑fledged EM learner for a 2‑regime, 2‑factor dynamic factor model
(3 observed variables).

Improvements (reflecting user‑reported error fix):
* Log‑space Hamilton filter / Kim smoother now **broadcast‑safe**
* `log_hamilton_filter` returns `(log_alpha, m_filt, P_filt, ll_t)` so
  the same likelihood array is reused (no double call)
* Kim smoother recursion uses a fully vectorised formula that avoids
  shape mismatch (`log_beta[t]` now computed from a (M×M) matrix)
* Train loop updated accordingly and no longer recomputes the filter
* PSD clip and sticky Dirichlet prior retained

Run with `python switching_dfm_full.py` to verify.
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from scipy.special import logsumexp

# ---------------------------------------------------------------------
# Utilities
# ---------------------------------------------------------------------

def psd_clip(S, eps: float = 1e-6):
    vals, vecs = np.linalg.eigh(S)
    return vecs @ np.diag(np.clip(vals, eps, None)) @ vecs.T

# ---------------------------------------------------------------------
# Synthetic data generator (unchanged)
# ---------------------------------------------------------------------

def simulate_slds(T=300, seed=42):
    rng = np.random.default_rng(seed)
    n_f, n_y, M = 2, 3, 2
    P_true = np.array([[0.9, 0.1], [0.1, 0.9]])
    A_true = np.array([[[0.8, 0.], [0., 0.7]], [[0.5, 0.], [0., 0.4]]])
    Q_true = np.array([0.2 * np.eye(n_f) for _ in range(M)])
    C_true = np.array([[1., 0.], [0., 1.], [1., 1.]])
    R_true = np.array([0.1 * np.eye(n_y) for _ in range(M)])
    s = np.zeros(T, int); f = np.zeros((T, n_f)); y = np.zeros((T, n_y))
    s[0] = rng.integers(0, M)
    f[0] = rng.multivariate_normal(np.zeros(n_f), np.eye(n_f))
    y[0] = C_true @ f[0] + rng.multivariate_normal(np.zeros(n_y), R_true[s[0]])
    for t in range(1, T):
        s[t] = rng.choice(M, p=P_true[s[t-1]])
        f[t] = A_true[s[t]] @ f[t-1] + rng.multivariate_normal(np.zeros(n_f), Q_true[s[t]])
        y[t] = C_true @ f[t] + rng.multivariate_normal(np.zeros(n_y), R_true[s[t]])
    return y, f, s, dict(P=P_true, A=A_true, Q=Q_true, R=R_true, C=C_true)

# ---------------------------------------------------------------------
# Kalman predict/update
# ---------------------------------------------------------------------

def kf_predict(m, P, A, Q):
    mp = A @ m
    Pp = psd_clip(A @ P @ A.T + Q)
    return mp, Pp

def kf_update(y, mp, Pp, C, R):
    S = psd_clip(C @ Pp @ C.T + R)
    K = Pp @ C.T @ np.linalg.inv(S)
    innov = y - C @ mp
    mf = mp + K @ innov
    Pf = psd_clip(Pp - K @ C @ Pp)
    ll = multivariate_normal.logpdf(innov, mean=np.zeros(len(innov)), cov=S, allow_singular=True)
    return mf, Pf, ll

# ---------------------------------------------------------------------
# Log‑space Hamilton filter
# ---------------------------------------------------------------------

def log_hamilton_filter(Y, par):
    T, n_y = Y.shape; M, n_f = par['A'].shape[:2]
    log_alpha = np.full((T, M), -np.inf)
    mf = np.zeros((T, M, n_f)); Pf = np.zeros((T, M, n_f, n_f))
    ll_t = np.zeros((T, M))
    # t=0
    for j in range(M):
        mp, Pp = kf_predict(np.zeros(n_f), np.eye(n_f), par['A'][j], par['Q'][j])
        mf[0, j], Pf[0, j], ll_t[0, j] = kf_update(Y[0], mp, Pp, par['C'], par['R'][j])
    log_alpha[0] = np.log(1 / M) + ll_t[0]
    log_alpha[0] -= logsumexp(log_alpha[0])
    logP = np.log(par['P'] + 1e-12)
    # recursion
    for t in range(1, T):
        for j in range(M):
            # compute in log, sum over previous i
            log_mix = log_alpha[t-1][:, None] + logP[:, [j]]  # M×1
            log_mix -= logsumexp(log_mix)                     # log weights
            mu_mix = np.zeros(n_f); Sigma_mix = np.zeros((n_f, n_f))
            for i in range(M):
                mp, Pp = kf_predict(mf[t-1, i], Pf[t-1, i], par['A'][j], par['Q'][j])
                wi = np.exp(log_mix[i, 0])
                mu_mix += wi * mp
            for i in range(M):
                mp, Pp = kf_predict(mf[t-1, i], Pf[t-1, i], par['A'][j], par['Q'][j])
                wi = np.exp(log_mix[i, 0]); diff = mp - mu_mix
                Sigma_mix += wi * (Pp + np.outer(diff, diff))
            mf[t, j], Pf[t, j], ll_t[t, j] = kf_update(Y[t], mu_mix, Sigma_mix, par['C'], par['R'][j])
        log_alpha[t] = logsumexp(log_alpha[t-1][:, None] + logP, axis=0) + ll_t[t]
        log_alpha[t] -= logsumexp(log_alpha[t])
    return log_alpha, mf, Pf, ll_t

# ---------------------------------------------------------------------
# Log‑space Kim smoother
# ---------------------------------------------------------------------

def log_kim_smoother(log_alpha, par, ll_t):
    T, M = log_alpha.shape
    log_beta = np.zeros_like(log_alpha)
    log_gamma = np.copy(log_alpha)  # will be corrected in loop
    logP = np.log(par['P'] + 1e-12)
    log_xi = np.zeros((T - 1, M, M))
    for t in range(T - 2, -1, -1):
        # log_beta_t(i) = logsum_j [ logP_ij + ll_{t+1}(j) + log_beta_{t+1}(j) ]
        tmp = logP + ll_t[t + 1] + log_beta[t + 1]
        log_beta[t] = logsumexp(tmp, axis=1)
        log_beta[t] -= logsumexp(log_beta[t])
        log_gamma[t] = log_alpha[t] + log_beta[t]
        log_gamma[t] -= logsumexp(log_gamma[t])
        log_xi[t] = (log_alpha[t][:, None] + logP + ll_t[t + 1] + log_beta[t + 1])
        log_xi[t] -= logsumexp(log_xi[t])
    return np.exp(log_gamma), np.exp(log_xi)

# ---------------------------------------------------------------------
# M‑step (sticky prior)
# ---------------------------------------------------------------------

def m_step(Y, par, gamma, xi, mf, Pf, kappa=5.0):
    T, n_y = Y.shape; M, n_f = par['A'].shape[:2]
    par['P'] = xi.sum(0) + kappa * np.eye(M)
    par['P'] /= par['P'].sum(axis=1, keepdims=True)
    numC = np.zeros((n_y, n_f)); denC = np.zeros((n_f, n_f))
    for t in range(T):
        for j in range(M):
            numC += gamma[t, j] * np.outer(Y[t], mf[t, j])
            denC += gamma[t, j] * (Pf[t, j] + np.outer(mf[t, j], mf[t, j]))
    par['C'] = numC @ np.linalg.inv(denC + 1e-12 * np.eye(n_f))
    for j in range(M):
        numR = np.zeros((n_y, n_y)); denR = gamma[:, j].sum()
        for t in range(T):
            res = Y[t] - par['C'] @ mf[t, j]
            numR += gamma[t, j] * (np.outer(res, res) + par['C'] @ Pf[t, j] @ par['C'].T)
        par['R'][j] = psd_clip(numR / max(denR, 1e-12))
        numA = np.zeros((n_f, n_f)); denA = np.zeros((n_f, n_f))
        for t in range(1, T):
            numA += gamma[t, j] * np.outer(mf[t, j], mf[t - 1, j])
            denA += gamma[t, j] * (Pf[t - 1, j] + np.outer(mf[t - 1, j], mf[t - 1, j]))
        par['A'][j] = numA @ np.linalg.inv(denA + 1e-12 * np.eye(n_f))
        numQ = np.zeros((n_f, n_f)); denQ = 0.0
        for t in range(1, T):
            diff = mf[t, j] - par['A'][j] @ mf[t - 1, j]
            numQ += gamma[t, j] * (np.outer(diff, diff) + Pf[t, j])
            denQ += gamma[t, j]
        par['Q'][j] = psd_clip(numQ / max(denQ, 1e-12))
    return par

# ---------------------------------------------------------------------
# EM training loop
# ---------------------------------------------------------------------

def train_em(Y, n_iter=15, seed=0):
    np.random.seed(seed)
    n_y = Y.shape[1]; n_f = 2; M = 2
    par = {
        "P": np.array([[0.8, 0.2], [0.2, 0.8]]),
        "A": np.array([np.eye(n_f) * 0.6, np.eye(n_f) * 0.4]),
        "Q": np.array([np.eye(n_f) * 0.3, np.eye(n_f) * 0.3]),
        "R": np.array([np.eye(n_y) * 0.5, np.eye(n_y) * 0.5]),
        "C": np.array([[1., 0.], [0., 1.], [1., 1.]])
    }
    loglik_trace = []
    for _ in range(n_iter):
        log_alpha, mf, Pf, ll_t = log_hamilton_filter(Y, par)
        gamma, xi = log_kim_smoother(log_alpha, par, ll_t)
        par = m_step(Y, par, gamma, xi, mf, Pf)
        loglik_trace.append(np.sum(logsumexp(log_alpha, axis=1)))
    return par, gamma, loglik_trace

# ---------------------------------------------------------------------
# Demo
# ---------------------------------------------------------------------

if __name__ == "__main__":
    Y, _, s_true, true_par = simulate_slds(T=300, seed=7)
    par_hat, gamma, _ = train_em(Y, n_iter=20)
    est_state = (gamma[:, 1] > 0.5).astype(int)
    acc = (est_state == s_true).mean()
    print("True P:\n", true_par['P'])
    print("Estimated P:\n", np.round(par_hat['P'], 3))
    print(f"Classification accuracy: {acc:.3f}")
    plt.figure(figsize=(9, 3))
    plt.plot(gamma[:, 1], label="Smoothed Prob(Regime 1)")
    plt.plot(s_true, "k:", label="True Regime (0/1)")
    plt.legend(); plt.tight_layout(); plt.show()
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""
Switching‑DFM  (2 regimes × 2 factors × 3 observables)
完全版 EM ‑ Python 3.9+
-------------------------------------------------------
• log‑space Hamilton フィルタ
• log‑space Kim スムーザ + RTS クロス共分散
• Sticky‑Dirichlet 事前   (κ)
• Inverse‑Wishart MAP     (ν,Ψ)  for Q,R
• マルチリスタート EM    (N 回)
固定パラメータ:  T=300, restarts=20, κ=25, tol=1e‑4
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import logsumexp
from scipy.stats import multivariate_normal

# ------------------------------------------------------------
# 0.  数値安定ユーティリティ
# ------------------------------------------------------------
def psd_clip(S: np.ndarray, eps: float = 1e-6) -> np.ndarray:
    """固有値をクリップして PSD を保証"""
    vals, vecs = np.linalg.eigh(S)
    return vecs @ np.diag(np.clip(vals, eps, None)) @ vecs.T

# ------------------------------------------------------------
# 1.  人工データ生成  (真パラメータは固定)
# ------------------------------------------------------------
def simulate_slds(T=300, seed=7):
    rng = np.random.default_rng(seed)
    n_f, n_y, M = 2, 3, 2

    P_true = np.array([[0.9, 0.1],
                       [0.1, 0.9]])
    A_true = np.array([[[0.8, 0.0],
                        [0.0, 0.7]],
                       [[0.5, 0.0],
                        [0.0, 0.4]]])
    Q_true = np.array([0.2 * np.eye(n_f) for _ in range(M)])
    C_true = np.array([[1., 0.],
                       [0., 1.],
                       [1., 1.]])
    R_true = np.array([0.1 * np.eye(n_y) for _ in range(M)])

    s = np.zeros(T, int)
    f = np.zeros((T, n_f))
    y = np.zeros((T, n_y))

    s[0] = rng.integers(0, M)
    f[0] = rng.multivariate_normal(np.zeros(n_f), np.eye(n_f))
    y[0] = C_true @ f[0] + rng.multivariate_normal(np.zeros(n_y), R_true[s[0]])

    for t in range(1, T):
        s[t] = rng.choice(M, p=P_true[s[t - 1]])
        f[t] = A_true[s[t]] @ f[t - 1] + rng.multivariate_normal(np.zeros(n_f), Q_true[s[t]])
        y[t] = C_true @ f[t] + rng.multivariate_normal(np.zeros(n_y), R_true[s[t]])

    return y, s

# ------------------------------------------------------------
# 2.  Kalman 基本関数
# ------------------------------------------------------------
def kf_predict(m, P, A, Q):
    mp = A @ m
    Pp = psd_clip(A @ P @ A.T + Q)
    return mp, Pp


def kf_update(y, mp, Pp, C, R):
    S = psd_clip(C @ Pp @ C.T + R)
    K = Pp @ C.T @ np.linalg.inv(S)
    innov = y - C @ mp
    mf = mp + K @ innov
    Pf = psd_clip(Pp - K @ C @ Pp)
    ll = multivariate_normal.logpdf(
        innov, mean=np.zeros(len(innov)), cov=S, allow_singular=True)
    return mf, Pf, ll

# ------------------------------------------------------------
# 3.  log‑space Hamilton フィルタ
# ------------------------------------------------------------
def log_hamilton_filter(Y, par):
    """
    Returns
    -------
    log_alpha : (T,M)  フィルタ後 log p(s_t=j | y_1:t)
    mf, Pf    : (T,M,n_f), (T,M,n_f,n_f)  フィルタ平均と共分散
    mpred, Ppred : 1ステップ予測 (E[f_t|s_t=j,y_1:t-1])
    ll_t      : (T,M)  寄与対数尤度 log p(y_t | s_t=j, y_1:t-1)
    """
    T, n_y = Y.shape
    M, n_f = par['A'].shape[:2]

    log_alpha = np.full((T, M), -np.inf)
    mf = np.zeros((T, M, n_f))
    Pf = np.zeros((T, M, n_f, n_f))
    mpred = np.zeros_like(mf)
    Ppred = np.zeros_like(Pf)
    ll_t  = np.zeros((T, M))

    # t = 0
    for j in range(M):
        mp, Pp = kf_predict(np.zeros(n_f), np.eye(n_f), par['A'][j], par['Q'][j])
        mpred[0, j], Ppred[0, j] = mp, Pp
        mf[0, j], Pf[0, j], ll_t[0, j] = kf_update(Y[0], mp, Pp, par['C'], par['R'][j])
    log_alpha[0] = np.log(1 / M) + ll_t[0]
    log_alpha[0] -= logsumexp(log_alpha[0])          # normalize

    logP = np.log(par['P'] + 1e-12)

    for t in range(1, T):
        for j in range(M):
            log_mix = log_alpha[t - 1] + logP[:, j]
            w = np.exp(log_mix - logsumexp(log_mix))  # weights to mix Gaussians

            # moment‑matched mixture mean & cov
            mu_mix = np.zeros(n_f)
            Sig_mix = np.zeros((n_f, n_f))
            preds = []
            for i in range(M):
                mp, Pp = kf_predict(mf[t - 1, i], Pf[t - 1, i],
                                    par['A'][j], par['Q'][j])
                preds.append((mp, Pp))
                mu_mix += w[i] * mp
            for i, (mp, Pp) in enumerate(preds):
                diff = mp - mu_mix
                Sig_mix += w[i] * (Pp + np.outer(diff, diff))

            mpred[t, j], Ppred[t, j] = mu_mix, psd_clip(Sig_mix)
            mf[t, j], Pf[t, j], ll_t[t, j] = kf_update(
                Y[t], mu_mix, Sig_mix, par['C'], par['R'][j])

        log_alpha[t] = logsumexp(log_alpha[t - 1][:, None] + logP, axis=0) + ll_t[t]
        log_alpha[t] -= logsumexp(log_alpha[t])

    return log_alpha, mf, Pf, mpred, Ppred, ll_t


# ------------------------------------------------------------
# 4.  log‑space Kim スムーザ + RTS cross covariance
# ------------------------------------------------------------
def log_kim_smoother(log_alpha, par, ll_t, mf, Pf, mpred, Ppred):
    T, M = log_alpha.shape
    n_f = mf.shape[2]
    log_beta = np.zeros_like(log_alpha)
    log_gamma = np.copy(log_alpha)
    log_xi = np.zeros((T - 1, M, M))
    logP = np.log(par['P'] + 1e-12)

    ms = np.zeros_like(mf)
    Ps = np.zeros_like(Pf)
    P_pair = np.zeros_like(Pf)   # cross‑covariance E[f_t f_{t-1}ᵀ]

    ms[-1] = mf[-1]
    Ps[-1] = Pf[-1]

    for t in range(T - 2, -1, -1):
        mat = logP + ll_t[t + 1] + log_beta[t + 1]
        log_beta[t] = logsumexp(mat, axis=1)  # size M
        log_beta[t] -= logsumexp(log_beta[t])
        log_gamma[t] = log_alpha[t] + log_beta[t]
        log_gamma[t] -= logsumexp(log_gamma[t])
        log_xi[t] = (log_alpha[t][:, None] + logP +
                     ll_t[t + 1] + log_beta[t + 1])
        log_xi[t] -= logsumexp(log_xi[t])

        # Smoothed states with RTS
        for j in range(M):
            J = Pf[t, j] @ par['A'][j].T @ np.linalg.inv(Ppred[t + 1, j])
            ms[t, j] = mf[t, j] + J @ (ms[t + 1, j] - mpred[t + 1, j])
            Ps[t, j] = Pf[t, j] + J @ (Ps[t + 1, j] - Ppred[t + 1, j]) @ J.T
            P_pair[t + 1, j] = Ps[t + 1, j] @ J.T    # E[f_t  f_{t-1}ᵀ]

    return (np.exp(log_gamma),
            np.exp(log_xi),
            ms, Ps, P_pair)

# ------------------------------------------------------------
# 5.  M‑step  (Sticky Dirichlet + MAP for  Q, R)
# ------------------------------------------------------------
def m_step(Y, par, gamma, xi, ms, Ps, P_pair,
           kappa=25,       # Sticky strength
           nu=4,           # inv‑Wishart ν
           psi_scale=0.01  # inv‑Wishart Ψ = ψ·I
           ):
    T, n_y = Y.shape
    M, n_f = par['A'].shape[:2]

    # ----- (a) transition matrix P with sticky prior -----
    par['P'] = xi.sum(0) + kappa * np.eye(M)
    par['P'] /= par['P'].sum(axis=1, keepdims=True)

    # ----- (b) common C -----
    numC = np.zeros((n_y, n_f))
    denC = np.zeros((n_f, n_f))
    for t in range(T):
        for j in range(M):
            numC += gamma[t, j] * np.outer(Y[t], ms[t, j])
            denC += gamma[t, j] * (Ps[t, j] + np.outer(ms[t, j], ms[t, j]))
    par['C'] = numC @ np.linalg.inv(denC + 1e-12 * np.eye(n_f))

    Psi_R = psi_scale * np.eye(n_y)
    Psi_Q = psi_scale * np.eye(n_f)

    # ----- (c) per‑regime A, Q, R -----
    for j in range(M):
        Nj = gamma[:, j].sum()

        # R MAP
        SR = np.zeros((n_y, n_y))
        for t in range(T):
            res = Y[t] - par['C'] @ ms[t, j]
            SR += gamma[t, j] * (np.outer(res, res) +
                                 par['C'] @ Ps[t, j] @ par['C'].T)
        par['R'][j] = psd_clip((SR + Psi_R) / (Nj + nu + n_y + 1))

        # A (using cross‑covariance)
        numA = np.zeros((n_f, n_f))
        denA = np.zeros((n_f, n_f))
        for t in range(1, T):
            numA += gamma[t, j] * P_pair[t, j].T
            denA += gamma[t, j] * (Ps[t - 1, j] +
                                   np.outer(ms[t - 1, j], ms[t - 1, j]))
        par['A'][j] = numA @ np.linalg.inv(denA + 1e-12 * np.eye(n_f))

        # Q MAP
        SQ = np.zeros((n_f, n_f))
        for t in range(1, T):
            diff = ms[t, j] - par['A'][j] @ ms[t - 1, j]
            SQ += gamma[t, j] * (np.outer(diff, diff) + Ps[t, j])
        par['Q'][j] = psd_clip((SQ + Psi_Q) / (Nj + nu + n_f + 1))

    return par


# ------------------------------------------------------------
# 6.  EM (single seed)
# ------------------------------------------------------------
def em_single(Y, max_iter=100, tol=1e-4, seed=0,
              kappa=25, nu=4, psi_scale=0.01):
    np.random.seed(seed)
    n_y = Y.shape[1]

    # initial parameters
    par = {
        'P': np.array([[0.8, 0.2],
                       [0.2, 0.8]]),
        'A': np.array([np.eye(2) * 0.6,
                       np.eye(2) * 0.4]),
        'Q': np.array([np.eye(2) * 0.3,
                       np.eye(2) * 0.3]),
        'R': np.array([np.eye(n_y) * 0.3,
                       np.eye(n_y) * 0.3]),
        'C': np.array([[1., 0.],
                       [0., 1.],
                       [1., 1.]])
    }

    prev_ll = -np.inf
    for _ in range(max_iter):
        log_alpha, mf, Pf, mp, Pp, ll_t = log_hamilton_filter(Y, par)
        gamma, xi, ms, Ps, P_pair = log_kim_smoother(
            log_alpha, par, ll_t, mf, Pf, mp, Pp)
        ll = logsumexp(log_alpha, axis=1).sum()
        par = m_step(Y, par, gamma, xi, ms, Ps, P_pair,
                     kappa=kappa, nu=nu, psi_scale=psi_scale)
        if abs(ll - prev_ll) < tol:
            break
        prev_ll = ll
    return par, gamma, ll


# ------------------------------------------------------------
# 7.  Multi‑restart wrapper
# ------------------------------------------------------------
def train_em(Y, restarts=20, **kwargs):
    best_ll = -np.inf
    best_par = best_gamma = None
    for r in range(restarts):
        par, gamma, ll = em_single(Y, seed=r, **kwargs)
        if ll > best_ll:
            best_ll, best_par, best_gamma = ll, par, gamma
    return best_par, best_gamma, best_ll


# ------------------------------------------------------------
# 8.  Main demo  (parameters fixed)
# ------------------------------------------------------------
if __name__ == "__main__":
    # ----- fixed hyperparameters -----
    T = 300
    RESTARTS = 20
    KAPPA = 25
    TOL = 1e-4

    # ----- data -----
    Y, s_true = simulate_slds(T=T, seed=7)

    # ----- train -----
    par_hat, gamma_hat, ll = train_em(
        Y, restarts=RESTARTS, kappa=KAPPA, tol=TOL)

    # ----- evaluation -----
    state_est = (gamma_hat[:, 1] > 0.5).astype(int)
    accuracy = (state_est == s_true).mean()

    print("Estimated P:\n", np.round(par_hat['P'], 3))
    print(f"Regime classification accuracy: {accuracy:.3f}")

    # ----- plot -----
    t = np.arange(T)
    plt.figure(figsize=(10, 3))
    plt.plot(t, gamma_hat[:, 1], label="Smoothed Prob(Regime 1)")
    plt.plot(t, s_true, "k:", lw=1, label="True Regime (0/1)")
    plt.xlabel("time"); plt.ylim(-0.05, 1.05)
    plt.title("Smoothed Regime Probability vs Truth")
    plt.legend(); plt.tight_layout(); plt.show()
```
