```python
import numpy as np
import matplotlib.pyplot as plt

class RegimeSwitchingDFM:
    """
    AR(1)
    EM

    :
        Y_t = c * f_t + _t,   _t ~ N(0, R_s_t)
    :
        f_t = _s_t + a_s_t * f_{t-1} + v_t,   v_t ~ N(0, Q_s_t)
    :
        Pr(s_t=j | s_{t-1}=i) = _{ij}

    E: 
    M:  ((26)-(31))
    """
    def __init__(self, n_regimes=2, max_iter=100, tol=1e-4):
        self.M = n_regimes
        self.max_iter = max_iter
        self.tol = tol
        # 
        self.pi = np.full((self.M, self.M), 1/self.M)
        self.mu = np.zeros(self.M)
        self.a = np.zeros(self.M)
        self.Q = np.ones(self.M)
        self.c = 1.0
        self.R = 1.0

    def fit(self, y):
        T = len(y)
        # 
        self.pi[:] = 0.9*np.eye(self.M) + 0.1*(np.ones((self.M,self.M))-np.eye(self.M))/ (self.M-1)
        self.mu[:] = np.linspace(-1,1,self.M)
        self.a[:] = 0.5
        self.Q[:] = 1.0
        self.c = 1.0
        self.R = 1.0

        # 
        gamma = np.full((T,self.M),1/self.M)
        xi = np.full((T,self.M,self.M),1/(self.M**2))

        ll_old = -np.inf
        for it in range(self.max_iter):
            # E, xi, 
            gamma, xi, f_filt, P_filt, P_lag = self._expectation(y, gamma)
            # M
            self._maximization(y, gamma, xi, f_filt, P_filt, P_lag)
            # 
            ll = self._log_likelihood(y, gamma)
            if abs(ll-ll_old) < self.tol:
                break
            ll_old = ll
        return gamma

    def _expectation(self, y, gamma_prev):
        T = len(y)
        M = self.M
        # 
        f_pred = np.zeros((T,M))
        P_pred = np.zeros((T,M))
        f_filt = np.zeros((T,M))
        P_filt = np.zeros((T,M))
        P_lag  = np.zeros((T,M))
        xi = np.zeros((T,M,M))
        gamma = np.zeros((T,M))

        # 
        f_prev = np.zeros(M)
        P_prev = np.ones(M)*1e1
        w_prev = np.full(M,1/M)

        for t in range(T):
            w_ij = np.zeros((M,M))
            f_ij = np.zeros((M,M))
            P_ij = np.zeros((M,M))
            cov_ij = np.zeros((M,M))
            for i in range(M):
                for j in range(M):
                    w_pred = w_prev[i]*self.pi[i,j]
                    # 
                    f_pr = self.mu[j] + self.a[j]*f_prev[i]
                    P_pr = self.a[j]**2*P_prev[i] + self.Q[j]
                    # 
                    S = self.c**2 * P_pr + self.R
                    # 
                    L = np.exp(-0.5*(y[t]-self.c*f_pr)**2/S)/np.sqrt(2*np.pi*S)
                    w_upd = w_pred * L
                    K = P_pr*self.c/S
                    f_up = f_pr + K*(y[t]-self.c*f_pr)
                    P_up = (1-K*self.c)*P_pr
                    cov = (1-K*self.c)*self.a[j]*P_prev[i]
                    w_ij[i,j] = w_upd
                    f_ij[i,j] = f_up
                    P_ij[i,j] = P_up
                    cov_ij[i,j] = cov
            total = w_ij.sum()
            xi[t] = w_ij/total
            gamma[t] = xi[t].sum(axis=0)
            # collapse
            for j in range(M):
                # 
                f_filt[t,j] = (w_ij[:,j]*f_ij[:,j]).sum()/w_ij[:,j].sum()
                # 
                P_filt[t,j] = (w_ij[:,j]*(P_ij[:,j]+(f_ij[:,j]-f_filt[t,j])**2)).sum()/w_ij[:,j].sum()
                # 
                P_lag[t,j] = (w_ij[:,j]*(cov_ij[:,j])).sum()/w_ij[:,j].sum()
            # 
            f_prev = f_filt[t].copy()
            P_prev = P_filt[t].copy()
            w_prev = gamma[t].copy()
        return gamma, xi, f_filt, P_filt, P_lag

    def _maximization(self, y, gamma, xi, f_filt, P_filt, P_lag):
        T = len(y)
        M = self.M
        # 
        for i in range(M):
            for j in range(M):
                self.pi[i,j] = xi[1:,i,j].sum()/gamma[:-1,i].sum()
        #  c, R
        num_c = 0; den_c=0; num_R=0
        for t in range(T):
            for j in range(M):
                num_c += gamma[t,j]*y[t]*f_filt[t,j]
                den_c += gamma[t,j]*(f_filt[t,j]**2+P_filt[t,j])
        self.c = num_c/den_c
        for t in range(T):
            for j in range(M):
                pred = self.c*f_filt[t,j]
                num_R += gamma[t,j]*((y[t]-pred)**2 + self.c**2*P_filt[t,j])
        self.R = num_R/gamma.sum()
        #  , a, Q
        for j in range(M):
            w = gamma[1:,j]
            N_j = w.sum()
            S0 = (w*f_filt[1:,j]).sum(); S1 = (w*f_filt[:-1,j]).sum()
            S01= (w*(f_filt[1:,j]*f_filt[:-1,j]+P_lag[1:,j])).sum()
            S11= (w*(f_filt[:-1,j]**2+P_filt[:-1,j])).sum()
            Mmat = np.array([[N_j, S1],[S1, S11]])
            bvec = np.array([S0, S01])
            sol = np.linalg.solve(Mmat,bvec)
            self.mu[j], self.a[j] = sol
            # Q
            num_Q=0
            for t in range(1,T):
                v = f_filt[t,j] - self.mu[j] - self.a[j]*f_filt[t-1,j]
                num_Q += gamma[t,j]*(v**2 + P_filt[t,j] + self.a[j]**2*P_filt[t-1,j] - 2*self.a[j]*P_lag[t,j])
            self.Q[j] = num_Q/N_j

    def _log_likelihood(self, y, gamma):
        # : _t_j _tj log(_tj)
        return -np.sum(gamma*np.log(gamma+1e-16))


if __name__ == '__main__':
    # 
    np.random.seed(1)
    T = 300
    true_pi = np.array([[0.9,0.1],[0.1,0.9]])
    true_mu = np.array([-0.5,0.5])
    true_a  = np.array([0.95,0.5])
    true_Q  = np.array([0.1,0.1])
    true_c  = 1.0; true_R=0.5
    s_true = np.zeros(T,dtype=int)
    y = np.zeros(T); f=np.zeros(T)
    for t in range(1,T):
        s_true[t] = np.random.choice(2,p=true_pi[s_true[t-1]])
        j=s_true[t]
        f[t] = true_mu[j]+true_a[j]*f[t-1]+np.sqrt(true_Q[j])*np.random.randn()
        y[t] = true_c*f[t]+np.sqrt(true_R)*np.random.randn()
    # 
    model = RegimeSwitchingDFM(n_regimes=2)
    gamma = model.fit(y)
    # 
    plt.figure(figsize=(10,5))
    plt.plot(s_true, label='True Regime', lw=1)
    plt.plot(gamma[:,1], label='Est Prob Regime1', lw=1)
    plt.xlabel('Time'); plt.legend(); plt.tight_layout(); plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


class RS_MixedFreqDFM:
    """
    RegimeSwitching Dynamic Factor Model
         latent daily factor f_t  (AR(1))
         n_D  daily series:           y_D[:,d]  = c_D[d] * f_t + _t,d
         n_M  monthly series (monthend only):
              y_M[:,k]  = c_M[k] * (1/N) _{d=0}^{N-1} f_{t-d} + _t,k
     F_t = [f_t, f_{t-1}, , f_{t-L+1}]  (L  31 )
    EM  / 
    """

    def __init__(self, n_regimes=2, L=31, max_iter=100, tol=1e-4):
        self.M = n_regimes
        self.L = L                # 31 
        self.max_iter = max_iter
        self.tol = tol

        # ---  fit()  ---
        self.pi = np.full((self.M, self.M), 1 / self.M)
        self.mu = np.zeros(self.M)
        self.a  = np.zeros(self.M)
        self.Q  = np.ones(self.M)

        # ---  fit()  ---
        self.c_D = None         # shape (n_D,)
        self.c_M = None         # shape (n_M,)
        self.R_D = None         # shape (n_D,)
        self.R_M = None         # shape (n_M,)

    # ===================================================================
    #                              EM 
    # ===================================================================
    def fit(self, y_D, y_M, dates):
        """
        Parameters
        ----------
        y_D : (T, n_D) ndarray            
        y_M : (T, n_M) ndarray             NaN
        dates : pandas.DatetimeIndex        (len = T)

        Returns
        -------
        gamma : (T, M) ndarray            
        """
        y_D = np.asarray(y_D)
        y_M = np.asarray(y_M)
        T, n_D = y_D.shape
        _, n_M = y_M.shape
        self.n_D, self.n_M = n_D, n_M

        # ----------  ----------
        self.pi[:] = 0.9 * np.eye(self.M) + 0.1 / (self.M - 1) * (1 - np.eye(self.M))
        self.mu[:] = np.linspace(-0.5, 0.5, self.M)
        self.a[:] = 0.6
        self.Q[:] = 0.3

        self.c_D = np.ones(n_D)
        self.c_M = np.ones(n_M)
        self.R_D = np.ones(n_D) * 0.5
        self.R_M = np.ones(n_M) * 0.5

        # ---------- EM  ----------
        gamma = np.full((T, self.M), 1 / self.M)
        xi = np.full((T, self.M, self.M), 1 / self.M ** 2)
        ll_old = -np.inf

        for _ in range(self.max_iter):
            gamma, xi, F, P, P_lag = self._expectation(y_D, y_M, dates, gamma)
            self._maximization(y_D, y_M, dates, gamma, xi, F, P, P_lag)
            ll = -np.sum(gamma * np.log(gamma + 1e-16))
            if abs(ll - ll_old) < self.tol:
                break
            ll_old = ll
        return gamma

    # ===================================================================
    #                           Estep
    # ===================================================================
    def _expectation(self, y_D, y_M, dates, gamma_prev):
        T = len(y_D)
        L, M = self.L, self.M
        n_D, n_M = self.n_D, self.n_M

        F_filt = np.zeros((T, M, L))
        P_filt = np.zeros((T, M, L, L))
        P_lag  = np.zeros((T, M))
        xi     = np.zeros((T, M, M))
        gamma  = np.zeros((T, M))

        # mixture 
        F_prev = np.zeros((M, L))
        P_prev = np.tile(np.eye(L) * 10.0, (M, 1, 1))
        w_prev = np.full(M, 1 / M)
        I_L = np.eye(L)

        # regime
        A_mat = np.zeros((M, L, L))
        for j in range(M):
            A = np.zeros((L, L))
            A[0, 0] = self.a[j]
            A[1:, :-1] = np.eye(L - 1)
            A_mat[j] = A

        for t in range(T):
            # ----------------  C_t ----------------
            obs_rows = []
            obs_vals = []
            obs_var  = []

            #    daily rows
            for d in range(n_D):
                row = np.zeros(L)
                row[0] = self.c_D[d]
                obs_rows.append(row)
                obs_vals.append(y_D[t, d])
                obs_var.append(self.R_D[d])

            #    monthend rows
            if not np.isnan(y_M[t]).all():           # 
                N = self._month_length(dates[t])
                w = np.zeros(L); w[:N] = 1.0 / N
                for k in range(n_M):
                    if np.isnan(y_M[t, k]):          # 
                        continue
                    obs_rows.append(self.c_M[k] * w)
                    obs_vals.append(y_M[t, k])
                    obs_var.append(self.R_M[k])

            C_t = np.vstack(obs_rows)                     # kL
            R_t = np.diag(obs_var)                        # kk
            k = C_t.shape[0]

            # ----------------  ----------------
            w_ij   = np.zeros((M, M))
            F_ij   = np.zeros((M, M, L))
            P_ij   = np.zeros((M, M, L, L))
            cov_ij = np.zeros((M, M))                     # cov(f_t, f_{t-1})

            for i in range(M):
                for j in range(M):
                    w_pred = w_prev[i] * self.pi[i, j]

                    #   
                    F_pr = self.mu[j] + self.a[j] * F_prev[i, 0]
                    F_pr = np.concatenate([[F_pr], F_prev[i, :-1]])
                    P_pr = A_mat[j] @ P_prev[i] @ A_mat[j].T
                    P_pr[0, 0] += self.Q[j]

                    #   
                    S = C_t @ P_pr @ C_t.T + R_t
                    invS = np.linalg.inv(S)
                    res = np.asarray(obs_vals) - C_t @ F_pr
                    lh = (np.exp(-0.5 * res @ invS @ res) /
                          np.sqrt((2 * np.pi) ** k * np.linalg.det(S)))
                    K = P_pr @ C_t.T @ invS
                    F_up = F_pr + K @ res
                    P_up = (I_L - K @ C_t) @ P_pr

                    w_upd = w_pred * lh

                    w_ij[i, j] = w_upd
                    F_ij[i, j] = F_up
                    P_ij[i, j] = P_up
                    cov_ij[i, j] = P_up[0, 1]

            # ----------- mixture collapse -----------
            total = w_ij.sum()
            xi[t] = w_ij / total
            gamma[t] = xi[t].sum(axis=0)
            for j in range(M):
                wj = w_ij[:, j]; W = wj.sum()
                F_filt[t, j] = (wj[:, None] * F_ij[:, j]).sum(axis=0) / W
                diff = F_ij[:, j] - F_filt[t, j]
                P_filt[t, j] = (wj[:, None, None] *
                                (P_ij[:, j] + diff[:, :, None] * diff[:, None, :])
                                ).sum(axis=0) / W
                P_lag[t, j] = (wj * cov_ij[:, j]).sum() / W

            F_prev, P_prev, w_prev = F_filt[t].copy(), P_filt[t].copy(), gamma[t].copy()

        return gamma, xi, F_filt, P_filt, P_lag

    # ===================================================================
    #                           Mstep
    # ===================================================================
    def _maximization(self, y_D, y_M, dates, gamma, xi, F, P, P_lag):
        T, n_D = y_D.shape
        n_M = y_M.shape[1]
        M = self.M

        # ---- (1)  ----
        for i in range(M):
            den = gamma[:-1, i].sum()
            self.pi[i] = xi[1:, i].sum(axis=0) / den

        # ---- (2) daily loadings / variances ----
        num_cD = np.zeros(n_D); den_cD = 0.0
        num_RD = np.zeros(n_D)
        for t in range(T):
            for j in range(M):
                f0 = F[t, j, 0]
                den_part = f0 ** 2 + P[t, j, 0, 0]
                den_cD += gamma[t, j] * den_part
                for d in range(n_D):
                    num_cD[d] += gamma[t, j] * y_D[t, d] * f0
                    num_RD[d] += gamma[t, j] * ((y_D[t, d] - self.c_D[d] * f0) ** 2 +
                                                self.c_D[d] ** 2 * P[t, j, 0, 0])
        self.c_D = num_cD / den_cD
        self.R_D = num_RD / gamma.sum()

        # ---- (3) monthly loadings / variances ----
        num_cM = np.zeros(n_M); den_cM = np.zeros(n_M); num_RM = np.zeros(n_M)
        obs_cnt = np.zeros(n_M, dtype=int)

        for t in range(T):
            if np.isnan(y_M[t]).all():
                continue
            N = self._month_length(dates[t])
            w = np.zeros(self.L); w[:N] = 1.0 / N
            for j in range(M):
                agg = w @ F[t, j]
                var_agg = w @ P[t, j] @ w
                common_den = gamma[t, j] * (agg ** 2 + var_agg)
                for k in range(n_M):
                    if np.isnan(y_M[t, k]):
                        continue
                    num_cM[k] += gamma[t, j] * y_M[t, k] * agg
                    den_cM[k] += common_den
                    num_RM[k] += gamma[t, j] * ((y_M[t, k] - self.c_M[k] * agg) ** 2 +
                                                self.c_M[k] ** 2 * var_agg)
                    obs_cnt[k] += 1

        with np.errstate(divide="ignore", invalid="ignore"):
            self.c_M = np.where(den_cM > 0, num_cM / den_cM, self.c_M)
            self.R_M = np.where(obs_cnt > 0,
                                num_RM / (gamma.sum() * obs_cnt / T),
                                self.R_M)

        # ---- (4) state params , a, Q ----
        for j in range(M):
            w = gamma[1:, j]; Nj = w.sum()
            if Nj < 1e-10:
                continue
            S0 = (w * F[1:, j, 0]).sum()
            S1 = (w * F[:-1, j, 0]).sum()
            S01 = (w * (F[1:, j, 0] * F[:-1, j, 0] + P_lag[1:, j])).sum()
            S11 = (w * (F[:-1, j, 0] ** 2 + P[:-1, j, 0, 0])).sum()
            self.mu[j], self.a[j] = np.linalg.solve(
                np.array([[Nj, S1], [S1, S11]]), np.array([S0, S01]))
            numQ = 0.0
            for t in range(1, T):
                v = F[t, j, 0] - self.mu[j] - self.a[j] * F[t - 1, j, 0]
                numQ += gamma[t, j] * (
                    v ** 2 + P[t, j, 0, 0] +
                    self.a[j] ** 2 * P[t - 1, j, 0, 0] -
                    2 * self.a[j] * P_lag[t, j])
            self.Q[j] = numQ / Nj

    # -------------------------------------------------------------------
    #                       
    # -------------------------------------------------------------------
    @staticmethod
    def _month_length(ts):
        """datetime  """
        start = ts - pd.offsets.MonthBegin(n=1)
        return (ts - start).days

# -----------------------------------------------------------------
#           Daily (2 ) + Monthly (3 )
# -----------------------------------------------------------------
np.random.seed(1)
dates = pd.date_range("2021-01-01", "2022-12-31", freq="D")
T = len(dates)

# ----  () ----
Mtrue = 2
pi_true  = np.array([[0.95, 0.05], [0.04, 0.96]])
mu_true  = np.array([-0.3, 0.5])
a_true   = np.array([0.97, 0.65])
Q_true   = np.array([0.06, 0.11])

c_D_true = np.array([1.0, -0.8])            # 2 daily series
c_M_true = np.array([0.9, -0.7, 0.5])       # 3 monthly series
R_D_true = np.array([0.4, 0.35])
R_M_true = np.array([0.3, 0.25, 0.28])

# ----  ----
s = np.zeros(T, dtype=int)
f = np.zeros(T)
y_D = np.zeros((T, 2))
y_M = np.full((T, 3), np.nan)

for t in range(1, T):
    s[t] = np.random.choice(2, p=pi_true[s[t - 1]])
    j = s[t]
    f[t] = mu_true[j] + a_true[j] * f[t - 1] + np.sqrt(Q_true[j]) * np.random.randn()
    y_D[t] = c_D_true * f[t] + np.sqrt(R_D_true) * np.random.randn(2)

    if dates[t].is_month_end:
        N = (dates[t] - (dates[t]-pd.offsets.MonthBegin(1))).days
        avg_f = f[t - N + 1:t + 1].mean()
        y_M[t] = c_M_true * avg_f + np.sqrt(R_M_true) * np.random.randn(3)

# ----  ----
model = RS_MixedFreqDFM(L=31)
gamma = model.fit(y_D, y_M, dates)

# ----  ----
plt.figure(figsize=(10, 4))
plt.plot(s, label="True Regime", lw=0.6)
plt.plot(gamma[:, 1], label="Est. Prob(Regime 1)", lw=0.8)
plt.legend(); plt.tight_layout(); plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


class RS_MixedFreqDFM:
    """
    RegimeSwitching Dynamic Factor Model
         latent daily factor f_t  (AR(1))
         n_D  daily series:           y_D[:,d]  = c_D[d] * f_t + _t,d
         n_M  monthly series (monthend only):
              y_M[:,k]  = c_M[k] * (1/N) _{d=0}^{N-1} f_{t-d} + _t,k
     F_t = [f_t, f_{t-1}, , f_{t-L+1}]  (L  31 )
    EM  / 
    """

    def __init__(self, n_regimes=2, L=31, max_iter=100, tol=1e-4):
        self.M = n_regimes
        self.L = L                # 31 
        self.max_iter = max_iter
        self.tol = tol

        # ---  fit()  ---
        self.pi = np.full((self.M, self.M), 1 / self.M)
        self.mu = np.zeros(self.M)
        self.a  = np.zeros(self.M)
        self.Q  = np.ones(self.M)

        # ---  fit()  ---
        self.c_D = None         # shape (n_D,)
        self.c_M = None         # shape (n_M,)
        self.R_D = None         # shape (n_D,)
        self.R_M = None         # shape (n_M,)

    # ===================================================================
    #                              EM 
    # ===================================================================
    def fit(self, y_D, y_M, dates, s=None):  # 
        """
        Parameters
        ----------
        y_D : (T, n_D) ndarray            
        y_M : (T, n_M) ndarray             NaN
        dates : pandas.DatetimeIndex        (len = T)
        s : (T,) int array, optional       

        Returns
        -------
        gamma : (T, M) ndarray            
        """
        #  one-hot 
        if s is not None:
            T = len(s)
            gamma = np.zeros((T, self.M))
            gamma[np.arange(T), s] = 1
            return gamma

        y_D = np.asarray(y_D)
        y_M = np.asarray(y_M)
        T, n_D = y_D.shape
        _, n_M = y_M.shape
        self.n_D, self.n_M = n_D, n_M

        # ----------  ----------
        self.pi[:] = 0.9 * np.eye(self.M) + 0.1 / (self.M - 1) * (1 - np.eye(self.M))
        self.mu[:] = np.linspace(-0.5, 0.5, self.M)
        self.a[:] = 0.6
        self.Q[:] = 0.3

        self.c_D = np.ones(n_D)
        self.c_M = np.ones(n_M)
        self.R_D = np.ones(n_D) * 0.5
        self.R_M = np.ones(n_M) * 0.5

        # ---------- EM  ----------
        gamma = np.full((T, self.M), 1 / self.M)
        xi = np.full((T, self.M, self.M), 1 / self.M ** 2)
        ll_old = -np.inf

        for _ in range(self.max_iter):
            gamma, xi, F, P, P_lag = self._expectation(y_D, y_M, dates, gamma)
            self._maximization(y_D, y_M, dates, gamma, xi, F, P, P_lag)
            ll = -np.sum(gamma * np.log(gamma + 1e-16))
            if abs(ll - ll_old) < self.tol:
                break
            ll_old = ll
        return gamma

    # ===================================================================
    #                           Estep
    # ===================================================================
    def _expectation(self, y_D, y_M, dates, gamma_prev):
        T = len(y_D)
        L, M = self.L, self.M
        n_D, n_M = self.n_D, self.n_M

        F_filt = np.zeros((T, M, L))
        P_filt = np.zeros((T, M, L, L))
        P_lag  = np.zeros((T, M))
        xi     = np.zeros((T, M, M))
        gamma  = np.zeros((T, M))

        # mixture 
        F_prev = np.zeros((M, L))
        P_prev = np.tile(np.eye(L) * 10.0, (M, 1, 1))
        w_prev = np.full(M, 1 / M)
        I_L = np.eye(L)

        # regime
        A_mat = np.zeros((M, L, L))
        for j in range(M):
            A = np.zeros((L, L))
            A[0, 0] = self.a[j]
            A[1:, :-1] = np.eye(L - 1)
            A_mat[j] = A

        for t in range(T):
            # ----------------  C_t ----------------
            obs_rows = []
            obs_vals = []
            obs_var  = []

            #    daily rows
            for d in range(n_D):
                row = np.zeros(L)
                row[0] = self.c_D[d]
                obs_rows.append(row)
                obs_vals.append(y_D[t, d])
                obs_var.append(self.R_D[d])

            #    monthend rows
            if not np.isnan(y_M[t]).all():           # 
                N = self._month_length(dates[t])
                w = np.zeros(L); w[:N] = 1.0 / N
                for k in range(n_M):
                    if np.isnan(y_M[t, k]):          # 
                        continue
                    obs_rows.append(self.c_M[k] * w)
                    obs_vals.append(y_M[t, k])
                    obs_var.append(self.R_M[k])

            C_t = np.vstack(obs_rows)                     # kL
            R_t = np.diag(obs_var)                        # kk
            k = C_t.shape[0]

            # ----------------  ----------------
            w_ij   = np.zeros((M, M))
            F_ij   = np.zeros((M, M, L))
            P_ij   = np.zeros((M, M, L, L))
            cov_ij = np.zeros((M, M))                     # cov(f_t, f_{t-1})

            for i in range(M):
                for j in range(M):
                    w_pred = w_prev[i] * self.pi[i, j]

                    #   
                    F_pr = self.mu[j] + self.a[j] * F_prev[i, 0]
                    F_pr = np.concatenate([[F_pr], F_prev[i, :-1]])
                    P_pr = A_mat[j] @ P_prev[i] @ A_mat[j].T
                    P_pr[0, 0] += self.Q[j]

                    #   
                    S = C_t @ P_pr @ C_t.T + R_t
                    invS = np.linalg.inv(S)
                    res = np.asarray(obs_vals) - C_t @ F_pr
                    lh = (np.exp(-0.5 * res @ invS @ res) /
                          np.sqrt((2 * np.pi) ** k * np.linalg.det(S)))
                    K = P_pr @ C_t.T @ invS
                    F_up = F_pr + K @ res
                    P_up = (I_L - K @ C_t) @ P_pr

                    w_upd = w_pred * lh

                    w_ij[i, j] = w_upd
                    F_ij[i, j] = F_up
                    P_ij[i, j] = P_up
                    cov_ij[i, j] = P_up[0, 1]

            # ----------- mixture collapse -----------
            total = w_ij.sum()
            xi[t] = w_ij / total
            gamma[t] = xi[t].sum(axis=0)
            for j in range(M):
                wj = w_ij[:, j]; W = wj.sum()
                F_filt[t, j] = (wj[:, None] * F_ij[:, j]).sum(axis=0) / W
                diff = F_ij[:, j] - F_filt[t, j]
                P_filt[t, j] = (wj[:, None, None] *
                                (P_ij[:, j] + diff[:, :, None] * diff[:, None, :])
                                ).sum(axis=0) / W
                P_lag[t, j] = (wj * cov_ij[:, j]).sum() / W

            F_prev, P_prev, w_prev = F_filt[t].copy(), P_filt[t].copy(), gamma[t].copy()

        return gamma, xi, F_filt, P_filt, P_lag

    # ===================================================================
    #                           Mstep
    # ===================================================================
    def _maximization(self, y_D, y_M, dates, gamma, xi, F, P, P_lag):
        T, n_D = y_D.shape
        n_M = y_M.shape[1]
        M = self.M

        # ---- (1)  ----
        for i in range(M):
            den = gamma[:-1, i].sum()
            self.pi[i] = xi[1:, i].sum(axis=0) / den

        # ---- (2) daily loadings / variances ----
        num_cD = np.zeros(n_D); den_cD = 0.0
        num_RD = np.zeros(n_D)
        for t in range(T):
            for j in range(M):
                f0 = F[t, j, 0]
                den_part = f0 ** 2 + P[t, j, 0, 0]
                den_cD += gamma[t, j] * den_part
                for d in range(n_D):
                    num_cD[d] += gamma[t, j] * y_D[t, d] * f0
                    num_RD[d] += gamma[t, j] * ((y_D[t, d] - self.c_D[d] * f0) ** 2 +
                                                self.c_D[d] ** 2 * P[t, j, 0, 0])
        self.c_D = num_cD / den_cD
        self.R_D = num_RD / gamma.sum()

        # ---- (3) monthly loadings / variances ----
        num_cM = np.zeros(n_M); den_cM = np.zeros(n_M); num_RM = np.zeros(n_M)
        obs_cnt = np.zeros(n_M, dtype=int)

        for t in range(T):
            if np.isnan(y_M[t]).all():
                continue
            N = self._month_length(dates[t])
            w = np.zeros(self.L); w[:N] = 1.0 / N
            for j in range(M):
                agg = w @ F[t, j]
                var_agg = w @ P[t, j] @ w
                common_den = gamma[t, j] * (agg ** 2 + var_agg)
                for k in range(n_M):
                    if np.isnan(y_M[t, k]):
                        continue
                    num_cM[k] += gamma[t, j] * y_M[t, k] * agg
                    den_cM[k] += common_den
                    num_RM[k] += gamma[t, j] * ((y_M[t, k] - self.c_M[k] * agg) ** 2 +
                                                self.c_M[k] ** 2 * var_agg)
                    obs_cnt[k] += 1

        with np.errstate(divide="ignore", invalid="ignore"):
            self.c_M = np.where(den_cM > 0, num_cM / den_cM, self.c_M)
            self.R_M = np.where(obs_cnt > 0,
                                num_RM / (gamma.sum() * obs_cnt / T),
                                self.R_M)

        # ---- (4) state params , a, Q ----
        for j in range(M):
            w = gamma[1:, j]; Nj = w.sum()
            if Nj < 1e-10:
                continue
            S0 = (w * F[1:, j, 0]).sum()
            S1 = (w * F[:-1, j, 0]).sum()
            S01 = (w * (F[1:, j, 0] * F[:-1, j, 0] + P_lag[1:, j])).sum()
            S11 = (w * (F[:-1, j, 0] ** 2 + P[:-1, j, 0, 0])).sum()
            self.mu[j], self.a[j] = np.linalg.solve(
                np.array([[Nj, S1], [S1, S11]]), np.array([S0, S01]))
            numQ = 0.0
            for t in range(1, T):
                v = F[t, j, 0] - self.mu[j] - self.a[j] * F[t - 1, j, 0]
                numQ += gamma[t, j] * (
                    v ** 2 + P[t, j, 0, 0] +
                    self.a[j] ** 2 * P[t - 1, j, 0, 0] -
                    2 * self.a[j] * P_lag[t, j])
            self.Q[j] = numQ / Nj


    # -------------------------------------------------------------------
    #                       
    # -------------------------------------------------------------------
    @staticmethod
    def _month_length(ts):
        """datetime  """
        start = ts - pd.offsets.MonthBegin(n=1)
        return (ts - start).days

# -----------------------------------------------------------------
#           Daily (2 ) + Monthly (3 )
# -----------------------------------------------------------------
np.random.seed(1)
dates = pd.date_range("2021-01-01", "2022-12-31", freq="D")
T = len(dates)

# ----  () ----
Mtrue = 2
pi_true  = np.array([[0.95, 0.05], [0.04, 0.96]])
mu_true  = np.array([-0.3, 0.5])
a_true   = np.array([0.97, 0.65])
Q_true   = np.array([0.06, 0.11])

c_D_true = np.array([1.0, -0.8])            # 2 daily series
c_M_true = np.array([0.9, -0.7, 0.5])       # 3 monthly series
R_D_true = np.array([0.4, 0.35])
R_M_true = np.array([0.3, 0.25, 0.28])

# ----  ----
s = np.zeros(T, dtype=int)
f = np.zeros(T)
y_D = np.zeros((T, 2))
y_M = np.full((T, 3), np.nan)

for t in range(1, T):
    s[t] = np.random.choice(2, p=pi_true[s[t - 1]])
    j = s[t]
    f[t] = mu_true[j] + a_true[j] * f[t - 1] + np.sqrt(Q_true[j]) * np.random.randn()
    y_D[t] = c_D_true * f[t] + np.sqrt(R_D_true) * np.random.randn(2)

    if dates[t].is_month_end:
        N = (dates[t] - (dates[t]-pd.offsets.MonthBegin(1))).days
        avg_f = f[t - N + 1:t + 1].mean()
        y_M[t] = c_M_true * avg_f + np.sqrt(R_M_true) * np.random.randn(3)

# ----  ----
model = RS_MixedFreqDFM(L=31)
gamma = model.fit(y_D, y_M, dates, s=s)

# ----  ----
plt.figure(figsize=(10, 4))
plt.plot(s, label="True Regime", lw=0.6)
plt.plot(gamma[:, 1], label="Est. Prob(Regime 1)", lw=0.8)
plt.legend(); plt.tight_layout(); plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


class RS_MixedFreqDFM:
    """
    Regime-Switching Dynamic Factor Model
         latent daily factor f_t  (AR(1))
         n_D  daily series:           y_D[:,d]  = c_D[d] * f_t + _t,d
         n_M  monthly series (month-end only):
              y_M[:,k]  = c_M[k] * (1/N) _{d=0}^{N-1} f_{t-d} + _t,k
     F_t = [f_t, f_{t-1}, , f_{t-L+1}]  (L  31 )
    EM  / 
    
    """

    def __init__(self, n_regimes=2, L=31, max_iter=100, tol=1e-4):
        self.M = n_regimes
        self.L = L                # 31 
        self.max_iter = max_iter
        self.tol = tol

        # ---  fit()  ---
        self.pi = np.full((self.M, self.M), 1 / self.M)
        self.mu = np.zeros(self.M)
        self.a  = np.zeros(self.M)
        self.Q  = np.ones(self.M)

        # ---  fit()  ---
        self.c_D = None         # shape (n_D,)
        self.c_M = None         # shape (n_M,)
        self.R_D = None         # shape (n_D,)
        self.R_M = None         # shape (n_M,)

    # ===================================================================
    #                              EM 
    # ===================================================================
    def fit(self, y_D, y_M, dates, s=None):  # 
        """
        Parameters
        ----------
        y_D : (T, n_D) ndarray            
        y_M : (T, n_M) ndarray             NaN
        dates : pandas.DatetimeIndex        (len = T)
        s : (T,) int array, optional       

        Returns
        -------
        gamma : (T, M) ndarray            
        """
        #  one-hot 
        if s is not None:
            T = len(s)
            gamma = np.zeros((T, self.M))
            gamma[np.arange(T), s] = 1
            return gamma

        y_D = np.asarray(y_D)
        y_M = np.asarray(y_M)
        T, n_D = y_D.shape
        _, n_M = y_M.shape
        self.n_D, self.n_M = n_D, n_M

        # ----------  ----------
        self.pi[:] = 0.9 * np.eye(self.M) + 0.1 / (self.M - 1) * (1 - np.eye(self.M))
        self.mu[:] = np.linspace(-0.5, 0.5, self.M)
        self.a[:] = 0.6
        self.Q[:] = 0.3

        self.c_D = np.ones(n_D)
        self.c_M = np.ones(n_M)
        self.R_D = np.ones(n_D) * 0.5
        self.R_M = np.ones(n_M) * 0.5

        # ---------- EM  ----------
        gamma = np.full((T, self.M), 1 / self.M)
        xi = np.full((T, self.M, self.M), 1 / self.M ** 2)
        ll_old = -np.inf

        for _ in range(self.max_iter):
            gamma, xi, F, P, P_lag = self._expectation(y_D, y_M, dates, gamma)
            self._maximization(y_D, y_M, dates, gamma, xi, F, P, P_lag)
            ll = -np.sum(gamma * np.log(gamma + 1e-16))
            if abs(ll - ll_old) < self.tol:
                break
            ll_old = ll
        return gamma

    # ===================================================================
    #                           E-step
    # ===================================================================
    def _expectation(self, y_D, y_M, dates, gamma_prev):
        T = len(y_D)
        L, M = self.L, self.M
        # n_D, n_M known 
        n_D = y_D.shape[1]
        n_M = y_M.shape[1]

        F_filt = np.zeros((T, M, L))
        P_filt = np.zeros((T, M, L, L))
        P_lag  = np.zeros((T, M))
        xi     = np.zeros((T, M, M))
        gamma  = np.zeros((T, M))

        # mixture 
        F_prev = np.zeros((M, L))
        P_prev = np.tile(np.eye(L) * 10.0, (M, 1, 1))
        w_prev = np.full(M, 1 / M)
        I_L = np.eye(L)

        # regime
        A_mat = np.zeros((M, L, L))
        for j in range(M):
            A = np.zeros((L, L))
            A[0, 0] = self.a[j]
            A[1:, :-1] = np.eye(L - 1)
            A_mat[j] = A

        for t in range(T):
            # ----------------  C_t ----------------
            obs_rows = []
            obs_vals = []
            obs_var  = []

            #    daily rows
            for d in range(n_D):
                row = np.zeros(L)
                row[0] = self.c_D[d]
                obs_rows.append(row)
                obs_vals.append(y_D[t, d])
                obs_var.append(self.R_D[d])

            #    monthend rows
            if not np.isnan(y_M[t]).all():           # 
                N = self._month_length(dates[t])
                w = np.zeros(L); w[:N] = 1.0 / N
                for k in range(n_M):
                    if np.isnan(y_M[t, k]):          # 
                        continue
                    obs_rows.append(self.c_M[k] * w)
                    obs_vals.append(y_M[t, k])
                    obs_var.append(self.R_M[k])

            C_t = np.vstack(obs_rows)                     # kL
            R_t = np.diag(obs_var)                        # kk
            k = C_t.shape[0]

            # ----------------  ----------------
            w_ij   = np.zeros((M, M))
            F_ij   = np.zeros((M, M, L))
            P_ij   = np.zeros((M, M, L, L))
            cov_ij = np.zeros((M, M))                     # cov(f_t, f_{t-1})

            for i in range(M):
                for j in range(M):
                    w_pred = w_prev[i] * self.pi[i, j]

                    #   
                    F_pr = self.mu[j] + self.a[j] * F_prev[i, 0]
                    F_pr = np.concatenate([[F_pr], F_prev[i, :-1]])
                    P_pr = A_mat[j] @ P_prev[i] @ A_mat[j].T
                    P_pr[0, 0] += self.Q[j]

                    #   
                    S = C_t @ P_pr @ C_t.T + R_t
                    invS = np.linalg.inv(S)
                    res = np.asarray(obs_vals) - C_t @ F_pr
                    lh = (np.exp(-0.5 * res @ invS @ res) /
                          np.sqrt((2 * np.pi) ** k * np.linalg.det(S)))
                    K = P_pr @ C_t.T @ invS
                    F_up = F_pr + K @ res
                    P_up = (I_L - K @ C_t) @ P_pr

                    w_upd = w_pred * lh

                    w_ij[i, j] = w_upd
                    F_ij[i, j] = F_up
                    P_ij[i, j] = P_up
                    cov_ij[i, j] = P_up[0, 1]

            # ----------- mixture collapse -----------
            total = w_ij.sum()
            xi[t] = w_ij / total
            gamma[t] = xi[t].sum(axis=0)
            for j in range(M):
                wj = w_ij[:, j]; W = wj.sum()
                F_filt[t, j] = (wj[:, None] * F_ij[:, j]).sum(axis=0) / W
                diff = F_ij[:, j] - F_filt[t, j]
                P_filt[t, j] = (wj[:, None, None] *
                                (P_ij[:, j] + diff[:, :, None] * diff[:, None, :])
                                ).sum(axis=0) / W
                P_lag[t, j] = (wj * cov_ij[:, j]).sum() / W

            F_prev, P_prev, w_prev = F_filt[t].copy(), P_filt[t].copy(), gamma[t].copy()

        return gamma, xi, F_filt, P_filt, P_lag

    # ===================================================================
    #                           M-step
    # ===================================================================
    def _maximization(self, y_D, y_M, dates, gamma, xi, F, P, P_lag):
        T, n_D = y_D.shape
        n_M = y_M.shape[1]
        M = self.M

        # ---- (1)  ----
        for i in range(M):
            den = gamma[:-1, i].sum()
            self.pi[i] = xi[1:, i].sum(axis=0) / den

        # ---- (2) daily loadings / variances ----
        num_cD = np.zeros(n_D); den_cD = 0.0
        num_RD = np.zeros(n_D)
        for t in range(T):
            for j in range(M):
                f0 = F[t, j, 0]
                den_part = f0 ** 2 + P[t, j, 0, 0]
                den_cD += gamma[t, j] * den_part
                for d in range(n_D):
                    num_cD[d] += gamma[t, j] * y_D[t, d] * f0
                    num_RD[d] += gamma[t, j] * ((y_D[t, d] - self.c_D[d] * f0) ** 2 +
                                                self.c_D[d] ** 2 * P[t, j, 0, 0])
        self.c_D = num_cD / den_cD
        self.R_D = num_RD / gamma.sum()

        # ---- (3) monthly loadings / variances ----
        num_cM = np.zeros(n_M); den_cM = np.zeros(n_M); num_RM = np.zeros(n_M)
        obs_cnt = np.zeros(n_M, dtype=int)

        for t in range(T):
            if np.isnan(y_M[t]).all():
                continue
            N = self._month_length(dates[t])
            w = np.zeros(self.L); w[:N] = 1.0 / N
            for j in range(M):
                agg = w @ F[t, j]
                var_agg = w @ P[t, j] @ w
                common_den = gamma[t, j] * (agg ** 2 + var_agg)
                for k in range(n_M):
                    if np.isnan(y_M[t, k]):
                        continue
                    num_cM[k] += gamma[t, j] * y_M[t, k] * agg
                    den_cM[k] += common_den
                    num_RM[k] += gamma[t, j] * ((y_M[t, k] - self.c_M[k] * agg) ** 2 +
                                                self.c_M[k] ** 2 * var_agg)
                    obs_cnt[k] += 1

        with np.errstate(divide="ignore", invalid="ignore"):
            self.c_M = np.where(den_cM > 0, num_cM / den_cM, self.c_M)
            self.R_M = np.where(obs_cnt > 0,
                                num_RM / (gamma.sum() * obs_cnt / T),
                                self.R_M)

        # ---- (4) state params , a, Q ----
        for j in range(M):
            w = gamma[1:, j]; Nj = w.sum()
            if Nj < 1e-10:
                continue
            S0 = (w * F[1:, j, 0]).sum()
            S1 = (w * F[:-1, j, 0]).sum()
            S01 = (w * (F[1:, j, 0] * F[:-1, j, 0] + P_lag[1:, j])).sum()
            S11 = (w * (F[:-1, j, 0] ** 2 + P[:-1, j, 0, 0])).sum()
            self.mu[j], self.a[j] = np.linalg.solve(
                np.array([[Nj, S1], [S1, S11]]), np.array([S0, S01]))
            numQ = 0.0
            for t in range(1, T):
                v = F[t, j, 0] - self.mu[j] - self.a[j] * F[t - 1, j, 0]
                numQ += gamma[t, j] * (
                    v ** 2 + P[t, j, 0, 0] +
                    self.a[j] ** 2 * P[t - 1, j, 0, 0] -
                    2 * self.a[j] * P_lag[t, j])
            self.Q[j] = numQ / Nj

    # -------------------------------------------------------------------
    #                       
    # -------------------------------------------------------------------
    @staticmethod
    def _month_length(ts):
        """datetime  """
        start = ts - pd.offsets.MonthBegin(n=1)
        return (ts - start).days

# -----------------------------------------------------------------
#           Daily (2 ) + Monthly (3 )
# -----------------------------------------------------------------
np.random.seed(1)
dates = pd.date_range("2021-01-01", "2022-12-31", freq="D")
T = len(dates)

# ----  () ----
Mtrue = 2
pi_true  = np.array([[0.95, 0.05], [0.04, 0.96]])
mu_true  = np.array([-0.3, 0.5])
a_true   = np.array([0.97, 0.65])
Q_true   = np.array([0.06, 0.11])

c_D_true = np.array([1.0, -0.8])            # 2 daily series
c_M_true = np.array([0.9, -0.7, 0.5])       # 3 monthly series
R_D_true = np.array([0.4, 0.35])
R_M_true = np.array([0.3, 0.25, 0.28])

# ----  ----
s = np.zeros(T, dtype=int)
f = np.zeros(T)
y_D = np.zeros((T, 2))
y_M = np.full((T, 3), np.nan)

for t in range(1, T):
    s[t] = np.random.choice(2, p=pi_true[s[t - 1]])
    j = s[t]
    f[t] = mu_true[j] + a_true[j] * f[t - 1] + np.sqrt(Q_true[j]) * np.random.randn()
    y_D[t] = c_D_true * f[t] + np.sqrt(R_D_true) * np.random.randn(2)

    if dates[t].is_month_end:
        N = (dates[t] - (dates[t]-pd.offsets.MonthBegin(1))).days
        avg_f = f[t - N + 1:t + 1].mean()
        y_M[t] = c_M_true * avg_f + np.sqrt(R_M_true) * np.random.randn(3)

# ----  ----
model = RS_MixedFreqDFM(L=31)
gamma = model.fit(y_D, y_M, dates, s=s)

# ----  ----
plt.figure(figsize=(10, 4))
plt.plot(s, label="True Regime", lw=0.6)
plt.plot(gamma[:, 1], label="Est. Prob(Regime 1)", lw=0.8)
plt.legend(); plt.tight_layout(); plt.show()

# ----  ----
# t = 
_, _, F_filt, P_filt, _ = model._expectation(y_D, y_M, dates, gamma)
t = T - 1

gamma_t = gamma[t]                # shape (M,)
f_tj    = F_filt[t,:,0]          # 
P_tj    = P_filt[t,:,0,0]        # 

# 
f_pred_j = model.mu + model.a * f_tj  # shape (M,)
# 
f_pred   = (gamma_t * f_pred_j).sum()

# 
yD_pred  = model.c_D * f_pred     # shape (n_D,)

# 
L        = model.L
avg_pred = np.mean([f_pred] + list(f_tj[1:L]))
yM_pred  = model.c_M * avg_pred   # shape (n_M,)

print("1 f_pred:", f_pred)
print(" yD_pred:", yD_pred)
print(" yM_pred:", yM_pred)
```
