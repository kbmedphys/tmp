```python
import numpy as np
import matplotlib.pyplot as plt

class RegimeSwitchingDFM:
    """
    レジームスイッチング単因子動的ファクターモデル（AR(1)因子）を
    EMアルゴリズムで推定

    観測方程式:
        Y_t = c * f_t + ε_t,   ε_t ~ N(0, R_s_t)
    状態方程式:
        f_t = μ_s_t + a_s_t * f_{t-1} + v_t,   v_t ~ N(0, Q_s_t)
    レジーム遷移:
        Pr(s_t=j | s_{t-1}=i) = π_{ij}

    Eステップ: 修正カルマンフィルタ＋ラグ共分散で事後期待値を推定
    Mステップ: 閉形式でパラメータ更新 (式(26)-(31))
    """
    def __init__(self, n_regimes=2, max_iter=100, tol=1e-4):
        self.M = n_regimes
        self.max_iter = max_iter
        self.tol = tol
        # パラメータ初期化
        self.pi = np.full((self.M, self.M), 1/self.M)
        self.mu = np.zeros(self.M)
        self.a = np.zeros(self.M)
        self.Q = np.ones(self.M)
        self.c = 1.0
        self.R = 1.0

    def fit(self, y):
        T = len(y)
        # 初期推定値
        self.pi[:] = 0.9*np.eye(self.M) + 0.1*(np.ones((self.M,self.M))-np.eye(self.M))/ (self.M-1)
        self.mu[:] = np.linspace(-1,1,self.M)
        self.a[:] = 0.5
        self.Q[:] = 1.0
        self.c = 1.0
        self.R = 1.0

        # 初期事後確率
        gamma = np.full((T,self.M),1/self.M)
        xi = np.full((T,self.M,self.M),1/(self.M**2))

        ll_old = -np.inf
        for it in range(self.max_iter):
            # Eステップ：フィルタ・スムーザによるガンマ, xi, 平滑化因子推定
            gamma, xi, f_filt, P_filt, P_lag = self._expectation(y, gamma)
            # Mステップ：パラメータ更新
            self._maximization(y, gamma, xi, f_filt, P_filt, P_lag)
            # 対数尤度
            ll = self._log_likelihood(y, gamma)
            if abs(ll-ll_old) < self.tol:
                break
            ll_old = ll
        return gamma

    def _expectation(self, y, gamma_prev):
        T = len(y)
        M = self.M
        # フィルタ
        f_pred = np.zeros((T,M))
        P_pred = np.zeros((T,M))
        f_filt = np.zeros((T,M))
        P_filt = np.zeros((T,M))
        P_lag  = np.zeros((T,M))
        xi = np.zeros((T,M,M))
        gamma = np.zeros((T,M))

        # 初期値
        f_prev = np.zeros(M)
        P_prev = np.ones(M)*1e1
        w_prev = np.full(M,1/M)

        for t in range(T):
            w_ij = np.zeros((M,M))
            f_ij = np.zeros((M,M))
            P_ij = np.zeros((M,M))
            cov_ij = np.zeros((M,M))
            for i in range(M):
                for j in range(M):
                    w_pred = w_prev[i]*self.pi[i,j]
                    # 状態予測
                    f_pr = self.mu[j] + self.a[j]*f_prev[i]
                    P_pr = self.a[j]**2*P_prev[i] + self.Q[j]
                    # 観測予測
                    S = self.c**2 * P_pr + self.R
                    # 観測尤度
                    L = np.exp(-0.5*(y[t]-self.c*f_pr)**2/S)/np.sqrt(2*np.pi*S)
                    w_upd = w_pred * L
                    K = P_pr*self.c/S
                    f_up = f_pr + K*(y[t]-self.c*f_pr)
                    P_up = (1-K*self.c)*P_pr
                    cov = (1-K*self.c)*self.a[j]*P_prev[i]
                    w_ij[i,j] = w_upd
                    f_ij[i,j] = f_up
                    P_ij[i,j] = P_up
                    cov_ij[i,j] = cov
            total = w_ij.sum()
            xi[t] = w_ij/total
            gamma[t] = xi[t].sum(axis=0)
            # collapse
            for j in range(M):
                # フィルタ済み因子
                f_filt[t,j] = (w_ij[:,j]*f_ij[:,j]).sum()/w_ij[:,j].sum()
                # 分散
                P_filt[t,j] = (w_ij[:,j]*(P_ij[:,j]+(f_ij[:,j]-f_filt[t,j])**2)).sum()/w_ij[:,j].sum()
                # ラグ共分散
                P_lag[t,j] = (w_ij[:,j]*(cov_ij[:,j])).sum()/w_ij[:,j].sum()
            # 次準備
            f_prev = f_filt[t].copy()
            P_prev = P_filt[t].copy()
            w_prev = gamma[t].copy()
        return gamma, xi, f_filt, P_filt, P_lag

    def _maximization(self, y, gamma, xi, f_filt, P_filt, P_lag):
        T = len(y)
        M = self.M
        # π
        for i in range(M):
            for j in range(M):
                self.pi[i,j] = xi[1:,i,j].sum()/gamma[:-1,i].sum()
        # 観測 c, R
        num_c = 0; den_c=0; num_R=0
        for t in range(T):
            for j in range(M):
                num_c += gamma[t,j]*y[t]*f_filt[t,j]
                den_c += gamma[t,j]*(f_filt[t,j]**2+P_filt[t,j])
        self.c = num_c/den_c
        for t in range(T):
            for j in range(M):
                pred = self.c*f_filt[t,j]
                num_R += gamma[t,j]*((y[t]-pred)**2 + self.c**2*P_filt[t,j])
        self.R = num_R/gamma.sum()
        # 状態 μ, a, Q
        for j in range(M):
            w = gamma[1:,j]
            N_j = w.sum()
            S0 = (w*f_filt[1:,j]).sum(); S1 = (w*f_filt[:-1,j]).sum()
            S01= (w*(f_filt[1:,j]*f_filt[:-1,j]+P_lag[1:,j])).sum()
            S11= (w*(f_filt[:-1,j]**2+P_filt[:-1,j])).sum()
            Mmat = np.array([[N_j, S1],[S1, S11]])
            bvec = np.array([S0, S01])
            sol = np.linalg.solve(Mmat,bvec)
            self.mu[j], self.a[j] = sol
            # Q
            num_Q=0
            for t in range(1,T):
                v = f_filt[t,j] - self.mu[j] - self.a[j]*f_filt[t-1,j]
                num_Q += gamma[t,j]*(v**2 + P_filt[t,j] + self.a[j]**2*P_filt[t-1,j] - 2*self.a[j]*P_lag[t,j])
            self.Q[j] = num_Q/N_j

    def _log_likelihood(self, y, gamma):
        # 簡易対数尤度: −∑_t∑_j γ_tj log(γ_tj)
        return -np.sum(gamma*np.log(gamma+1e-16))


if __name__ == '__main__':
    # シミュレーション
    np.random.seed(1)
    T = 300
    true_pi = np.array([[0.9,0.1],[0.1,0.9]])
    true_mu = np.array([-0.5,0.5])
    true_a  = np.array([0.95,0.5])
    true_Q  = np.array([0.1,0.1])
    true_c  = 1.0; true_R=0.5
    s_true = np.zeros(T,dtype=int)
    y = np.zeros(T); f=np.zeros(T)
    for t in range(1,T):
        s_true[t] = np.random.choice(2,p=true_pi[s_true[t-1]])
        j=s_true[t]
        f[t] = true_mu[j]+true_a[j]*f[t-1]+np.sqrt(true_Q[j])*np.random.randn()
        y[t] = true_c*f[t]+np.sqrt(true_R)*np.random.randn()
    # 推定
    model = RegimeSwitchingDFM(n_regimes=2)
    gamma = model.fit(y)
    # プロット
    plt.figure(figsize=(10,5))
    plt.plot(s_true, label='True Regime', lw=1)
    plt.plot(gamma[:,1], label='Est Prob Regime1', lw=1)
    plt.xlabel('Time'); plt.legend(); plt.tight_layout(); plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


class RS_MixedFreqDFM:
    """
    Regime‑Switching Dynamic Factor Model
        • latent daily factor f_t  (AR(1))
        • n_D  daily series:           y_D[:,d]  = c_D[d] * f_t + ε_t,d
        • n_M  monthly series (month‑end only):
              y_M[:,k]  = c_M[k] * (1/N) Σ_{d=0}^{N-1} f_{t-d} + ε_t,k
    状態ベクトル F_t = [f_t, f_{t-1}, …, f_{t-L+1}]ᵀ  (L ≥ 31 推奨)
    EM アルゴリズム（修正カルマンフィルタ / スムーザ）で推定
    """

    def __init__(self, n_regimes=2, L=31, max_iter=100, tol=1e-4):
        self.M = n_regimes
        self.L = L                # ラグ長（31 で「月平均」可）
        self.max_iter = max_iter
        self.tol = tol

        # --- レジーム別状態パラメータ（初期値は fit() で上書き） ---
        self.pi = np.full((self.M, self.M), 1 / self.M)
        self.mu = np.zeros(self.M)
        self.a  = np.zeros(self.M)
        self.Q  = np.ones(self.M)

        # --- 観測パラメータ（サイズは fit() 内で決定） ---
        self.c_D = None         # shape (n_D,)
        self.c_M = None         # shape (n_M,)
        self.R_D = None         # shape (n_D,)
        self.R_M = None         # shape (n_M,)

    # ===================================================================
    #                              EM 外部ループ
    # ===================================================================
    def fit(self, y_D, y_M, dates):
        """
        Parameters
        ----------
        y_D : (T, n_D) ndarray            毎日観測（欠測なし）
        y_M : (T, n_M) ndarray            月末のみ値、他は NaN
        dates : pandas.DatetimeIndex       日次インデックス (len = T)

        Returns
        -------
        gamma : (T, M) ndarray            平滑化レジーム確率
        """
        y_D = np.asarray(y_D)
        y_M = np.asarray(y_M)
        T, n_D = y_D.shape
        _, n_M = y_M.shape
        self.n_D, self.n_M = n_D, n_M

        # ---------- パラメータ初期化 ----------
        self.pi[:] = 0.9 * np.eye(self.M) + 0.1 / (self.M - 1) * (1 - np.eye(self.M))
        self.mu[:] = np.linspace(-0.5, 0.5, self.M)
        self.a[:] = 0.6
        self.Q[:] = 0.3

        self.c_D = np.ones(n_D)
        self.c_M = np.ones(n_M)
        self.R_D = np.ones(n_D) * 0.5
        self.R_M = np.ones(n_M) * 0.5

        # ---------- EM ループ ----------
        gamma = np.full((T, self.M), 1 / self.M)
        xi = np.full((T, self.M, self.M), 1 / self.M ** 2)
        ll_old = -np.inf

        for _ in range(self.max_iter):
            gamma, xi, F, P, P_lag = self._expectation(y_D, y_M, dates, gamma)
            self._maximization(y_D, y_M, dates, gamma, xi, F, P, P_lag)
            ll = -np.sum(gamma * np.log(gamma + 1e-16))
            if abs(ll - ll_old) < self.tol:
                break
            ll_old = ll
        return gamma

    # ===================================================================
    #                           E‑step
    # ===================================================================
    def _expectation(self, y_D, y_M, dates, gamma_prev):
        T = len(y_D)
        L, M = self.L, self.M
        n_D, n_M = self.n_D, self.n_M

        F_filt = np.zeros((T, M, L))
        P_filt = np.zeros((T, M, L, L))
        P_lag  = np.zeros((T, M))
        xi     = np.zeros((T, M, M))
        gamma  = np.zeros((T, M))

        # mixture の初期値
        F_prev = np.zeros((M, L))
        P_prev = np.tile(np.eye(L) * 10.0, (M, 1, 1))
        w_prev = np.full(M, 1 / M)
        I_L = np.eye(L)

        # regime‑別遷移行列
        A_mat = np.zeros((M, L, L))
        for j in range(M):
            A = np.zeros((L, L))
            A[0, 0] = self.a[j]
            A[1:, :-1] = np.eye(L - 1)
            A_mat[j] = A

        for t in range(T):
            # ---------------- 観測行列 C_t ----------------
            obs_rows = []
            obs_vals = []
            obs_var  = []

            #   ‑‑ daily rows
            for d in range(n_D):
                row = np.zeros(L)
                row[0] = self.c_D[d]
                obs_rows.append(row)
                obs_vals.append(y_D[t, d])
                obs_var.append(self.R_D[d])

            #   ‑‑ month‑end rows
            if not np.isnan(y_M[t]).all():           # 少なくとも一列は値あり
                N = self._month_length(dates[t])
                w = np.zeros(L); w[:N] = 1.0 / N
                for k in range(n_M):
                    if np.isnan(y_M[t, k]):          # 欠測は飛ばす
                        continue
                    obs_rows.append(self.c_M[k] * w)
                    obs_vals.append(y_M[t, k])
                    obs_var.append(self.R_M[k])

            C_t = np.vstack(obs_rows)                     # k×L
            R_t = np.diag(obs_var)                        # k×k
            k = C_t.shape[0]

            # ---------------- 各パス毎の更新 ----------------
            w_ij   = np.zeros((M, M))
            F_ij   = np.zeros((M, M, L))
            P_ij   = np.zeros((M, M, L, L))
            cov_ij = np.zeros((M, M))                     # cov(f_t, f_{t-1})

            for i in range(M):
                for j in range(M):
                    w_pred = w_prev[i] * self.pi[i, j]

                    # — 状態予測 —
                    F_pr = self.mu[j] + self.a[j] * F_prev[i, 0]
                    F_pr = np.concatenate([[F_pr], F_prev[i, :-1]])
                    P_pr = A_mat[j] @ P_prev[i] @ A_mat[j].T
                    P_pr[0, 0] += self.Q[j]

                    # — 観測更新 —
                    S = C_t @ P_pr @ C_t.T + R_t
                    invS = np.linalg.inv(S)
                    res = np.asarray(obs_vals) - C_t @ F_pr
                    lh = (np.exp(-0.5 * res @ invS @ res) /
                          np.sqrt((2 * np.pi) ** k * np.linalg.det(S)))
                    K = P_pr @ C_t.T @ invS
                    F_up = F_pr + K @ res
                    P_up = (I_L - K @ C_t) @ P_pr

                    w_upd = w_pred * lh

                    w_ij[i, j] = w_upd
                    F_ij[i, j] = F_up
                    P_ij[i, j] = P_up
                    cov_ij[i, j] = P_up[0, 1]

            # ----------- mixture collapse -----------
            total = w_ij.sum()
            xi[t] = w_ij / total
            gamma[t] = xi[t].sum(axis=0)
            for j in range(M):
                wj = w_ij[:, j]; W = wj.sum()
                F_filt[t, j] = (wj[:, None] * F_ij[:, j]).sum(axis=0) / W
                diff = F_ij[:, j] - F_filt[t, j]
                P_filt[t, j] = (wj[:, None, None] *
                                (P_ij[:, j] + diff[:, :, None] * diff[:, None, :])
                                ).sum(axis=0) / W
                P_lag[t, j] = (wj * cov_ij[:, j]).sum() / W

            F_prev, P_prev, w_prev = F_filt[t].copy(), P_filt[t].copy(), gamma[t].copy()

        return gamma, xi, F_filt, P_filt, P_lag

    # ===================================================================
    #                           M‑step
    # ===================================================================
    def _maximization(self, y_D, y_M, dates, gamma, xi, F, P, P_lag):
        T, n_D = y_D.shape
        n_M = y_M.shape[1]
        M = self.M

        # ---- (1) π ----
        for i in range(M):
            den = gamma[:-1, i].sum()
            self.pi[i] = xi[1:, i].sum(axis=0) / den

        # ---- (2) daily loadings / variances ----
        num_cD = np.zeros(n_D); den_cD = 0.0
        num_RD = np.zeros(n_D)
        for t in range(T):
            for j in range(M):
                f0 = F[t, j, 0]
                den_part = f0 ** 2 + P[t, j, 0, 0]
                den_cD += gamma[t, j] * den_part
                for d in range(n_D):
                    num_cD[d] += gamma[t, j] * y_D[t, d] * f0
                    num_RD[d] += gamma[t, j] * ((y_D[t, d] - self.c_D[d] * f0) ** 2 +
                                                self.c_D[d] ** 2 * P[t, j, 0, 0])
        self.c_D = num_cD / den_cD
        self.R_D = num_RD / gamma.sum()

        # ---- (3) monthly loadings / variances ----
        num_cM = np.zeros(n_M); den_cM = np.zeros(n_M); num_RM = np.zeros(n_M)
        obs_cnt = np.zeros(n_M, dtype=int)

        for t in range(T):
            if np.isnan(y_M[t]).all():
                continue
            N = self._month_length(dates[t])
            w = np.zeros(self.L); w[:N] = 1.0 / N
            for j in range(M):
                agg = w @ F[t, j]
                var_agg = w @ P[t, j] @ w
                common_den = gamma[t, j] * (agg ** 2 + var_agg)
                for k in range(n_M):
                    if np.isnan(y_M[t, k]):
                        continue
                    num_cM[k] += gamma[t, j] * y_M[t, k] * agg
                    den_cM[k] += common_den
                    num_RM[k] += gamma[t, j] * ((y_M[t, k] - self.c_M[k] * agg) ** 2 +
                                                self.c_M[k] ** 2 * var_agg)
                    obs_cnt[k] += 1

        with np.errstate(divide="ignore", invalid="ignore"):
            self.c_M = np.where(den_cM > 0, num_cM / den_cM, self.c_M)
            self.R_M = np.where(obs_cnt > 0,
                                num_RM / (gamma.sum() * obs_cnt / T),
                                self.R_M)

        # ---- (4) state params μ, a, Q ----
        for j in range(M):
            w = gamma[1:, j]; Nj = w.sum()
            if Nj < 1e-10:
                continue
            S0 = (w * F[1:, j, 0]).sum()
            S1 = (w * F[:-1, j, 0]).sum()
            S01 = (w * (F[1:, j, 0] * F[:-1, j, 0] + P_lag[1:, j])).sum()
            S11 = (w * (F[:-1, j, 0] ** 2 + P[:-1, j, 0, 0])).sum()
            self.mu[j], self.a[j] = np.linalg.solve(
                np.array([[Nj, S1], [S1, S11]]), np.array([S0, S01]))
            numQ = 0.0
            for t in range(1, T):
                v = F[t, j, 0] - self.mu[j] - self.a[j] * F[t - 1, j, 0]
                numQ += gamma[t, j] * (
                    v ** 2 + P[t, j, 0, 0] +
                    self.a[j] ** 2 * P[t - 1, j, 0, 0] -
                    2 * self.a[j] * P_lag[t, j])
            self.Q[j] = numQ / Nj

    # -------------------------------------------------------------------
    #                       ユーティリティ
    # -------------------------------------------------------------------
    @staticmethod
    def _month_length(ts):
        """datetime → その月の日数"""
        start = ts - pd.offsets.MonthBegin(n=1)
        return (ts - start).days

# -----------------------------------------------------------------
#         シミュレーション：  Daily (2 列) + Monthly (3 列)
# -----------------------------------------------------------------
np.random.seed(1)
dates = pd.date_range("2021-01-01", "2022-12-31", freq="D")
T = len(dates)

# ---- 真のパラメータ (例) ----
Mtrue = 2
pi_true  = np.array([[0.95, 0.05], [0.04, 0.96]])
mu_true  = np.array([-0.3, 0.5])
a_true   = np.array([0.97, 0.65])
Q_true   = np.array([0.06, 0.11])

c_D_true = np.array([1.0, -0.8])            # 2 daily series
c_M_true = np.array([0.9, -0.7, 0.5])       # 3 monthly series
R_D_true = np.array([0.4, 0.35])
R_M_true = np.array([0.3, 0.25, 0.28])

# ---- データ生成 ----
s = np.zeros(T, dtype=int)
f = np.zeros(T)
y_D = np.zeros((T, 2))
y_M = np.full((T, 3), np.nan)

for t in range(1, T):
    s[t] = np.random.choice(2, p=pi_true[s[t - 1]])
    j = s[t]
    f[t] = mu_true[j] + a_true[j] * f[t - 1] + np.sqrt(Q_true[j]) * np.random.randn()
    y_D[t] = c_D_true * f[t] + np.sqrt(R_D_true) * np.random.randn(2)

    if dates[t].is_month_end:
        N = (dates[t] - (dates[t]-pd.offsets.MonthBegin(1))).days
        avg_f = f[t - N + 1:t + 1].mean()
        y_M[t] = c_M_true * avg_f + np.sqrt(R_M_true) * np.random.randn(3)

# ---- 推定 ----
model = RS_MixedFreqDFM(L=31)
gamma = model.fit(y_D, y_M, dates)

# ---- 結果可視化 ----
plt.figure(figsize=(10, 4))
plt.plot(s, label="True Regime", lw=0.6)
plt.plot(gamma[:, 1], label="Est. Prob(Regime 1)", lw=0.8)
plt.legend(); plt.tight_layout(); plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


class RS_MixedFreqDFM:
    """
    Regime‑Switching Dynamic Factor Model
        • latent daily factor f_t  (AR(1))
        • n_D  daily series:           y_D[:,d]  = c_D[d] * f_t + ε_t,d
        • n_M  monthly series (month‑end only):
              y_M[:,k]  = c_M[k] * (1/N) Σ_{d=0}^{N-1} f_{t-d} + ε_t,k
    状態ベクトル F_t = [f_t, f_{t-1}, …, f_{t-L+1}]ᵀ  (L ≥ 31 推奨)
    EM アルゴリズム（修正カルマンフィルタ / スムーザ）で推定
    """

    def __init__(self, n_regimes=2, L=31, max_iter=100, tol=1e-4):
        self.M = n_regimes
        self.L = L                # ラグ長（31 で「月平均」可）
        self.max_iter = max_iter
        self.tol = tol

        # --- レジーム別状態パラメータ（初期値は fit() で上書き） ---
        self.pi = np.full((self.M, self.M), 1 / self.M)
        self.mu = np.zeros(self.M)
        self.a  = np.zeros(self.M)
        self.Q  = np.ones(self.M)

        # --- 観測パラメータ（サイズは fit() 内で決定） ---
        self.c_D = None         # shape (n_D,)
        self.c_M = None         # shape (n_M,)
        self.R_D = None         # shape (n_D,)
        self.R_M = None         # shape (n_M,)

    # ===================================================================
    #                              EM 外部ループ
    # ===================================================================
    def fit(self, y_D, y_M, dates, s=None):  # レジームシークエンスをオプション引数に追加
        """
        Parameters
        ----------
        y_D : (T, n_D) ndarray            毎日観測（欠測なし）
        y_M : (T, n_M) ndarray            月末のみ値、他は NaN
        dates : pandas.DatetimeIndex       日次インデックス (len = T)
        s : (T,) int array, optional       既知のレジームシークエンス

        Returns
        -------
        gamma : (T, M) ndarray            平滑化レジーム確率
        """
        # 既知シークエンスが指定された場合は one-hot を返す
        if s is not None:
            T = len(s)
            gamma = np.zeros((T, self.M))
            gamma[np.arange(T), s] = 1
            return gamma

        y_D = np.asarray(y_D)
        y_M = np.asarray(y_M)
        T, n_D = y_D.shape
        _, n_M = y_M.shape
        self.n_D, self.n_M = n_D, n_M

        # ---------- パラメータ初期化 ----------
        self.pi[:] = 0.9 * np.eye(self.M) + 0.1 / (self.M - 1) * (1 - np.eye(self.M))
        self.mu[:] = np.linspace(-0.5, 0.5, self.M)
        self.a[:] = 0.6
        self.Q[:] = 0.3

        self.c_D = np.ones(n_D)
        self.c_M = np.ones(n_M)
        self.R_D = np.ones(n_D) * 0.5
        self.R_M = np.ones(n_M) * 0.5

        # ---------- EM ループ ----------
        gamma = np.full((T, self.M), 1 / self.M)
        xi = np.full((T, self.M, self.M), 1 / self.M ** 2)
        ll_old = -np.inf

        for _ in range(self.max_iter):
            gamma, xi, F, P, P_lag = self._expectation(y_D, y_M, dates, gamma)
            self._maximization(y_D, y_M, dates, gamma, xi, F, P, P_lag)
            ll = -np.sum(gamma * np.log(gamma + 1e-16))
            if abs(ll - ll_old) < self.tol:
                break
            ll_old = ll
        return gamma

    # ===================================================================
    #                           E‑step
    # ===================================================================
    def _expectation(self, y_D, y_M, dates, gamma_prev):
        T = len(y_D)
        L, M = self.L, self.M
        n_D, n_M = self.n_D, self.n_M

        F_filt = np.zeros((T, M, L))
        P_filt = np.zeros((T, M, L, L))
        P_lag  = np.zeros((T, M))
        xi     = np.zeros((T, M, M))
        gamma  = np.zeros((T, M))

        # mixture の初期値
        F_prev = np.zeros((M, L))
        P_prev = np.tile(np.eye(L) * 10.0, (M, 1, 1))
        w_prev = np.full(M, 1 / M)
        I_L = np.eye(L)

        # regime‑別遷移行列
        A_mat = np.zeros((M, L, L))
        for j in range(M):
            A = np.zeros((L, L))
            A[0, 0] = self.a[j]
            A[1:, :-1] = np.eye(L - 1)
            A_mat[j] = A

        for t in range(T):
            # ---------------- 観測行列 C_t ----------------
            obs_rows = []
            obs_vals = []
            obs_var  = []

            #   ‑‑ daily rows
            for d in range(n_D):
                row = np.zeros(L)
                row[0] = self.c_D[d]
                obs_rows.append(row)
                obs_vals.append(y_D[t, d])
                obs_var.append(self.R_D[d])

            #   ‑‑ month‑end rows
            if not np.isnan(y_M[t]).all():           # 少なくとも一列は値あり
                N = self._month_length(dates[t])
                w = np.zeros(L); w[:N] = 1.0 / N
                for k in range(n_M):
                    if np.isnan(y_M[t, k]):          # 欠測は飛ばす
                        continue
                    obs_rows.append(self.c_M[k] * w)
                    obs_vals.append(y_M[t, k])
                    obs_var.append(self.R_M[k])

            C_t = np.vstack(obs_rows)                     # k×L
            R_t = np.diag(obs_var)                        # k×k
            k = C_t.shape[0]

            # ---------------- 各パス毎の更新 ----------------
            w_ij   = np.zeros((M, M))
            F_ij   = np.zeros((M, M, L))
            P_ij   = np.zeros((M, M, L, L))
            cov_ij = np.zeros((M, M))                     # cov(f_t, f_{t-1})

            for i in range(M):
                for j in range(M):
                    w_pred = w_prev[i] * self.pi[i, j]

                    # — 状態予測 —
                    F_pr = self.mu[j] + self.a[j] * F_prev[i, 0]
                    F_pr = np.concatenate([[F_pr], F_prev[i, :-1]])
                    P_pr = A_mat[j] @ P_prev[i] @ A_mat[j].T
                    P_pr[0, 0] += self.Q[j]

                    # — 観測更新 —
                    S = C_t @ P_pr @ C_t.T + R_t
                    invS = np.linalg.inv(S)
                    res = np.asarray(obs_vals) - C_t @ F_pr
                    lh = (np.exp(-0.5 * res @ invS @ res) /
                          np.sqrt((2 * np.pi) ** k * np.linalg.det(S)))
                    K = P_pr @ C_t.T @ invS
                    F_up = F_pr + K @ res
                    P_up = (I_L - K @ C_t) @ P_pr

                    w_upd = w_pred * lh

                    w_ij[i, j] = w_upd
                    F_ij[i, j] = F_up
                    P_ij[i, j] = P_up
                    cov_ij[i, j] = P_up[0, 1]

            # ----------- mixture collapse -----------
            total = w_ij.sum()
            xi[t] = w_ij / total
            gamma[t] = xi[t].sum(axis=0)
            for j in range(M):
                wj = w_ij[:, j]; W = wj.sum()
                F_filt[t, j] = (wj[:, None] * F_ij[:, j]).sum(axis=0) / W
                diff = F_ij[:, j] - F_filt[t, j]
                P_filt[t, j] = (wj[:, None, None] *
                                (P_ij[:, j] + diff[:, :, None] * diff[:, None, :])
                                ).sum(axis=0) / W
                P_lag[t, j] = (wj * cov_ij[:, j]).sum() / W

            F_prev, P_prev, w_prev = F_filt[t].copy(), P_filt[t].copy(), gamma[t].copy()

        return gamma, xi, F_filt, P_filt, P_lag

    # ===================================================================
    #                           M‑step
    # ===================================================================
    def _maximization(self, y_D, y_M, dates, gamma, xi, F, P, P_lag):
        T, n_D = y_D.shape
        n_M = y_M.shape[1]
        M = self.M

        # ---- (1) π ----
        for i in range(M):
            den = gamma[:-1, i].sum()
            self.pi[i] = xi[1:, i].sum(axis=0) / den

        # ---- (2) daily loadings / variances ----
        num_cD = np.zeros(n_D); den_cD = 0.0
        num_RD = np.zeros(n_D)
        for t in range(T):
            for j in range(M):
                f0 = F[t, j, 0]
                den_part = f0 ** 2 + P[t, j, 0, 0]
                den_cD += gamma[t, j] * den_part
                for d in range(n_D):
                    num_cD[d] += gamma[t, j] * y_D[t, d] * f0
                    num_RD[d] += gamma[t, j] * ((y_D[t, d] - self.c_D[d] * f0) ** 2 +
                                                self.c_D[d] ** 2 * P[t, j, 0, 0])
        self.c_D = num_cD / den_cD
        self.R_D = num_RD / gamma.sum()

        # ---- (3) monthly loadings / variances ----
        num_cM = np.zeros(n_M); den_cM = np.zeros(n_M); num_RM = np.zeros(n_M)
        obs_cnt = np.zeros(n_M, dtype=int)

        for t in range(T):
            if np.isnan(y_M[t]).all():
                continue
            N = self._month_length(dates[t])
            w = np.zeros(self.L); w[:N] = 1.0 / N
            for j in range(M):
                agg = w @ F[t, j]
                var_agg = w @ P[t, j] @ w
                common_den = gamma[t, j] * (agg ** 2 + var_agg)
                for k in range(n_M):
                    if np.isnan(y_M[t, k]):
                        continue
                    num_cM[k] += gamma[t, j] * y_M[t, k] * agg
                    den_cM[k] += common_den
                    num_RM[k] += gamma[t, j] * ((y_M[t, k] - self.c_M[k] * agg) ** 2 +
                                                self.c_M[k] ** 2 * var_agg)
                    obs_cnt[k] += 1

        with np.errstate(divide="ignore", invalid="ignore"):
            self.c_M = np.where(den_cM > 0, num_cM / den_cM, self.c_M)
            self.R_M = np.where(obs_cnt > 0,
                                num_RM / (gamma.sum() * obs_cnt / T),
                                self.R_M)

        # ---- (4) state params μ, a, Q ----
        for j in range(M):
            w = gamma[1:, j]; Nj = w.sum()
            if Nj < 1e-10:
                continue
            S0 = (w * F[1:, j, 0]).sum()
            S1 = (w * F[:-1, j, 0]).sum()
            S01 = (w * (F[1:, j, 0] * F[:-1, j, 0] + P_lag[1:, j])).sum()
            S11 = (w * (F[:-1, j, 0] ** 2 + P[:-1, j, 0, 0])).sum()
            self.mu[j], self.a[j] = np.linalg.solve(
                np.array([[Nj, S1], [S1, S11]]), np.array([S0, S01]))
            numQ = 0.0
            for t in range(1, T):
                v = F[t, j, 0] - self.mu[j] - self.a[j] * F[t - 1, j, 0]
                numQ += gamma[t, j] * (
                    v ** 2 + P[t, j, 0, 0] +
                    self.a[j] ** 2 * P[t - 1, j, 0, 0] -
                    2 * self.a[j] * P_lag[t, j])
            self.Q[j] = numQ / Nj


    # -------------------------------------------------------------------
    #                       ユーティリティ
    # -------------------------------------------------------------------
    @staticmethod
    def _month_length(ts):
        """datetime → その月の日数"""
        start = ts - pd.offsets.MonthBegin(n=1)
        return (ts - start).days

# -----------------------------------------------------------------
#         シミュレーション：  Daily (2 列) + Monthly (3 列)
# -----------------------------------------------------------------
np.random.seed(1)
dates = pd.date_range("2021-01-01", "2022-12-31", freq="D")
T = len(dates)

# ---- 真のパラメータ (例) ----
Mtrue = 2
pi_true  = np.array([[0.95, 0.05], [0.04, 0.96]])
mu_true  = np.array([-0.3, 0.5])
a_true   = np.array([0.97, 0.65])
Q_true   = np.array([0.06, 0.11])

c_D_true = np.array([1.0, -0.8])            # 2 daily series
c_M_true = np.array([0.9, -0.7, 0.5])       # 3 monthly series
R_D_true = np.array([0.4, 0.35])
R_M_true = np.array([0.3, 0.25, 0.28])

# ---- データ生成 ----
s = np.zeros(T, dtype=int)
f = np.zeros(T)
y_D = np.zeros((T, 2))
y_M = np.full((T, 3), np.nan)

for t in range(1, T):
    s[t] = np.random.choice(2, p=pi_true[s[t - 1]])
    j = s[t]
    f[t] = mu_true[j] + a_true[j] * f[t - 1] + np.sqrt(Q_true[j]) * np.random.randn()
    y_D[t] = c_D_true * f[t] + np.sqrt(R_D_true) * np.random.randn(2)

    if dates[t].is_month_end:
        N = (dates[t] - (dates[t]-pd.offsets.MonthBegin(1))).days
        avg_f = f[t - N + 1:t + 1].mean()
        y_M[t] = c_M_true * avg_f + np.sqrt(R_M_true) * np.random.randn(3)

# ---- 推定 ----
model = RS_MixedFreqDFM(L=31)
gamma = model.fit(y_D, y_M, dates, s=s)

# ---- 結果可視化 ----
plt.figure(figsize=(10, 4))
plt.plot(s, label="True Regime", lw=0.6)
plt.plot(gamma[:, 1], label="Est. Prob(Regime 1)", lw=0.8)
plt.legend(); plt.tight_layout(); plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


class RS_MixedFreqDFM:
    """
    Regime-Switching Dynamic Factor Model
        • latent daily factor f_t  (AR(1))
        • n_D  daily series:           y_D[:,d]  = c_D[d] * f_t + ε_t,d
        • n_M  monthly series (month-end only):
              y_M[:,k]  = c_M[k] * (1/N) Σ_{d=0}^{N-1} f_{t-d} + ε_t,k
    状態ベクトル F_t = [f_t, f_{t-1}, …, f_{t-L+1}]ᵀ  (L ≥ 31 推奨)
    EM アルゴリズム（修正カルマンフィルタ / スムーザ）で推定
    ※プログラムへの修正は最小限にしてください
    """

    def __init__(self, n_regimes=2, L=31, max_iter=100, tol=1e-4):
        self.M = n_regimes
        self.L = L                # ラグ長（31 で「月平均」可）
        self.max_iter = max_iter
        self.tol = tol

        # --- レジーム別状態パラメータ（初期値は fit() で上書き） ---
        self.pi = np.full((self.M, self.M), 1 / self.M)
        self.mu = np.zeros(self.M)
        self.a  = np.zeros(self.M)
        self.Q  = np.ones(self.M)

        # --- 観測パラメータ（サイズは fit() 内で決定） ---
        self.c_D = None         # shape (n_D,)
        self.c_M = None         # shape (n_M,)
        self.R_D = None         # shape (n_D,)
        self.R_M = None         # shape (n_M,)

    # ===================================================================
    #                              EM 外部ループ
    # ===================================================================
    def fit(self, y_D, y_M, dates, s=None):  # レジームシークエンスをオプション引数に追加
        """
        Parameters
        ----------
        y_D : (T, n_D) ndarray            毎日観測（欠測なし）
        y_M : (T, n_M) ndarray            月末のみ値、他は NaN
        dates : pandas.DatetimeIndex       日次インデックス (len = T)
        s : (T,) int array, optional       既知のレジームシークエンス

        Returns
        -------
        gamma : (T, M) ndarray            平滑化レジーム確率
        """
        # 既知シークエンスが指定された場合は one-hot を返す
        if s is not None:
            T = len(s)
            gamma = np.zeros((T, self.M))
            gamma[np.arange(T), s] = 1
            return gamma

        y_D = np.asarray(y_D)
        y_M = np.asarray(y_M)
        T, n_D = y_D.shape
        _, n_M = y_M.shape
        self.n_D, self.n_M = n_D, n_M

        # ---------- パラメータ初期化 ----------
        self.pi[:] = 0.9 * np.eye(self.M) + 0.1 / (self.M - 1) * (1 - np.eye(self.M))
        self.mu[:] = np.linspace(-0.5, 0.5, self.M)
        self.a[:] = 0.6
        self.Q[:] = 0.3

        self.c_D = np.ones(n_D)
        self.c_M = np.ones(n_M)
        self.R_D = np.ones(n_D) * 0.5
        self.R_M = np.ones(n_M) * 0.5

        # ---------- EM ループ ----------
        gamma = np.full((T, self.M), 1 / self.M)
        xi = np.full((T, self.M, self.M), 1 / self.M ** 2)
        ll_old = -np.inf

        for _ in range(self.max_iter):
            gamma, xi, F, P, P_lag = self._expectation(y_D, y_M, dates, gamma)
            self._maximization(y_D, y_M, dates, gamma, xi, F, P, P_lag)
            ll = -np.sum(gamma * np.log(gamma + 1e-16))
            if abs(ll - ll_old) < self.tol:
                break
            ll_old = ll
        return gamma

    # ===================================================================
    #                           E-step
    # ===================================================================
    def _expectation(self, y_D, y_M, dates, gamma_prev):
        T = len(y_D)
        L, M = self.L, self.M
        # n_D, n_M は入力データから取得（known モード後にも期待値計算可）
        n_D = y_D.shape[1]
        n_M = y_M.shape[1]

        F_filt = np.zeros((T, M, L))
        P_filt = np.zeros((T, M, L, L))
        P_lag  = np.zeros((T, M))
        xi     = np.zeros((T, M, M))
        gamma  = np.zeros((T, M))

        # mixture の初期値
        F_prev = np.zeros((M, L))
        P_prev = np.tile(np.eye(L) * 10.0, (M, 1, 1))
        w_prev = np.full(M, 1 / M)
        I_L = np.eye(L)

        # regime‑別遷移行列
        A_mat = np.zeros((M, L, L))
        for j in range(M):
            A = np.zeros((L, L))
            A[0, 0] = self.a[j]
            A[1:, :-1] = np.eye(L - 1)
            A_mat[j] = A

        for t in range(T):
            # ---------------- 観測行列 C_t ----------------
            obs_rows = []
            obs_vals = []
            obs_var  = []

            #   ‑‑ daily rows
            for d in range(n_D):
                row = np.zeros(L)
                row[0] = self.c_D[d]
                obs_rows.append(row)
                obs_vals.append(y_D[t, d])
                obs_var.append(self.R_D[d])

            #   ‑‑ month‑end rows
            if not np.isnan(y_M[t]).all():           # 少なくとも一列は値あり
                N = self._month_length(dates[t])
                w = np.zeros(L); w[:N] = 1.0 / N
                for k in range(n_M):
                    if np.isnan(y_M[t, k]):          # 欠測は飛ばす
                        continue
                    obs_rows.append(self.c_M[k] * w)
                    obs_vals.append(y_M[t, k])
                    obs_var.append(self.R_M[k])

            C_t = np.vstack(obs_rows)                     # k×L
            R_t = np.diag(obs_var)                        # k×k
            k = C_t.shape[0]

            # ---------------- 各パス毎の更新 ----------------
            w_ij   = np.zeros((M, M))
            F_ij   = np.zeros((M, M, L))
            P_ij   = np.zeros((M, M, L, L))
            cov_ij = np.zeros((M, M))                     # cov(f_t, f_{t-1})

            for i in range(M):
                for j in range(M):
                    w_pred = w_prev[i] * self.pi[i, j]

                    # — 状態予測 —
                    F_pr = self.mu[j] + self.a[j] * F_prev[i, 0]
                    F_pr = np.concatenate([[F_pr], F_prev[i, :-1]])
                    P_pr = A_mat[j] @ P_prev[i] @ A_mat[j].T
                    P_pr[0, 0] += self.Q[j]

                    # — 観測更新 —
                    S = C_t @ P_pr @ C_t.T + R_t
                    invS = np.linalg.inv(S)
                    res = np.asarray(obs_vals) - C_t @ F_pr
                    lh = (np.exp(-0.5 * res @ invS @ res) /
                          np.sqrt((2 * np.pi) ** k * np.linalg.det(S)))
                    K = P_pr @ C_t.T @ invS
                    F_up = F_pr + K @ res
                    P_up = (I_L - K @ C_t) @ P_pr

                    w_upd = w_pred * lh

                    w_ij[i, j] = w_upd
                    F_ij[i, j] = F_up
                    P_ij[i, j] = P_up
                    cov_ij[i, j] = P_up[0, 1]

            # ----------- mixture collapse -----------
            total = w_ij.sum()
            xi[t] = w_ij / total
            gamma[t] = xi[t].sum(axis=0)
            for j in range(M):
                wj = w_ij[:, j]; W = wj.sum()
                F_filt[t, j] = (wj[:, None] * F_ij[:, j]).sum(axis=0) / W
                diff = F_ij[:, j] - F_filt[t, j]
                P_filt[t, j] = (wj[:, None, None] *
                                (P_ij[:, j] + diff[:, :, None] * diff[:, None, :])
                                ).sum(axis=0) / W
                P_lag[t, j] = (wj * cov_ij[:, j]).sum() / W

            F_prev, P_prev, w_prev = F_filt[t].copy(), P_filt[t].copy(), gamma[t].copy()

        return gamma, xi, F_filt, P_filt, P_lag

    # ===================================================================
    #                           M-step
    # ===================================================================
    def _maximization(self, y_D, y_M, dates, gamma, xi, F, P, P_lag):
        T, n_D = y_D.shape
        n_M = y_M.shape[1]
        M = self.M

        # ---- (1) π ----
        for i in range(M):
            den = gamma[:-1, i].sum()
            self.pi[i] = xi[1:, i].sum(axis=0) / den

        # ---- (2) daily loadings / variances ----
        num_cD = np.zeros(n_D); den_cD = 0.0
        num_RD = np.zeros(n_D)
        for t in range(T):
            for j in range(M):
                f0 = F[t, j, 0]
                den_part = f0 ** 2 + P[t, j, 0, 0]
                den_cD += gamma[t, j] * den_part
                for d in range(n_D):
                    num_cD[d] += gamma[t, j] * y_D[t, d] * f0
                    num_RD[d] += gamma[t, j] * ((y_D[t, d] - self.c_D[d] * f0) ** 2 +
                                                self.c_D[d] ** 2 * P[t, j, 0, 0])
        self.c_D = num_cD / den_cD
        self.R_D = num_RD / gamma.sum()

        # ---- (3) monthly loadings / variances ----
        num_cM = np.zeros(n_M); den_cM = np.zeros(n_M); num_RM = np.zeros(n_M)
        obs_cnt = np.zeros(n_M, dtype=int)

        for t in range(T):
            if np.isnan(y_M[t]).all():
                continue
            N = self._month_length(dates[t])
            w = np.zeros(self.L); w[:N] = 1.0 / N
            for j in range(M):
                agg = w @ F[t, j]
                var_agg = w @ P[t, j] @ w
                common_den = gamma[t, j] * (agg ** 2 + var_agg)
                for k in range(n_M):
                    if np.isnan(y_M[t, k]):
                        continue
                    num_cM[k] += gamma[t, j] * y_M[t, k] * agg
                    den_cM[k] += common_den
                    num_RM[k] += gamma[t, j] * ((y_M[t, k] - self.c_M[k] * agg) ** 2 +
                                                self.c_M[k] ** 2 * var_agg)
                    obs_cnt[k] += 1

        with np.errstate(divide="ignore", invalid="ignore"):
            self.c_M = np.where(den_cM > 0, num_cM / den_cM, self.c_M)
            self.R_M = np.where(obs_cnt > 0,
                                num_RM / (gamma.sum() * obs_cnt / T),
                                self.R_M)

        # ---- (4) state params μ, a, Q ----
        for j in range(M):
            w = gamma[1:, j]; Nj = w.sum()
            if Nj < 1e-10:
                continue
            S0 = (w * F[1:, j, 0]).sum()
            S1 = (w * F[:-1, j, 0]).sum()
            S01 = (w * (F[1:, j, 0] * F[:-1, j, 0] + P_lag[1:, j])).sum()
            S11 = (w * (F[:-1, j, 0] ** 2 + P[:-1, j, 0, 0])).sum()
            self.mu[j], self.a[j] = np.linalg.solve(
                np.array([[Nj, S1], [S1, S11]]), np.array([S0, S01]))
            numQ = 0.0
            for t in range(1, T):
                v = F[t, j, 0] - self.mu[j] - self.a[j] * F[t - 1, j, 0]
                numQ += gamma[t, j] * (
                    v ** 2 + P[t, j, 0, 0] +
                    self.a[j] ** 2 * P[t - 1, j, 0, 0] -
                    2 * self.a[j] * P_lag[t, j])
            self.Q[j] = numQ / Nj

    # -------------------------------------------------------------------
    #                       ユーティリティ
    # -------------------------------------------------------------------
    @staticmethod
    def _month_length(ts):
        """datetime → その月の日数"""
        start = ts - pd.offsets.MonthBegin(n=1)
        return (ts - start).days

# -----------------------------------------------------------------
#         シミュレーション：  Daily (2 列) + Monthly (3 列)
# -----------------------------------------------------------------
np.random.seed(1)
dates = pd.date_range("2021-01-01", "2022-12-31", freq="D")
T = len(dates)

# ---- 真のパラメータ (例) ----
Mtrue = 2
pi_true  = np.array([[0.95, 0.05], [0.04, 0.96]])
mu_true  = np.array([-0.3, 0.5])
a_true   = np.array([0.97, 0.65])
Q_true   = np.array([0.06, 0.11])

c_D_true = np.array([1.0, -0.8])            # 2 daily series
c_M_true = np.array([0.9, -0.7, 0.5])       # 3 monthly series
R_D_true = np.array([0.4, 0.35])
R_M_true = np.array([0.3, 0.25, 0.28])

# ---- データ生成 ----
s = np.zeros(T, dtype=int)
f = np.zeros(T)
y_D = np.zeros((T, 2))
y_M = np.full((T, 3), np.nan)

for t in range(1, T):
    s[t] = np.random.choice(2, p=pi_true[s[t - 1]])
    j = s[t]
    f[t] = mu_true[j] + a_true[j] * f[t - 1] + np.sqrt(Q_true[j]) * np.random.randn()
    y_D[t] = c_D_true * f[t] + np.sqrt(R_D_true) * np.random.randn(2)

    if dates[t].is_month_end:
        N = (dates[t] - (dates[t]-pd.offsets.MonthBegin(1))).days
        avg_f = f[t - N + 1:t + 1].mean()
        y_M[t] = c_M_true * avg_f + np.sqrt(R_M_true) * np.random.randn(3)

# ---- 推定 ----
model = RS_MixedFreqDFM(L=31)
gamma = model.fit(y_D, y_M, dates, s=s)

# ---- 結果可視化 ----
plt.figure(figsize=(10, 4))
plt.plot(s, label="True Regime", lw=0.6)
plt.plot(gamma[:, 1], label="Est. Prob(Regime 1)", lw=0.8)
plt.legend(); plt.tight_layout(); plt.show()

# ---- １ステップ先予測（最小限修正） ----
# t = 最後の観測インデックス
_, _, F_filt, P_filt, _ = model._expectation(y_D, y_M, dates, gamma)
t = T - 1

gamma_t = gamma[t]                # shape (M,)
f_tj    = F_filt[t,:,0]          # レジームごとの因子
P_tj    = P_filt[t,:,0,0]        # 分散

# 各レジームの１ステップ先因子予測
f_pred_j = model.mu + model.a * f_tj  # shape (M,)
# 混合予測
f_pred   = (gamma_t * f_pred_j).sum()

# 日次観測の予測
yD_pred  = model.c_D * f_pred     # shape (n_D,)

# 月次観測の予測（月末なら）
L        = model.L
avg_pred = np.mean([f_pred] + list(f_tj[1:L]))
yM_pred  = model.c_M * avg_pred   # shape (n_M,)

print("1ステップ先因子予測 f_pred:", f_pred)
print("日次予測 yD_pred:", yD_pred)
print("月次予測 yM_pred:", yM_pred)
```

```python
import numpy as np
import matplotlib.pyplot as plt

class RegimeSwitchingDFM:
    """
    レジームスイッチング多因子動的ファクターモデル（AR(1)因子）を
    EMアルゴリズムで推定

    観測方程式:
        Y_t = c^T f_t + \u03B5_t,   \u03B5_t ~ N(0, R)
    状態方程式:
        f_t = \u03BC_{s_t} + A_{s_t} f_{t-1} + v_t,   v_t ~ N(0, Q_{s_t})
    レジーム遷移:
        Pr(s_t=j | s_{t-1}=i) = \u03C0_{ij}
    """
    def __init__(self, n_regimes=2, n_factors=1, max_iter=100, tol=1e-4):
        self.M = n_regimes       # レジーム数
        self.k = n_factors       # 因子数
        self.max_iter = max_iter
        self.tol = tol
        # パラメータ初期化
        self.pi = np.full((self.M, self.M), 1/self.M)
        self.mu = np.zeros((self.M, self.k))
        self.a = np.zeros((self.M, self.k))
        self.Q = np.ones((self.M, self.k))
        self.c = np.ones(self.k)
        self.R = 1.0

    def fit(self, y):
        T = len(y)
        # 初期推定値
        self.pi[:] = 0.9*np.eye(self.M) + 0.1*(np.ones((self.M, self.M)) - np.eye(self.M))/(self.M-1)
        self.mu[:] = np.linspace(-1, 1, self.M*self.k).reshape(self.M, self.k)
        self.a[:] = 0.5
        self.Q[:] = 1.0
        self.c[:] = 1.0
        self.R = 1.0

        gamma = np.full((T, self.M), 1/self.M)
        xi = np.full((T, self.M, self.M), 1/(self.M**2))
        ll_old = -np.inf

        for it in range(self.max_iter):
            gamma, xi, f_filt, P_filt, P_lag = self._expectation(y, gamma)
            self._maximization(y, gamma, xi, f_filt, P_filt, P_lag)
            ll = self._log_likelihood(gamma)
            if abs(ll - ll_old) < self.tol:
                break
            ll_old = ll
        return gamma

    def _expectation(self, y, gamma_prev):
        T = len(y)
        M, k = self.M, self.k
        f_pred = np.zeros((T, M, k))
        P_pred = np.zeros((T, M, k))
        f_filt = np.zeros((T, M, k))
        P_filt = np.zeros((T, M, k))
        P_lag  = np.zeros((T, M, k))
        xi = np.zeros((T, M, M))
        gamma = np.zeros((T, M))

        f_prev = np.zeros((M, k))
        P_prev = np.ones((M, k)) * 1e1
        w_prev = np.full(M, 1/M)

        for t in range(T):
            w_ij = np.zeros((M, M))
            f_ij = np.zeros((M, M, k))
            P_ij = np.zeros((M, M, k))
            cov_ij = np.zeros((M, M, k))

            for i in range(M):
                for j in range(M):
                    w_pred = w_prev[i] * self.pi[i, j]
                    f_pr = self.mu[j] + self.a[j] * f_prev[i]
                    P_pr = self.a[j]**2 * P_prev[i] + self.Q[j]

                    # 観測予測
                    m = np.dot(self.c, f_pr)
                    S = np.dot(self.c**2, P_pr) + self.R

                    # 観測尤度
                    L = np.exp(-0.5 * (y[t] - m)**2 / S) / np.sqrt(2*np.pi*S)
                    w_upd = w_pred * L

                    # カルマン更新
                    K = (P_pr * self.c) / S
                    f_up = f_pr + K * (y[t] - m)
                    P_up = P_pr - K * self.c * P_pr
                    cov = (1 - K*self.c) * self.a[j] * P_prev[i]

                    w_ij[i, j] = w_upd
                    f_ij[i, j] = f_up
                    P_ij[i, j] = P_up
                    cov_ij[i, j] = cov

            total = w_ij.sum()
            xi[t] = w_ij / total
            gamma[t] = xi[t].sum(axis=0)

            for j in range(M):
                weights = w_ij[:, j]
                sum_w = weights.sum()
                f_filt[t, j] = (weights[:, None] * f_ij[:, j]).sum(axis=0) / sum_w
                P_filt[t, j] = (weights[:, None] * (P_ij[:, j] + (f_ij[:, j] - f_filt[t, j])**2)).sum(axis=0) / sum_w
                P_lag[t, j]  = (weights[:, None] * cov_ij[:, j]).sum(axis=0) / sum_w

            f_prev = f_filt[t].copy()
            P_prev = P_filt[t].copy()
            w_prev = gamma[t].copy()

        return gamma, xi, f_filt, P_filt, P_lag

    def _maximization(self, y, gamma, xi, f_filt, P_filt, P_lag):
        T = len(y)
        M, k = self.M, self.k

        # π
        self.pi = xi[1:].sum(axis=0) / gamma[:-1].sum(axis=0)[:, None]

        # c, R
        num_c = np.zeros(k)
        den_c = np.zeros(k)
        num_R = 0.0
        for t in range(T):
            for j in range(M):
                num_c += gamma[t, j] * y[t] * f_filt[t, j]
                den_c += gamma[t, j] * (f_filt[t, j]**2 + P_filt[t, j])
        self.c = num_c / den_c

        for t in range(T):
            for j in range(M):
                pred = np.dot(self.c, f_filt[t, j])
                num_R += gamma[t, j] * ((y[t] - pred)**2 + np.dot(self.c**2, P_filt[t, j]))
        self.R = num_R / gamma.sum()

        # μ, a, Q
        for j in range(M):
            w = gamma[1:, j]
            N_j = w.sum()
            for dim in range(k):
                S0 = (w * f_filt[1:, j, dim]).sum()
                S1 = (w * f_filt[:-1, j, dim]).sum()
                S01 = (w * (f_filt[1:, j, dim] * f_filt[:-1, j, dim] + P_lag[1:, j, dim])).sum()
                S11 = (w * (f_filt[:-1, j, dim]**2 + P_filt[:-1, j, dim])).sum()
                Mmat = np.array([[N_j, S1], [S1, S11]])
                bvec = np.array([S0, S01])
                sol = np.linalg.solve(Mmat, bvec)
                self.mu[j, dim], self.a[j, dim] = sol

                num_Q = 0.0
                for t in range(1, T):
                    v = f_filt[t, j, dim] - self.mu[j, dim] - self.a[j, dim] * f_filt[t-1, j, dim]
                    num_Q += gamma[t, j] * (v**2 + P_filt[t, j, dim] + self.a[j, dim]**2 * P_filt[t-1, j, dim] - 2 * self.a[j, dim] * P_lag[t, j, dim])
                self.Q[j, dim] = num_Q / N_j

    def _log_likelihood(self, gamma):
        return -np.sum(gamma * np.log(gamma + 1e-16))

# example usage remains for single-factor case
if __name__ == '__main__':
    np.random.seed(1)
    T = 300
    true_pi = np.array([[0.9, 0.1], [0.1, 0.9]])
    true_mu = np.array([-0.5, 0.5])
    true_a  = np.array([0.95, 0.5])
    true_Q  = np.array([0.1, 0.1])
    true_c  = 1.0; true_R = 0.5
    s_true = np.zeros(T, dtype=int)
    y = np.zeros(T); f = np.zeros(T)
    for t in range(1, T):
        s_true[t] = np.random.choice(2, p=true_pi[s_true[t-1]])
        j = s_true[t]
        f[t] = true_mu[j] + true_a[j]*f[t-1] + np.sqrt(true_Q[j])*np.random.randn()
        y[t] = true_c*f[t] + np.sqrt(true_R)*np.random.randn()

    model = RegimeSwitchingDFM(n_regimes=2, n_factors=1)
    gamma = model.fit(y)

    plt.figure(figsize=(10,5))
    plt.plot(s_true, label='True Regime')
    plt.plot(gamma[:,1], label='Est Prob Regime1')
    plt.xlabel('Time')
    plt.legend(); plt.tight_layout(); plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class RegimeSwitchingDFM:
    """
    レジームスイッチング多因子動的ファクターモデル（AR(1)因子）を
    EMアルゴリズムで推定

    観測方程式:
        Y_t = C_{s_t}^T f_t + ε_t,   ε_t ~ N(0, R)
    状態方程式:
        f_t = μ_{s_t} + A_{s_t} f_{t-1} + v_t,   v_t ~ N(0, Q_{s_t})
    レジーム遷移:
        Pr(s_t=j | s_{t-1}=i) = π_{ij}
    """
    def __init__(self, n_regimes=2, n_factors=1, n_obs=1, max_iter=100, tol=1e-4):
        self.M = n_regimes       # レジーム数
        self.k = n_factors       # 因子数
        self.p = n_obs           # 観測次元
        self.max_iter = max_iter
        self.tol = tol
        # パラメータ初期化
        self.pi = np.full((self.M, self.M), 1/self.M)
        self.mu = np.zeros((self.M, self.k))
        self.a = np.zeros((self.M, self.k))
        self.Q = np.ones((self.M, self.k))
        self.C = np.ones((self.M, self.p, self.k))  # レジーム×観測次元×因子数
        self.R = np.ones(self.p)                  # 観測ノイズ分散(対角)

    def fit(self, y):
        # y: shape (T, p)
        T = y.shape[0]
        # 初期推定値
        self.pi[:] = 0.9*np.eye(self.M) + 0.1*(np.ones((self.M, self.M)) - np.eye(self.M))/(self.M-1)
        self.mu[:] = np.linspace(-1, 1, self.M*self.k).reshape(self.M, self.k)
        self.a[:] = 0.5
        self.Q[:] = 1.0
        self.C[:] = 1.0
        self.R[:] = 1.0

        gamma = np.full((T, self.M), 1/self.M)
        xi = np.full((T, self.M, self.M), 1/(self.M**2))
        ll_old = -np.inf

        for _ in range(self.max_iter):
            gamma, xi, f_filt, P_filt, P_lag = self._expectation(y, gamma)
            self._maximization(y, gamma, xi, f_filt, P_filt, P_lag)
            ll = self._log_likelihood(gamma)
            if abs(ll - ll_old) < self.tol:
                break
            ll_old = ll
        return gamma

    def _expectation(self, y, gamma_prev):
        T = y.shape[0]
        M, k, p = self.M, self.k, self.p
        f_filt = np.zeros((T, M, k))
        P_filt = np.zeros((T, M, k))
        P_lag  = np.zeros((T, M, k))
        xi = np.zeros((T, M, M))
        gamma = np.zeros((T, M))
        eps = 1e-8

        f_prev = np.zeros((M, k))
        P_prev = np.ones((M, k)) * 1e1
        w_prev = np.full(M, 1/M)

        for t in range(T):
            w_ij = np.zeros((M, M))
            f_ij = np.zeros((M, M, k))
            P_ij = np.zeros((M, M, k))
            cov_ij = np.zeros((M, M, k))

            for i in range(M):
                for j in range(M):
                    w_pred = w_prev[i] * self.pi[i, j]
                    f_pr = self.mu[j] + self.a[j] * f_prev[i]
                    P_pr = self.a[j]**2 * P_prev[i] + self.Q[j]

                    # 観測予測
                    m = self.C[j].dot(f_pr)  # (p,)
                    S = (self.C[j]**2 * P_pr).sum(axis=1) + self.R  # (p,)
                    S_safe = np.maximum(S, eps)
                    resid = y[t] - m

                    # 観測尤度（独立次元として積）
                    L = np.prod(np.exp(-0.5 * resid**2 / S_safe) / np.sqrt(2*np.pi * S_safe))
                    w_upd = w_pred * L

                    # カルマン更新
                    K = (P_pr[:, None] * self.C[j].T) / S_safe[None, :]
                    f_up = f_pr + K.dot(resid)
                    P_up = P_pr * (1 - (self.C[j]**2 * P_pr / S_safe[:, None]).sum(axis=0))
                    cov = (1 - (self.C[j]**2 * P_pr / S_safe[:, None]).sum(axis=0)) * self.a[j] * P_prev[i]

                    w_ij[i, j] = w_upd
                    f_ij[i, j] = f_up
                    P_ij[i, j] = P_up
                    cov_ij[i, j] = cov

            total = w_ij.sum()
            xi[t] = w_ij / total
            gamma[t] = xi[t].sum(axis=0)

            for j in range(M):
                w = w_ij[:, j]
                sum_w = w.sum()
                f_filt[t, j] = (w[:, None] * f_ij[:, j]).sum(axis=0) / sum_w
                P_filt[t, j] = (w[:, None] * (P_ij[:, j] + (f_ij[:, j] - f_filt[t, j])**2)).sum(axis=0) / sum_w
                P_lag[t, j]  = (w[:, None] * cov_ij[:, j]).sum(axis=0) / sum_w

            f_prev = f_filt[t].copy()
            P_prev = P_filt[t].copy()
            w_prev = gamma[t].copy()

        return gamma, xi, f_filt, P_filt, P_lag

    def _maximization(self, y, gamma, xi, f_filt, P_filt, P_lag):
        T = y.shape[0]
        M, k, p = self.M, self.k, self.p

        # π
        self.pi = xi[1:].sum(axis=0) / gamma[:-1].sum(axis=0)[:, None]

        # C, R
        num_C = np.zeros((M, p, k))
        den_C = np.zeros((M, k))
        num_R = np.zeros(p)
        for t in range(T):
            for j in range(M):
                num_C[j] += gamma[t, j] * np.outer(y[t], f_filt[t, j])
                den_C[j] += gamma[t, j] * (f_filt[t, j]**2 + P_filt[t, j])
        self.C = num_C / den_C[:, None, :]

        for t in range(T):
            for j in range(M):
                m = self.C[j].dot(f_filt[t, j])
                num_R += gamma[t, j] * ((y[t] - m)**2 + (self.C[j]**2 * P_filt[t, j]).sum(axis=1))
        self.R = num_R / gamma.sum()

        # μ, a, Q
        for j in range(M):
            w = gamma[1:, j]
            N_j = w.sum()
            for dim in range(k):
                S0 = (w * f_filt[1:, j, dim]).sum()
                S1 = (w * f_filt[:-1, j, dim]).sum()
                S01 = (w * (f_filt[1:, j, dim] * f_filt[:-1, j, dim] + P_lag[1:, j, dim])).sum()
                S11 = (w * (f_filt[:-1, j, dim]**2 + P_filt[:-1, j, dim])).sum()
                Mmat = np.array([[N_j, S1], [S1, S11]])
                bvec = np.array([S0, S01])
                sol = np.linalg.solve(Mmat, bvec)
                self.mu[j, dim], self.a[j, dim] = sol

                num_Q = 0.0
                for t in range(1, T):
                    v = f_filt[t, j, dim] - self.mu[j, dim] - self.a[j, dim] * f_filt[t-1, j, dim]
                    num_Q += gamma[t, j] * (v**2 + P_filt[t, j, dim] + self.a[j, dim]**2 * P_filt[t-1, j, dim] - 2 * self.a[j, dim] * P_lag[t, j, dim])
                self.Q[j, dim] = num_Q / N_j

    def _log_likelihood(self, gamma):
        return -np.sum(gamma * np.log(gamma + 1e-16))


if __name__ == '__main__':
    # サンプル実行
    np.random.seed(1)
    T = 300
    p = 2         # 観測次元
    M = 2         # レジーム数
    k = 1         # 因子数

    # 真のパラメータ
    true_pi = np.array([[0.9, 0.1], [0.1, 0.9]])
    true_mu = np.array([[-0.5], [0.5]])
    true_a  = np.array([[0.95], [0.5]])
    true_Q  = np.array([[0.1], [0.1]])
    true_C  = np.array([[[1.0], [2.0]], [[1.5], [0.5]]])  # shape (M, p, k)
    true_R  = np.array([0.5, 0.7])

    # データ生成
    s_true = np.zeros(T, dtype=int)
    f = np.zeros((T, k))
    y = np.zeros((T, p))
    for t in range(1, T):
        s_true[t] = np.random.choice(M, p=true_pi[s_true[t-1]])
        j = s_true[t]
        f[t] = true_mu[j] + true_a[j] * f[t-1] + np.sqrt(true_Q[j]) * np.random.randn()
        y[t] = true_C[j].dot(f[t]) + np.sqrt(true_R) * np.random.randn(p)

    # モデル推定
    model = RegimeSwitchingDFM(n_regimes=M, n_factors=k, n_obs=p)
    gamma = model.fit(y)

    # プロット
    plt.figure(figsize=(10,5))
    plt.plot(s_true, label='True Regime')
    plt.plot(gamma[:,1], label='Estimated Prob Regime1')
    plt.xlabel('Time')
    plt.ylabel('Probability')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class RegimeSwitchingDFM:
    """
    レジームスイッチング単因子動的ファクターモデル（AR(1)因子）を
    EMアルゴリズムで推定

    観測方程式:
        Y_t = C_{s_t} f_t + ε_t,   ε_t ~ N(0, R_{s_t})
    状態方程式:
        f_t = μ_{s_t} + a_{s_t} f_{t-1} + v_t,   v_t ~ N(0, Q_{s_t})
    レジーム遷移:
        Pr(s_t=j | s_{t-1}=i) = π_{ij}
    """
    def __init__(self, n_regimes=2, n_obs=1, max_iter=100, tol=1e-4):
        self.M = n_regimes
        self.p = n_obs
        self.max_iter = max_iter
        self.tol = tol

        # パラメータ初期化
        self.pi = np.full((self.M, self.M), 1/self.M)
        self.mu = np.zeros(self.M)
        self.a = np.zeros(self.M)
        self.Q = np.ones(self.M)
        self.C = np.ones((self.M, self.p))
        self.R = np.array([np.eye(self.p) for _ in range(self.M)])

    def fit(self, y):
        y = np.asarray(y)
        if y.ndim == 1:
            y = y.reshape(-1,1)
        T, p = y.shape
        assert p == self.p, "Observation dimension mismatch."

        # 初期推定値
        self.pi[:] = 0.9*np.eye(self.M) + 0.1*(np.ones((self.M,self.M))-np.eye(self.M))/(self.M-1)
        self.mu[:] = np.linspace(-1,1,self.M)
        self.a[:] = 0.5
        self.Q[:] = 1.0
        self.C[:] = 1.0
        self.R[:] = np.array([np.eye(self.p) for _ in range(self.M)])

        gamma = np.full((T,self.M),1/self.M)
        xi    = np.full((T,self.M,self.M),1/(self.M**2))
        ll_old = -np.inf

        for it in range(self.max_iter):
            gamma, xi, f_filt, P_filt, P_lag = self._expectation(y, gamma)
            self._maximization(y, gamma, xi, f_filt, P_filt, P_lag)
            ll = self._log_likelihood(gamma)
            if abs(ll - ll_old) < self.tol:
                break
            ll_old = ll
        return gamma

    def _expectation(self, y, gamma_prev):
        T = y.shape[0]
        M = self.M

        f_filt = np.zeros((T,M))
        P_filt = np.zeros((T,M))
        P_lag  = np.zeros((T,M))
        xi     = np.zeros((T,M,M))
        gamma  = np.zeros((T,M))

        f_prev  = np.zeros(M)
        P_prev  = np.ones(M)*1e1
        w_prev  = np.full(M,1/M)

        for t in range(T):
            y_t = y[t]
            w_ij  = np.zeros((M,M))
            f_ij  = np.zeros((M,M))
            P_ij  = np.zeros((M,M))
            cov_ij= np.zeros((M,M))

            for i in range(M):
                for j in range(M):
                    w_pred = w_prev[i] * self.pi[i,j]
                    f_pr   = self.mu[j] + self.a[j] * f_prev[i]
                    P_pr   = self.a[j]**2 * P_prev[i] + self.Q[j]

                    Cj     = self.C[j]
                    Rj     = self.R[j]
                    S      = P_pr * np.outer(Cj, Cj) + Rj
                    diff   = y_t - Cj * f_pr
                    sign, logdet = np.linalg.slogdet(S)
                    invS   = np.linalg.inv(S)
                    logL   = -0.5 * (len(diff)*np.log(2*np.pi) + logdet + diff.T @ invS @ diff)
                    L      = np.exp(logL)

                    w_upd  = w_pred * L
                    K      = P_pr * (invS @ Cj)
                    f_up   = f_pr + K.dot(diff)
                    P_up   = P_pr - P_pr * (K.dot(Cj))
                    cov    = (1 - K.dot(Cj)) * self.a[j] * P_prev[i]

                    w_ij[i,j]   = w_upd
                    f_ij[i,j]   = f_up
                    P_ij[i,j]   = P_up
                    cov_ij[i,j] = cov

            total       = w_ij.sum()
            xi[t]       = w_ij / total
            gamma[t]    = xi[t].sum(axis=0)

            for j in range(M):
                weights       = w_ij[:,j]
                f_filt[t,j]  = (weights * f_ij[:,j]).sum() / weights.sum()
                P_filt[t,j]  = (weights * (P_ij[:,j] + (f_ij[:,j] - f_filt[t,j])**2)).sum() / weights.sum()
                P_lag[t,j]   = (weights * cov_ij[:,j]).sum() / weights.sum()

            f_prev = f_filt[t].copy()
            P_prev = P_filt[t].copy()
            w_prev = gamma[t].copy()

        return gamma, xi, f_filt, P_filt, P_lag

    def _maximization(self, y, gamma, xi, f_filt, P_filt, P_lag):
        T, p = y.shape
        M   = self.M

        # π の更新
        for i in range(M):
            for j in range(M):
                self.pi[i,j] = xi[1:,i,j].sum() / xi[:-1,i,:].sum()

        # 観測 C, R の更新
        for j in range(M):
            gamma_j  = gamma[:,j]
            W        = gamma_j.sum()
            sum_yf   = np.zeros(p)
            sum_f2   = 0.0
            sum_yyt  = np.zeros((p,p))
            for t in range(T):
                ft       = f_filt[t,j]
                Pt       = P_filt[t,j]
                y_t      = y[t]
                sum_yf  += gamma_j[t] * y_t * ft
                sum_f2  += gamma_j[t] * (ft**2 + Pt)
                sum_yyt += gamma_j[t] * np.outer(y_t, y_t)
            self.C[j] = sum_yf / sum_f2
            Cj        = self.C[j]
            term      = sum_yyt - np.outer(Cj, sum_yf) - np.outer(sum_yf, Cj) + sum_f2 * np.outer(Cj, Cj)
            self.R[j] = term / W

        # 状態 μ, a, Q の更新
        for j in range(M):
            gamma_j = gamma[1:,j]
            N_j     = gamma_j.sum()
            S0      = (gamma_j * f_filt[1:,j]).sum()
            S1      = (gamma_j * f_filt[:-1,j]).sum()
            S01     = (gamma_j * (f_filt[1:,j] * f_filt[:-1,j] + P_lag[1:,j])).sum()
            S11     = (gamma_j * (f_filt[:-1,j]**2 + P_filt[:-1,j])).sum()
            Mmat    = np.array([[N_j, S1],[S1, S11]])
            bvec    = np.array([S0, S01])
            sol     = np.linalg.solve(Mmat, bvec)
            self.mu[j], self.a[j] = sol

            num_Q = 0.0
            for t in range(1,T):
                v = f_filt[t,j] - self.mu[j] - self.a[j]*f_filt[t-1,j]
                num_Q += gamma[t,j] * (v**2 + P_filt[t,j] + self.a[j]**2 * P_filt[t-1,j] - 2*self.a[j]*P_lag[t,j])
            self.Q[j] = num_Q / N_j

    def _log_likelihood(self, gamma):
        return -np.sum(gamma * np.log(gamma + 1e-16))


if __name__ == '__main__':
    # サンプル実行
    np.random.seed(1)
    T = 300
    M = 2
    p = 2
    # 真のパラメータ
    true_pi = np.array([[0.9,0.1],[0.1,0.9]])
    true_mu = np.array([-0.5,0.5])
    true_a  = np.array([0.95,0.5])
    true_Q  = np.array([0.1,0.1])
    true_C  = np.array([[1.0, 0.5], [0.8, 1.2]])
    true_R  = np.array([np.eye(p)*0.5, np.eye(p)*0.8])

    # シミュレーション
    s_true = np.zeros(T, dtype=int)
    f = np.zeros(T)
    y = np.zeros((T,p))
    for t in range(1,T):
        s_true[t] = np.random.choice(M, p=true_pi[s_true[t-1]])
        j = s_true[t]
        f[t] = true_mu[j] + true_a[j]*f[t-1] + np.sqrt(true_Q[j])*np.random.randn()
        y[t] = true_C[j]*f[t] + np.random.multivariate_normal(np.zeros(p), true_R[j])

    # 推定
    model = RegimeSwitchingDFM(n_regimes=M, n_obs=p)
    gamma = model.fit(y)

    # プロット
    plt.figure(figsize=(10,5))
    plt.plot(s_true, label='True Regime', lw=1)
    plt.plot(gamma[:,1], label='Est Prob Regime1', lw=1)
    plt.xlabel('Time')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class RegimeSwitchingDFM:
    """
    混合頻度対応レジームスイッチングDFM
    - freq: 観測変数ごとの頻度リスト('m'=月次, 'q'=四半期)
    """
    def __init__(self, n_regimes=2, n_obs=1, freq=None, max_iter=100, tol=1e-4):
        self.M = n_regimes
        self.p = n_obs
        self.freq = freq
        self.max_iter = max_iter
        self.tol = tol

        # パラメータ初期化
        self.pi = np.full((self.M, self.M), 1/self.M)
        self.mu = np.zeros(self.M)
        self.a  = np.zeros(self.M)
        self.Q  = np.ones(self.M)
        self.C  = np.ones((self.M, self.p))
        self.R  = np.array([np.eye(self.p) for _ in range(self.M)])

    def fit(self, y):
        y = np.asarray(y, float)
        # 混合頻度前処理
        if self.freq is not None:
            y = self._preprocess_mixed_frequency(y, self.freq)

        if y.ndim == 1:
            y = y.reshape(-1,1)
        T, p = y.shape
        assert p == self.p, "Observation dimension mismatch."

        # 初期推定値設定
        self.pi[:] = 0.9*np.eye(self.M) + 0.1*(np.ones((self.M,self.M))-np.eye(self.M))/(self.M-1)
        self.mu[:] = np.linspace(-1,1,self.M)
        self.a[:]  = 0.5
        self.Q[:]  = 1.0
        self.C[:]  = 1.0
        self.R[:]  = np.array([np.eye(self.p) for _ in range(self.M)])

        gamma = np.full((T,self.M),1/self.M)
        xi    = np.full((T,self.M,self.M),1/(self.M**2))
        ll_old = -np.inf

        for it in range(self.max_iter):
            gamma, xi, f_filt, P_filt, P_lag = self._expectation(y, gamma)
            self._maximization(y, gamma, xi, f_filt, P_filt, P_lag)
            ll = self._log_likelihood(gamma)
            if abs(ll - ll_old) < self.tol:
                break
            ll_old = ll
        return gamma

    def _preprocess_mixed_frequency(self, y, freq):
        """
        四半期データを月次近似（Mariano–Murasawa 2003）
        Δ3 y_t ≃ (1/3) [1Δ1y_t + 2Δ1y_{t-1} + 3Δ1y_{t-2} + 2Δ1y_{t-3} + 1Δ1y_{t-4}]
        """
        y2 = y.copy()
        T, p = y.shape
        logy = np.log(y)
        d1 = np.vstack([np.zeros((1,p)), np.diff(logy, axis=0)])
        weight = [1,2,3,2,1]
        for j, f in enumerate(freq):
            if f == 'q':
                dm = np.zeros(T)
                for t in range(T):
                    wsum = 0.0
                    for k, w in enumerate(weight):
                        idx = t - k
                        if idx >= 0:
                            wsum += w * d1[idx, j]
                    dm[t] = wsum / 3.0
                y2[:, j] = np.exp(np.cumsum(dm))
        return y2

    def _expectation(self, y, gamma_prev):
        T = y.shape[0]
        M = self.M
        f_filt = np.zeros((T,M))
        P_filt = np.zeros((T,M))
        P_lag  = np.zeros((T,M))
        xi     = np.zeros((T,M,M))
        gamma  = np.zeros((T,M))
        f_prev  = np.zeros(M)
        P_prev  = np.ones(M)*1e1
        w_prev  = np.full(M,1/M)
        for t in range(T):
            y_t = y[t]
            w_ij  = np.zeros((M,M))
            f_ij  = np.zeros((M,M))
            P_ij  = np.zeros((M,M))
            cov_ij= np.zeros((M,M))
            for i in range(M):
                for j in range(M):
                    w_pred = w_prev[i] * self.pi[i,j]
                    f_pr   = self.mu[j] + self.a[j] * f_prev[i]
                    P_pr   = self.a[j]**2 * P_prev[i] + self.Q[j]
                    Cj     = self.C[j]
                    Rj     = self.R[j]
                    S      = P_pr * np.outer(Cj, Cj) + Rj
                    diff   = y_t - Cj * f_pr
                    sign, logdet = np.linalg.slogdet(S)
                    invS   = np.linalg.inv(S)
                    logL   = -0.5*(len(diff)*np.log(2*np.pi)+logdet+diff.T@invS@diff)
                    L      = np.exp(logL)
                    w_upd  = w_pred * L
                    K      = P_pr * (invS @ Cj)
                    f_up   = f_pr + K.dot(diff)
                    P_up   = P_pr - P_pr*(K.dot(Cj))
                    cov    = (1-K.dot(Cj))*self.a[j]*P_prev[i]
                    w_ij[i,j]   = w_upd
                    f_ij[i,j]   = f_up
                    P_ij[i,j]   = P_up
                    cov_ij[i,j] = cov
            total    = w_ij.sum()
            xi[t]    = w_ij / total
            gamma[t] = xi[t].sum(axis=0)
            for j in range(M):
                weights      = w_ij[:,j]
                f_filt[t,j] = (weights*f_ij[:,j]).sum()/weights.sum()
                P_filt[t,j] = (weights*(P_ij[:,j]+(f_ij[:,j]-f_filt[t,j])**2)).sum()/weights.sum()
                P_lag[t,j]  = (weights*cov_ij[:,j]).sum()/weights.sum()
            f_prev = f_filt[t].copy()
            P_prev = P_filt[t].copy()
            w_prev = gamma[t].copy()
        return gamma, xi, f_filt, P_filt, P_lag

    def _maximization(self, y, gamma, xi, f_filt, P_filt, P_lag):
        T, p = y.shape
        M = self.M
        # π の更新
        for i in range(M):
            for j in range(M):
                self.pi[i,j] = xi[1:,i,j].sum() / xi[:-1,i,:].sum()
        # 観測 C, R の更新（欠測値スキップ）
        for j in range(M):
            gamma_j = gamma[:, j]
            obs_idx = ~np.isnan(y[:,0])
            W = gamma_j[obs_idx].sum()
            sum_yf = np.zeros(p)
            sum_f2 = 0.0
            sum_yyt = np.zeros((p,p))
            for t in np.where(obs_idx)[0]:
                ft = f_filt[t,j]
                Pt = P_filt[t,j]
                yt = y[t]
                sum_yf  += gamma_j[t] * yt * ft
                sum_f2  += gamma_j[t] * (ft**2 + Pt)
                sum_yyt += gamma_j[t] * np.outer(yt, yt)
            self.C[j] = sum_yf / sum_f2
            Cj = self.C[j]
            term = sum_yyt - np.outer(Cj,sum_yf) - np.outer(sum_yf,Cj) + sum_f2*np.outer(Cj,Cj)
            self.R[j] = term / W
        # 状態 μ, a, Q の更新
        for j in range(M):
            gamma_j = gamma[1:,j]
            N_j     = gamma_j.sum()
            S0      = (gamma_j * f_filt[1:,j]).sum()
            S1      = (gamma_j * f_filt[:-1,j]).sum()
            S01     = (gamma_j * (f_filt[1:,j]*f_filt[:-1,j] + P_lag[1:,j])).sum()
            S11     = (gamma_j * (f_filt[:-1,j]**2 + P_filt[:-1,j])).sum()
            Mmat    = np.array([[N_j,S1],[S1,S11]])
            bvec    = np.array([S0,S01])
            sol     = np.linalg.solve(Mmat,bvec)
            self.mu[j], self.a[j] = sol
            num_Q = 0.0
            for t in range(1,T):
                v = f_filt[t,j] - self.mu[j] - self.a[j]*f_filt[t-1,j]
                num_Q += gamma[t,j] * (v**2 + P_filt[t,j] + self.a[j]**2*P_filt[t-1,j] - 2*self.a[j]*P_lag[t,j])
            self.Q[j] = num_Q / N_j

    def _log_likelihood(self, gamma):
        return -np.sum(gamma * np.log(gamma + 1e-16))

if __name__ == '__main__':
    # サンプル実行
    np.random.seed(1)
    T = 300
    M = 2
    p = 2
    freq = ['m'] * p  # 各系列の頻度 ('m'=月次,'q'=四半期)

    # 真のパラメータ
    true_pi = np.array([[0.9,0.1],[0.1,0.9]])
    true_mu = np.array([-0.5,0.5])
    true_a  = np.array([0.95,0.5])
    true_Q  = np.array([0.1,0.1])
    true_C  = np.array([[1.0,0.5],[0.8,1.2]])
    true_R  = np.array([np.eye(p)*0.5, np.eye(p)*0.8])

    # シミュレーションデータ生成
    s_true = np.zeros(T, dtype=int)
    f = np.zeros(T)
    y = np.zeros((T,p))
    for t in range(1,T):
        s_true[t] = np.random.choice(M, p=true_pi[s_true[t-1]])
        j = s_true[t]
        f[t]      = true_mu[j] + true_a[j]*f[t-1] + np.sqrt(true_Q[j])*np.random.randn()
        y[t]      = true_C[j]*f[t] + np.random.multivariate_normal(np.zeros(p), true_R[j])

    # 推定
    model = RegimeSwitchingDFM(n_regimes=M, n_obs=p, freq=freq)
    gamma = model.fit(y)

    # プロット
    plt.figure(figsize=(10,5))
    plt.plot(s_true, label='True Regime', lw=1)
    plt.plot(gamma[:,1], label='Est Prob Regime1', lw=1)
    plt.xlabel('Time')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class RegimeSwitchingDFM:
    """
    混合頻度対応レジームスイッチング動的因子モデル（Multi-Factor）
    - n_factors: 因子数
    - freq: 各観測変数の頻度リスト ('m'=月次, 'q'=四半期)
    """
    def __init__(self, n_regimes=2, n_factors=1, n_obs=1, freq=None, max_iter=100, tol=1e-4):
        self.M = n_regimes
        self.k = n_factors
        self.p = n_obs
        self.freq = freq
        self.max_iter = max_iter
        self.tol = tol

        # パラメータ初期化
        self.pi = np.full((self.M, self.M), 1/self.M)
        self.mu = np.zeros((self.M, self.k))
        self.A  = np.array([np.eye(self.k) for _ in range(self.M)])
        self.Q  = np.array([np.eye(self.k) for _ in range(self.M)])
        self.C  = np.ones((self.M, self.p, self.k))
        self.R  = np.array([np.eye(self.p) for _ in range(self.M)])

    def fit(self, y):
        y = np.asarray(y, float)
        if self.freq is not None:
            y = self._preprocess_mixed_frequency(y, self.freq)
        if y.ndim == 1:
            y = y.reshape(-1,1)
        T, p = y.shape
        assert p == self.p, "Observation dimension mismatch."

        # 初期推定値設定
        self.pi[:] = 0.9*np.eye(self.M) + 0.1*(np.ones((self.M,self.M))-np.eye(self.M))/(self.M-1)
        for j in range(self.M):
            self.mu[j] = np.linspace(-1,1,self.k)
            self.A[j] = np.eye(self.k)*0.5
            self.Q[j] = np.eye(self.k)
            self.C[j] = np.ones((self.p,self.k))
            self.R[j] = np.eye(self.p)

        gamma = np.full((T,self.M),1/self.M)
        xi    = np.full((T,self.M,self.M),1/(self.M**2))
        ll_old = -np.inf

        for it in range(self.max_iter):
            gamma, xi, f_filt, P_filt, P_lag = self._expectation(y, gamma)
            self._maximization(y, gamma, xi, f_filt, P_filt, P_lag)
            ll = self._log_likelihood(gamma)
            if abs(ll - ll_old) < self.tol:
                break
            ll_old = ll
        return gamma

    def _preprocess_mixed_frequency(self, y, freq):
        y2 = y.copy()
        T, p = y.shape
        logy = np.log(y)
        d1 = np.vstack([np.zeros((1,p)), np.diff(logy, axis=0)])
        weight = [1,2,3,2,1]
        for j, f in enumerate(freq):
            if f == 'q':
                dm = np.zeros(T)
                for t in range(T):
                    wsum = 0.0
                    for kk,w in enumerate(weight):
                        idx = t-kk
                        if idx>=0:
                            wsum += w * d1[idx,j]
                    dm[t] = wsum/3.0
                y2[:,j] = np.exp(np.cumsum(dm))
        return y2

    def _expectation(self, y, gamma_prev):
        T = y.shape[0]
        M, k, p = self.M, self.k, self.p
        f_filt = np.zeros((T,M,k))
        P_filt = np.zeros((T,M,k,k))
        P_lag  = np.zeros((T,M,k,k))
        xi     = np.zeros((T,M,M))
        gamma  = np.zeros((T,M))
        f_prev = np.zeros((M,k))
        P_prev = np.array([np.eye(k)*1e1 for _ in range(M)])
        w_prev = np.full(M,1/M)
        for t in range(T):
            y_t   = y[t]
            w_ij  = np.zeros((M,M))
            f_ij  = np.zeros((M,M,k))
            P_ij  = np.zeros((M,M,k,k))
            cov_ij= np.zeros((M,M,k,k))
            for i in range(M):
                for j in range(M):
                    w_pred = w_prev[i] * self.pi[i,j]
                    f_pr   = self.mu[j] + self.A[j] @ f_prev[i]
                    P_pr   = self.A[j] @ P_prev[i] @ self.A[j].T + self.Q[j]
                    Cj     = self.C[j]
                    Rj     = self.R[j]
                    S      = Cj @ P_pr @ Cj.T + Rj
                    diff   = y_t - Cj @ f_pr
                    sign, logdet = np.linalg.slogdet(S)
                    invS   = np.linalg.inv(S)
                    logL   = -0.5*(p*np.log(2*np.pi)+logdet+diff.T@invS@diff)
                    L      = np.exp(logL)
                    w_upd  = w_pred * L
                    K      = P_pr @ Cj.T @ invS
                    f_up   = f_pr + K @ diff
                    P_up   = P_pr - K @ Cj @ P_pr
                    cov    = (np.eye(k) - K @ Cj) @ self.A[j] @ P_prev[i]
                    w_ij[i,j]   = w_upd
                    f_ij[i,j]   = f_up
                    P_ij[i,j]   = P_up
                    cov_ij[i,j] = cov
            total    = w_ij.sum()
            xi[t]    = w_ij/total
            gamma[t] = xi[t].sum(axis=0)
            for j in range(M):
                weights = w_ij[:,j]
                W       = weights.sum()
                # 因子のフィルタ推定
                f_sum = (weights[:,None]*f_ij[:,j,:]).sum(axis=0)
                f_filt[t,j] = f_sum/W
                # 共分散更新
                P_sum = np.zeros((k,k))
                for i in range(M):
                    diff_f = f_ij[i,j] - f_filt[t,j]
                    P_sum += weights[i]*(P_ij[i,j] + np.outer(diff_f,diff_f))
                P_filt[t,j] = P_sum/W
                # 1ステップラグ共分散
                cov_sum = np.zeros((k,k))
                for i in range(M):
                    cov_sum += weights[i]*cov_ij[i,j]
                P_lag[t,j] = cov_sum/W
            f_prev = f_filt[t].copy()
            P_prev = P_filt[t].copy()
            w_prev = gamma[t].copy()
        return gamma, xi, f_filt, P_filt, P_lag

    def _maximization(self, y, gamma, xi, f_filt, P_filt, P_lag):
        T = y.shape[0]
        M, k, p = self.M, self.k, self.p
        # 遷移行列更新
        for i in range(M):
            for j in range(M):
                self.pi[i,j] = xi[1:,i,j].sum()/xi[:-1,i,:].sum()
        # 観測行列 C, 共分散 R の更新
        for j in range(M):
            gamma_j = gamma[:,j]
            obs_idx = ~np.isnan(y).any(axis=1)
            W       = gamma_j[obs_idx].sum()
            sum_yf  = np.zeros((p,k))
            sum_ff  = np.zeros((k,k))
            sum_yy  = np.zeros((p,p))
            for t in np.where(obs_idx)[0]:
                ft = f_filt[t,j]
                Pt = P_filt[t,j]
                yt = y[t]
                sum_yf += gamma_j[t]*np.outer(yt,ft)
                sum_ff += gamma_j[t]*(np.outer(ft,ft)+Pt)
                sum_yy += gamma_j[t]*np.outer(yt,yt)
            self.C[j] = sum_yf @ np.linalg.inv(sum_ff)
            term      = sum_yy - self.C[j]@sum_yf.T - sum_yf@self.C[j].T + self.C[j]@sum_ff@self.C[j].T
            self.R[j] = term/W
        # 状態方程式パラメータ mu, A, Q の更新
        for j in range(M):
            gamma_j = gamma[1:,j]
            N_j     = gamma_j.sum()
            sum_f1  = np.zeros(k)
            sum_f0  = np.zeros(k)
            sum_11  = np.zeros((k,k))
            sum_01  = np.zeros((k,k))
            for t in range(1,T):
                wjt = gamma[t,j]
                f1  = f_filt[t,j]
                f0  = f_filt[t-1,j]
                sum_f1 += wjt*f1
                sum_f0 += wjt*f0
                sum_11 += wjt*(np.outer(f0,f0)+P_filt[t-1,j])
                sum_01 += wjt*(np.outer(f1,f0)+P_lag[t,j])
            m0      = sum_f1/N_j
            m1      = sum_f0/N_j
            S11     = sum_11/N_j - np.outer(m1,m1)
            S01     = sum_01/N_j - np.outer(m0,m1)
            self.A[j]  = S01 @ np.linalg.inv(S11)
            self.mu[j] = m0 - self.A[j]@m1
            # Q の更新
            num_Q = np.zeros((k,k))
            for t in range(1,T):
                wjt = gamma[t,j]
                f1  = f_filt[t,j]
                f0  = f_filt[t-1,j]
                v   = f1 - self.mu[j] - self.A[j]@f0
                num_Q += wjt*(np.outer(v,v) + P_filt[t,j] + self.A[j]@P_filt[t-1,j]@self.A[j].T - self.A[j]@P_lag[t,j] - P_lag[t,j]@self.A[j].T)
            self.Q[j] = num_Q/N_j

    def _log_likelihood(self, gamma):
        return -np.sum(gamma * np.log(gamma+1e-16))

if __name__ == '__main__':
    np.random.seed(1)
    T, M, p, k = 300, 2, 2, 1
    freq = ['m']*p
    # 真のパラメータ設定
    true_pi = np.array([[0.9,0.1],[0.1,0.9]])
    true_mu = np.array([[-0.5,0.3],[0.5,-0.2]])
    true_A  = np.array([np.eye(k)*0.95, np.eye(k)*0.5])
    true_Q  = np.array([np.eye(k)*0.1, np.eye(k)*0.1])
    true_C  = np.array([[[1.0,0.5],[0.5,1.0]], [[0.8,1.2],[1.2,0.8]]])
    true_R  = np.array([np.eye(p)*0.5, np.eye(p)*0.8])

    # シミュレーションデータ生成
    s_true = np.zeros(T, dtype=int)
    f      = np.zeros((T,k))
    y      = np.zeros((T,p))
    for t in range(1,T):
        s_true[t] = np.random.choice(M, p=true_pi[s_true[t-1]])
        j = s_true[t]
        f[t] = true_mu[j] + true_A[j]@f[t-1] + np.random.multivariate_normal(np.zeros(k), true_Q[j])
        y[t] = true_C[j]@f[t] + np.random.multivariate_normal(np.zeros(p), true_R[j])

    # 推定実行
    model = RegimeSwitchingDFM(n_regimes=M, n_factors=k, n_obs=p, freq=freq)
    gamma = model.fit(y)

    # 結果プロット
    plt.figure(figsize=(10,5))
    plt.plot(s_true, label='True Regime', lw=1)
    plt.plot(gamma[:,1], label='Estim Prob Regime1', lw=1)
    plt.xlabel('Time')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
from scipy.linalg import inv
from numpy.linalg import slogdet
import matplotlib.pyplot as plt

class RegimeSwitchingDFM:
    """
    Mixed-Frequency Regime-Switching Dynamic Factor Model
    - M: number of regimes
    - r: factor dimension
    - q: VAR order (state dimension = r*q + 1)
    - N: number of observed series
    - freq: list of frequencies for each series ('m'=monthly, 'q'=quarterly)
    - max_iter: maximum EM iterations
    - tol: convergence tolerance for log-likelihood change
    """
    def __init__(self, M, r, q, N, freq=None, max_iter=50, tol=1e-4):
        self.M = M
        self.r = r
        self.q = q
        self.N = N
        self.freq = freq
        self.max_iter = max_iter
        self.tol = tol
        # state dimension (intercept + lags)
        self.dim = r * q + 1
        # initialize transition matrix
        self.Z = np.full((M, M), 1.0 / M)
        # regime-specific parameters
        self.C = np.random.randn(M, N, self.dim)      # factor loadings + intercept
        self.A = np.random.randn(M, self.dim, self.dim)  # state transition
        self.Q = np.array([np.eye(self.dim) for _ in range(M)])
        self.R = np.array([np.eye(N) for _ in range(M)])

    def _preprocess_mixed_frequency(self, y):
        """
        Apply Mariano–Murasawa mixed-frequency approximation for each quarterly series:
        Δ3 y_t ≃ (1/3)[1Δy_t + 2Δy_{t-1} + 3Δy_{t-2} + 2Δy_{t-3} + 1Δy_{t-4}]
        """
        T, N = y.shape
        y2 = y.copy()
        logy = np.log(y + 1e-6)
        d1 = np.vstack([np.zeros((1,N)), np.diff(logy, axis=0)])
        weights = np.array([1,2,3,2,1])
        for j, f in enumerate(self.freq):
            if f == 'q':
                dm = np.zeros(T)
                for t in range(T):
                    idx = np.arange(t, t-5, -1)
                    vals = d1[idx[idx>=0], j]
                    w = weights[:len(vals)]
                    dm[t] = (w * vals).sum() / 3.0
                y2[:, j] = np.exp(np.cumsum(dm))
        return y2

    def _kalman_filter(self, Y):
        T = Y.shape[0]
        dim = self.dim
        # storage
        a_filt = np.zeros((T, self.M, dim))
        P_filt = np.zeros((T, self.M, dim, dim))
        gamma = np.zeros((T, self.M))
        xi = np.zeros((T-1, self.M, self.M))
        # init
        a_prev = np.zeros((self.M, dim))
        P_prev = np.array([np.eye(dim)*1e6 for _ in range(self.M)])
        eigvals, eigvecs = np.linalg.eig(self.Z.T)
        stat = eigvecs[:, np.argmax(np.isclose(eigvals, 1))].real
        gamma_prev = stat / stat.sum()
        for t in range(T):
            # Predict & collapse (Pseudo-Bayesian r=2 collapse)
            a_pred = np.zeros((self.M, dim))
            P_pred = np.zeros((self.M, dim, dim))
            for j in range(self.M):
                w = gamma_prev * self.Z[:, j]
                w /= w.sum()
                # mixture of predictions
                for i in range(self.M):
                    a_pred[j] += w[i] * (self.A[j] @ a_prev[i])
                    P_pred[j] += w[i] * (self.A[j] @ P_prev[i] @ self.A[j].T)
                P_pred[j] += self.Q[j]
            # Update
            for j in range(self.M):
                S = self.C[j] @ P_pred[j] @ self.C[j].T + self.R[j]
                innov = Y[t] - self.C[j] @ a_pred[j]
                K = P_pred[j] @ self.C[j].T @ inv(S)
                a_filt[t,j] = a_pred[j] + K @ innov
                P_filt[t,j] = P_pred[j] - K @ self.C[j] @ P_pred[j]
            # Regime probabilities
            llk = np.zeros((self.M, self.M))
            for i in range(self.M):
                for j in range(self.M):
                    S = self.C[j] @ P_pred[j] @ self.C[j].T + self.R[j]
                    innov = Y[t] - self.C[j] @ a_pred[j]
                    sign, ld = slogdet(S)
                    llk[i,j] = -0.5*(ld + innov.T@inv(S)@innov + self.N*np.log(2*np.pi))
            w_mat = gamma_prev[:,None] * self.Z * np.exp(llk)
            if t>0:
                xi[t-1] = w_prev / w_prev.sum()
            gamma_t = w_mat.sum(axis=0)
            gamma[t] = gamma_t / gamma_t.sum()
            # carry
            gamma_prev = gamma[t].copy()
            a_prev = a_filt[t].copy()
            P_prev = P_filt[t].copy()
            w_prev = w_mat
        return a_filt, P_filt, gamma, xi

    def _maximization(self, Y, a_filt, P_filt, gamma, xi):
        T = Y.shape[0]
        dim = self.dim
        # Update Z
        for i in range(self.M):
            for j in range(self.M):
                self.Z[i,j] = xi[:,i,j].sum() / gamma[:-1,i].sum()
        # Update C, R
        for j in range(self.M):
            W = gamma[:,j]
            # Form sums
            S_xy = np.zeros((self.N, dim))
            S_xx = np.zeros((dim, dim))
            S_yy = np.zeros((self.N, self.N))
            for t in range(T):
                x = a_filt[t,j]
                y = Y[t]
                S_xy += W[t]*np.outer(y, x)
                S_xx += W[t]*np.outer(x, x)
                S_yy += W[t]*np.outer(y, y)
            self.C[j] = S_xy @ inv(S_xx)
            resid = Y - np.array([self.C[j]@a_filt[t,j] for t in range(T)])
            S = (resid.T * W) @ resid / W.sum()
            self.R[j] = np.diag(np.diag(S))
        # Update A, Q
        for j in range(self.M):
            W = gamma[1:,j]
            Xp = a_filt[:-1,j]
            Xf = a_filt[1:,j]
            S_xx = (Xp.T * W) @ Xp
            S_xy = (Xp.T * W) @ Xf
            self.A[j] = S_xy.T @ inv(S_xx)
            resid = Xf - (self.A[j]@Xp.T).T
            S = (resid.T * W) @ resid / W.sum()
            self.Q[j] = S

    def fit(self, y):
        # handle data
        y = np.asarray(y, float)
        if self.freq is not None:
            y = self._preprocess_mixed_frequency(y)
        if y.ndim==1:
            y = y.reshape(-1,1)
        # EM
        ll_old = -np.inf
        for it in range(self.max_iter):
            a_filt, P_filt, gamma, xi = self._kalman_filter(y)
            self._maximization(y, a_filt, P_filt, gamma, xi)
            ll = -np.sum(gamma*np.log(gamma+1e-16))
            if abs(ll-ll_old)<self.tol:
                print(f"Converged at iteration {it}")
                break
            ll_old = ll
        return gamma

if __name__ == '__main__':
    np.random.seed(1)
    T = 300; p = 2; M = 2; r = 1; q = 1
    freq = ['m','q']
    # true params
    pi = np.array([[0.9,0.1],[0.1,0.9]])
    mu = np.array([-0.5,0.5]); a = np.array([0.95,0.5])
    Q = np.array([0.1,0.1]); C = np.array([[1.0,0.5],[0.8,1.2]])
    R = np.array([np.eye(p)*0.5, np.eye(p)*0.8])
    # simulate
    s = np.zeros(T,dtype=int); f = np.zeros((T, r*q+1)); y = np.zeros((T,p))
    for t in range(1,T):
        s[t] = np.random.choice(M, p=pi[s[t-1]])
        j = s[t]
        f[t] = np.array([1, f[t-1,1]])*0
        f[t,1] = mu[j] + a[j]*f[t-1,1] + np.sqrt(Q[j])*np.random.randn()
        y[t] = C[j]*f[t,1] + np.random.multivariate_normal(np.zeros(p), R[j])
    # fit
    model = RegimeSwitchingDFM(M, r, q, p, freq, max_iter=50)
    gamma = model.fit(y)
    # plot
    plt.figure(figsize=(10,4))
    plt.plot(s, label='True Regime')
    plt.plot(gamma[:,1], label='Est Prob Regime1')
    plt.legend(); plt.tight_layout(); plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class RegimeSwitchingDFM:
    """
    混合頻度対応レジームスイッチングDFM
    - freq: 観測変数ごとの頻度リスト('m'=月次, 'q'=四半期)
    """
    def __init__(self, n_regimes=2, n_obs=1, freq=None, max_iter=100, tol=1e-4):
        self.M = n_regimes
        self.p = n_obs
        self.freq = freq
        self.max_iter = max_iter
        self.tol = tol

        # パラメータ初期化
        self.pi = np.full((self.M, self.M), 1/self.M)
        self.mu = np.zeros(self.M)
        self.a  = np.zeros(self.M)
        self.Q  = np.ones(self.M)
        self.C  = np.ones((self.M, self.p))
        self.R  = np.array([np.eye(self.p) for _ in range(self.M)])

    def fit(self, y):
        y = np.asarray(y, float)
        # 混合頻度前処理
        if self.freq is not None:
            y = self._preprocess_mixed_frequency(y, self.freq)

        if y.ndim == 1:
            y = y.reshape(-1,1)
        T, p = y.shape
        assert p == self.p, "Observation dimension mismatch."

        # 初期推定値設定
        self.pi[:] = 0.9*np.eye(self.M) + 0.1*(np.ones((self.M,self.M))-np.eye(self.M))/(self.M-1)
        self.mu[:] = np.linspace(-1,1,self.M)
        self.a[:]  = 0.5
        self.Q[:]  = 1.0
        self.C[:]  = 1.0
        self.R[:]  = np.array([np.eye(self.p) for _ in range(self.M)])

        gamma = np.full((T,self.M),1/self.M)
        xi    = np.full((T,self.M,self.M),1/(self.M**2))
        ll_old = -np.inf

        for it in range(self.max_iter):
            gamma, xi, f_filt, P_filt, P_lag = self._expectation(y, gamma)
            self._maximization(y, gamma, xi, f_filt, P_filt, P_lag)
            ll = self._log_likelihood(gamma)
            if abs(ll - ll_old) < self.tol:
                break
            ll_old = ll
        return gamma

    def _preprocess_mixed_frequency(self, y, freq):
        """
        四半期を月次に近似（Mariano–Murasawa 2003）
        """
        y2 = y.copy()
        T, p = y.shape
        logy = np.log(y2)
        # 四半期観測は同値で埋める (fill-forward)
        for j, f in enumerate(freq):
            if f == 'q':
                last = np.nan
                for t in range(T):
                    if not np.isnan(y[t,j]):
                        last = y[t,j]
                    else:
                        y2[t,j] = last
        logy = np.log(y2)
        d1 = np.vstack([np.zeros((1,p)), np.diff(logy, axis=0)])
        weight = [1,2,3,2,1]
        for j, f in enumerate(freq):
            if f == 'q':
                dm = np.zeros(T)
                for t in range(T):
                    wsum = 0.0
                    for k, w in enumerate(weight):
                        idx = t - k
                        if idx >= 0:
                            wsum += w * d1[idx, j]
                    dm[t] = wsum / 3.0
                # 差分を累積してレベルに戻す
                y2[:, j] = np.exp(np.cumsum(dm))
        return y2

    def _expectation(self, y, gamma_prev):
        T, p = y.shape
        M = self.M
        f_filt = np.zeros((T,M))
        P_filt = np.zeros((T,M))
        P_lag  = np.zeros((T,M))
        xi     = np.zeros((T,M,M))
        gamma  = np.zeros((T,M))
        f_prev  = np.zeros(M)
        P_prev  = np.ones(M)*1e1
        w_prev  = np.full(M,1/M)

        for t in range(T):
            y_t = y[t]
            w_ij  = np.zeros((M,M))
            f_ij  = np.zeros((M,M))
            P_ij  = np.zeros((M,M))
            cov_ij= np.zeros((M,M))
            for i in range(M):
                for j in range(M):
                    w_pred = w_prev[i] * self.pi[i,j]
                    f_pr   = self.mu[j] + self.a[j] * f_prev[i]
                    P_pr   = self.a[j]**2 * P_prev[i] + self.Q[j]
                    Cj     = self.C[j]
                    Rj     = self.R[j]
                    S      = P_pr * np.outer(Cj, Cj) + Rj
                    diff   = y_t - Cj * f_pr
                    sign, logdet = np.linalg.slogdet(S)
                    invS   = np.linalg.inv(S)
                    logL   = -0.5*(len(diff)*np.log(2*np.pi)+logdet+diff.T@invS@diff)
                    L      = np.exp(logL)
                    w_upd  = w_pred * L
                    K      = P_pr * (invS @ Cj)
                    f_up   = f_pr + K.dot(diff)
                    P_up   = P_pr - P_pr*(K.dot(Cj))
                    cov    = (1-K.dot(Cj))*self.a[j]*P_prev[i]
                    w_ij[i,j]   = w_upd
                    f_ij[i,j]   = f_up
                    P_ij[i,j]   = P_up
                    cov_ij[i,j] = cov
            total    = w_ij.sum()
            xi[t]    = w_ij / total
            gamma[t] = xi[t].sum(axis=0)
            for j in range(M):
                w_ = w_ij[:,j]
                f_filt[t,j] = (w_*f_ij[:,j]).sum()/w_.sum()
                P_filt[t,j] = (w_*(P_ij[:,j]+(f_ij[:,j]-f_filt[t,j])**2)).sum()/w_.sum()
                P_lag[t,j]  = (w_*cov_ij[:,j]).sum()/w_.sum()
            f_prev = f_filt[t].copy()
            P_prev = P_filt[t].copy()
            w_prev = gamma[t].copy()
        return gamma, xi, f_filt, P_filt, P_lag

    def _maximization(self, y, gamma, xi, f_filt, P_filt, P_lag):
        T, p = y.shape
        M = self.M
        # π の更新
        for i in range(M):
            for j in range(M):
                self.pi[i,j] = xi[1:,i,j].sum() / xi[:-1,i,:].sum()
        # C, R の更新
        for j in range(M):
            g = gamma[:,j]
            obs = ~np.isnan(y[:,0])
            W = g[obs].sum()
            sum_yf = np.zeros(p)
            sum_f2 = 0.0
            sum_yyt = np.zeros((p,p))
            for t in np.where(obs)[0]:
                ft = f_filt[t,j]
                Pt = P_filt[t,j]
                yt = y[t]
                sum_yf  += g[t]*yt*ft
                sum_f2  += g[t]*(ft**2+Pt)
                sum_yyt += g[t]*np.outer(yt,yt)
            self.C[j] = sum_yf/sum_f2
            Cj = self.C[j]
            term = sum_yyt - np.outer(Cj,sum_yf) - np.outer(sum_yf,Cj) + sum_f2*np.outer(Cj,Cj)
            self.R[j] = term/W
        # μ, a, Q の更新
        for j in range(M):
            gj = gamma[1:,j]
            Nj = gj.sum()
            S0 = (gj * f_filt[1:,j]).sum()
            S1 = (gj * f_filt[:-1,j]).sum()
            S01 = (gj * (f_filt[1:,j]*f_filt[:-1,j] + P_lag[1:,j])).sum()
            S11 = (gj * (f_filt[:-1,j]**2 + P_filt[:-1,j])).sum()
            Mmat = np.array([[Nj,S1],[S1,S11]])
            bvec = np.array([S0,S01])
            sol = np.linalg.solve(Mmat,bvec)
            self.mu[j], self.a[j] = sol
            num_Q = 0.0
            for t in range(1,T):
                v = f_filt[t,j] - self.mu[j] - self.a[j]*f_filt[t-1,j]
                num_Q += gamma[t,j]*(v**2 + P_filt[t,j] + self.a[j]**2*P_filt[t-1,j] - 2*self.a[j]*P_lag[t,j])
            self.Q[j] = num_Q/Nj

    def _log_likelihood(self, gamma):
        return -np.sum(gamma * np.log(gamma + 1e-16))


if __name__ == '__main__':
    # サンプル実行（混合頻度：1列目=m, 2列目=q）
    np.random.seed(42)
    T = 300
    M = 2
    p = 2
    freq = ['m','q']  # 1列目：月次, 2列目：四半期

    # 真のパラメータ
    true_pi = np.array([[0.9,0.1],[0.1,0.9]])
    true_mu = np.array([-0.5, 0.5])
    true_a  = np.array([0.95, 0.5])
    true_Q  = np.array([0.1, 0.1])
    true_C  = np.array([[1.0, 0.8], [0.5, 1.2]])  # 2 regimes × 2 obs
    true_R  = np.array([np.eye(p)*0.5, np.eye(p)*0.8])

    # シミュレーションデータ生成
    s_true = np.zeros(T, dtype=int)
    f = np.zeros(T)
    y = np.zeros((T,p))
    # 四半期系列はfill-forward用にnanを入れておく
    y[:,1] = np.nan
    for t in range(1,T):
        s_true[t] = np.random.choice(M, p=true_pi[s_true[t-1]])
        j = s_true[t]
        # 因子
        f[t] = true_mu[j] + true_a[j]*f[t-1] + np.sqrt(true_Q[j])*np.random.randn()
        # 月次系列
        y[t,0] = true_C[j,0]*f[t] + np.random.randn()*np.sqrt(true_R[j,0,0])
        # 四半期系列：3ヶ月に1回のみ観測し、それ以外は nan
        if (t % 3) == 0:
            y[t,1] = true_C[j,1]*f[t] + np.random.randn()*np.sqrt(true_R[j,1,1])

    # モデル推定
    model = RegimeSwitchingDFM(n_regimes=M, n_obs=p, freq=freq)
    gamma = model.fit(y)

    # プロット
    plt.figure(figsize=(10,5))
    plt.plot(s_true,       label='True Regime',       lw=1)
    plt.plot(gamma[:,1],   label='Est Prob Regime 1', lw=1)
    plt.xlabel('Time')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

class RS_MixedFreqDFM:
    """
    Regime-Switching Dynamic Factor Model
      • latent daily factor f_t  (AR(1))
      • n_D  daily series:           y_D[:,d]  = c_D[j,d] * f_t + ε_t,d
      • n_M  monthly series (month-end only):
            y_M[:,k]  = c_M[j,k] * (1/N) Σ_{d=0}^{N-1} f_{t-d} + ε_t,k
      • EM アルゴリズム（修正カルマンフィルタ＋スムーザ）
    観測行列 C_t およびノイズ分散 R_t をレジーム j ごとに切り替えます。
    """

    def __init__(self, n_regimes=2, L=31, max_iter=100, tol=1e-4):
        self.M = n_regimes
        self.L = L
        self.max_iter = max_iter
        self.tol = tol

        # --- レジーム別状態パラメータ ---
        self.pi = np.full((self.M, self.M), 1/self.M)
        self.mu = np.zeros(self.M)
        self.a  = np.zeros(self.M)
        self.Q  = np.ones(self.M)

        # --- レジーム別観測ローディング＆分散 ---
        # fit() 内で形状が決まる
        self.c_D = None    # shape (M, n_D)
        self.c_M = None    # shape (M, n_M)
        self.R_D = None    # shape (M, n_D)
        self.R_M = None    # shape (M, n_M)

    def fit(self, y_D, y_M, dates):
        y_D = np.asarray(y_D)
        y_M = np.asarray(y_M)
        T, n_D = y_D.shape
        _, n_M = y_M.shape
        self.n_D, self.n_M = n_D, n_M

        # ----- 初期化 -----
        # 遷移行列
        self.pi[:] = 0.9*np.eye(self.M) + 0.1/(self.M-1)*(1-np.eye(self.M))
        # AR(1)因子
        self.mu[:] = np.linspace(-0.5, 0.5, self.M)
        self.a[:]  = 0.6
        self.Q[:]  = 0.3
        # 観測ローディング・分散
        self.c_D = np.ones((self.M, n_D))
        self.c_M = np.ones((self.M, n_M))
        self.R_D = np.ones((self.M, n_D))*0.5
        self.R_M = np.ones((self.M, n_M))*0.5

        # ----- EM ループ -----
        gamma = np.full((T, self.M), 1/self.M)
        xi    = np.full((T, self.M, self.M), 1/self.M**2)
        ll_old = -np.inf

        for _ in range(self.max_iter):
            gamma, xi, F, P, P_lag = self._expectation(y_D, y_M, dates, gamma)
            self._maximization(y_D, y_M, dates, gamma, xi, F, P, P_lag)
            ll = -np.sum(gamma * np.log(gamma + 1e-16))
            if abs(ll - ll_old) < self.tol:
                break
            ll_old = ll

        return gamma

    def _expectation(self, y_D, y_M, dates, gamma_prev):
        T, L, M = len(y_D), self.L, self.M
        n_D, n_M = self.n_D, self.n_M
        I_L = np.eye(L)

        # 結果格納変数
        F_filt = np.zeros((T, M, L))
        P_filt = np.zeros((T, M, L, L))
        P_lag  = np.zeros((T, M))
        xi     = np.zeros((T, M, M))
        gamma  = np.zeros((T, M))

        # 初期値
        F_prev = np.zeros((M, L))
        P_prev = np.tile(np.eye(L)*10.0, (M,1,1))
        w_prev = np.full(M, 1/M)

        # レジーム別状態遷移行列 A を準備
        A_mat = np.zeros((M, L, L))
        for j in range(M):
            A = np.zeros((L,L))
            A[0,0] = self.a[j]
            A[1:, :-1] = np.eye(L-1)
            A_mat[j] = A

        for t in range(T):
            # 月次用ウェイト
            if not np.isnan(y_M[t]).all():
                N = self._month_length(dates[t])
                w_month = np.zeros(L); w_month[:N] = 1.0/N
            else:
                w_month = None

            # 各レジーム j ごとに観測行列 C_t^(j) と分散 R_t^(j) を構築
            C_list, R_list = [], []
            for j in range(M):
                obs_rows, obs_var = [], []
                # daily
                for d in range(n_D):
                    row = np.zeros(L)
                    row[0] = self.c_D[j,d]
                    obs_rows.append(row)
                    obs_var.append(self.R_D[j,d])
                # monthly
                if w_month is not None:
                    for k in range(n_M):
                        if np.isnan(y_M[t,k]): continue
                        row = w_month * self.c_M[j,k]
                        obs_rows.append(row)
                        obs_var.append(self.R_M[j,k])
                C_list.append(np.vstack(obs_rows))
                R_list.append(np.diag(obs_var))

            # パスごとのカルマンフィルタ
            w_ij   = np.zeros((M,M))
            F_ij   = np.zeros((M,M,L))
            P_ij   = np.zeros((M,M,L,L))
            cov_ij = np.zeros((M,M))

            for i in range(M):
                for j in range(M):
                    # 1ステップ予測
                    w_pred = w_prev[i] * self.pi[i,j]
                    F_pr = self.mu[j] + self.a[j]*F_prev[i,0]
                    F_pr = np.concatenate([[F_pr], F_prev[i,:-1]])
                    P_pr = A_mat[j] @ P_prev[i] @ A_mat[j].T
                    P_pr[0,0] += self.Q[j]

                    # 観測更新
                    C_t = C_list[j]
                    R_t = R_list[j]
                    # 観測値ベクトル
                    obs_vals = list(y_D[t])
                    if w_month is not None:
                        for k in range(n_M):
                            if not np.isnan(y_M[t,k]):
                                obs_vals.append(y_M[t,k])
                    obs_vals = np.asarray(obs_vals)
                    k_obs = obs_vals.size

                    S = C_t @ P_pr @ C_t.T + R_t
                    invS = np.linalg.inv(S)
                    resid = obs_vals - C_t @ F_pr
                    lh = np.exp(-0.5 * resid @ invS @ resid) / np.sqrt((2*np.pi)**k_obs * np.linalg.det(S))
                    K = P_pr @ C_t.T @ invS
                    F_up = F_pr + K @ resid
                    P_up = (I_L - K @ C_t) @ P_pr

                    w_ij[i,j]   = w_pred * lh
                    F_ij[i,j]   = F_up
                    P_ij[i,j]   = P_up
                    cov_ij[i,j] = P_up[0,1] if L>1 else 0.0

            # mixture collapse
            total = w_ij.sum()
            xi[t]   = w_ij / total
            gamma[t]= xi[t].sum(axis=0)
            for j in range(M):
                wj = w_ij[:,j]
                W  = wj.sum()
                F_filt[t,j] = (wj[:,None] * F_ij[:,j]).sum(axis=0) / W
                diff = F_ij[:,j] - F_filt[t,j]
                P_filt[t,j] = (wj[:,None,None] * (P_ij[:,j] + diff[:,:,None]*diff[:,None,:])).sum(axis=0) / W
                P_lag[t,j]  = (wj * cov_ij[:,j]).sum() / W

            # 次ステップへ
            F_prev, P_prev, w_prev = F_filt[t].copy(), P_filt[t].copy(), gamma[t].copy()

        return gamma, xi, F_filt, P_filt, P_lag

    def _maximization(self, y_D, y_M, dates, gamma, xi, F, P, P_lag):
        T, n_D, n_M, M = len(y_D), self.n_D, self.n_M, self.M

        # (1) π の更新
        for i in range(M):
            denom = gamma[:-1,i].sum()
            if denom > 0:
                self.pi[i] = xi[1:,i].sum(axis=0) / denom

        # (2) daily loadings & variances の更新
        num_cD   = np.zeros((M,n_D))
        den_cD   = np.zeros(M)
        num_R_D  = np.zeros((M,n_D))
        for t in range(T):
            for j in range(M):
                f0   = F[t,j,0]
                var0 = P[t,j,0,0]
                den_cD[j] += gamma[t,j] * (f0**2 + var0)
                for d in range(n_D):
                    num_cD[j,d]  += gamma[t,j] * y_D[t,d] * f0
                    num_R_D[j,d] += gamma[t,j] * ((y_D[t,d] - self.c_D[j,d]*f0)**2 + self.c_D[j,d]**2 * var0)
        for j in range(M):
            if den_cD[j] > 0:
                self.c_D[j] = num_cD[j]/den_cD[j]
                self.R_D[j] = num_R_D[j]/den_cD[j]

        # (3) monthly loadings & variances の更新
        num_cM   = np.zeros((M,n_M))
        den_cM   = np.zeros(M)
        num_R_M  = np.zeros((M,n_M))
        for t in range(T):
            if np.isnan(y_M[t]).all(): continue
            N = self._month_length(dates[t])
            w_month = np.zeros(self.L); w_month[:N]=1.0/N
            for j in range(M):
                agg     = w_month @ F[t,j]
                var_agg = w_month @ P[t,j] @ w_month
                den_cM[j] += gamma[t,j] * (agg**2 + var_agg)
                for k in range(n_M):
                    if np.isnan(y_M[t,k]): continue
                    num_cM[j,k]  += gamma[t,j] * y_M[t,k] * agg
                    num_R_M[j,k] += gamma[t,j] * ((y_M[t,k] - self.c_M[j,k]*agg)**2 + self.c_M[j,k]**2 * var_agg)
        for j in range(M):
            if den_cM[j] > 0:
                self.c_M[j] = num_cM[j]/den_cM[j]
                self.R_M[j] = num_R_M[j]/den_cM[j]

        # (4) μ, a, Q の更新
        for j in range(M):
            w = gamma[1:,j]
            Nj = w.sum()
            if Nj < 1e-8: 
                continue

            S0  = (w * F[1:,j,0]).sum()
            S1  = (w * F[:-1,j,0]).sum()
            S01 = (w * (F[1:,j,0]*F[:-1,j,0] + P_lag[1:,j])).sum()
            S11 = (w * (F[:-1,j,0]**2 + P[:-1,j,0,0])).sum()

            mat = np.array([[Nj, S1],[S1, S11]])
            vec = np.array([S0, S01])
            sol = np.linalg.solve(mat, vec)
            self.mu[j], self.a[j] = sol[0], sol[1]

            numQ = 0.0
            for t in range(1, T):
                v = F[t,j,0] - self.mu[j] - self.a[j]*F[t-1,j,0]
                numQ += gamma[t,j] * (
                    v**2 + P[t,j,0,0] +
                    self.a[j]**2 * P[t-1,j,0,0] -
                    2*self.a[j]*P_lag[t,j]
                )
            self.Q[j] = numQ / Nj

    @staticmethod
    def _month_length(ts):
        """その月の日数を返す"""
        start = ts - pd.offsets.MonthBegin(n=1)
        return (ts - start).days


# =====================
# シミュレーション & 推定例
# =====================
np.random.seed(1)
dates = pd.date_range("2021-01-01", "2022-12-31", freq="D")
T = len(dates)

# 真のパラメータ
Mtrue   = 2
pi_true = np.array([[0.95,0.05],[0.04,0.96]])
mu_true = np.array([-0.3,0.5])
a_true  = np.array([0.97,0.65])
Q_true  = np.array([0.06,0.11])

c_D_true = np.array([ 1.0, -0.8])       # 2 daily series
c_M_true = np.array([ 0.9, -0.7, 0.5])  # 3 monthly series
R_D_true = np.array([0.4, 0.35])
R_M_true = np.array([0.3, 0.25, 0.28])

# データ生成
s   = np.zeros(T, dtype=int)
f   = np.zeros(T)
y_D = np.zeros((T,2))
y_M = np.full((T,3), np.nan)

for t in range(1, T):
    s[t] = np.random.choice(2, p=pi_true[s[t-1]])
    j    = s[t]
    f[t] = mu_true[j] + a_true[j]*f[t-1] + np.sqrt(Q_true[j])*np.random.randn()
    y_D[t] = c_D_true * f[t] + np.sqrt(R_D_true)*np.random.randn(2)

    if dates[t].is_month_end:
        N     = (dates[t] - (dates[t]-pd.offsets.MonthBegin(1))).days
        avg_f = f[t-N+1:t+1].mean()
        y_M[t] = c_M_true * avg_f + np.sqrt(R_M_true)*np.random.randn(3)

# モデル推定
model = RS_MixedFreqDFM(n_regimes=2, L=31, max_iter=100, tol=1e-4)
gamma = model.fit(y_D, y_M, dates)

# 結果プロット
plt.figure(figsize=(10,4))
plt.plot(s,        label="True Regime",        lw=0.6)
plt.plot(gamma[:,1], label="Est. Prob(Regime 1)", lw=0.8)
plt.legend(); plt.tight_layout(); plt.show()
```
