# Cyclic Mispricing Timing Strategy

### 投資家の誤認知が市場の予測可能性を生むと仮定

## ２つの誤認知
- Correlation Neglect（相関無視）: 
企業決算が集中する「newsy month」（四半期の第1月：1,4,7,10月）と、それに続く月（第2・第3月）のリターンには、**継続（momentum）→反転（reversal）**という特異なパターンがある。

四半期ごとに 3 ヶ月のサイクルがあり、

1 ヶ月目：Newsy（本決算に近く、新しい情報が多い）

2 ヶ月目：Repetitive（1 ヶ月目の内容と高い相関を持つ追いニュース）

3 ヶ月目：3rd month（ニュースが乏しい）

という構造。Guo & Wachter (2025) では、2 ヶ月目のアナウンスは 1 ヶ月目と高相関なのに、投資家はそれを独立に近いと誤認し「相関を無視（correlation neglect）」する、という行動仮定が置かれ

## 無条件平均とミスプライシングの分解
市場リターン$R_t$を
$$R_t=R_t^{\text{fund}}+R_t^{\text{mis}}$$
に分解する。ここで：
- $R_t^{\text{fund}}$：合理的な割引係数とキャッシュフローのもとで決まるファンダメンタルリターン
- $R_t^{\text{mis}}$：earnings extrapolation / correlation neglect によって生じる期待の歪み

四半期ごとの「Newsy / Repetitive / Quiet」構造と earnings の自己相関から、
- Newsy 月のリターンは「新しい四半期の決算」であり過去との相関が低い
- Non-newsy 月（特に Repetitive）は過去の決算とほぼ同じ情報を再表明しているため相関が高い
にもかかわらず、投資家がこれを十分に区別しないために、earnings を誤った係数で外挿してしまい、それが $R_t^{\text{mis}}$ に現れる、という形で定式化

市場リターンは 本源的プレミアム + サイクル誤差 に分解される
$$r_t=\mu+\eta_t+\epsilon_t$$
ここで：
$\mu$：長期リスクプレミアム
$\eta_t$:：情報サイクル起源のミスプライシング成分
$\epsilon_t$：ノイズ

ミスプライシング成分は Newsy 月の累積ショックで近似
Newsy 月の総合指標：
$$S_t=\sum_{j=0}^{3}R_{N_{t-j}}$$

直近4回の Newsy 月のトータルリターンの合計）
しかし S そのままでは Newsy 月・非 Newsy 月で 符号が変わるため
一つのシグナルとして扱いづらい

3. x_t の構築（理論から導かれる Mispricing 指標）

Guo は 対称化した Mispricing Signal を定義：

$$x_t=\text{sign}(t)\cdot(S_{t-1}-\bar{S}_{t-1})$$

Newsy 月 ⇒ sign = −1
→ 過剰反応の修正方向を反映

非 Newsy 月 ⇒ sign = +1
→ 遅延反応（under-reaction）を反映

これにより 1 変数で全期間のミスプライシングを表現できる。

4. 予測モデルの導出（Method 6 の中核）

市場のミスプライシングが線形にリターンへ反映されると仮定すると：
$$r_{t+1}-\bar{r}_t=\beta x_t+\epsilon_{t+1}$$

- 左辺は「市場リターンのランニング平均からの乖離」
- 右辺は「情報サイクル起源の mispricing シグナル」
- $\beta$は expanding もしくは rolling OLS で時変推定

戦略シグナルは：
$$w_t^{\text{raw}}=\beta_tx_t$$

CMT（Cyclic Mispricing Timing）戦略の定義
$$w_t=k\cdot\beta_tx_t$$
ここで$k$はボラティリティターゲッティング係数
$$k=\frac{\sigma^*}{\sigma_{\text{strategy}}}$$

6. 理論仮定の実証的補強（Method 6 の妥当性チェック）

以下の 実証分析プログラムを用いることで、理論仮定を裏付けられる。

（1）S の符号反転（Newsy vs Non-Newsy）

Newsy 月：
$$\beta_N\lt0$$

Non-Newsy 月：
$$\beta_N\gt0$$

実証結果（プログラムで確認）
→ Guo が主張する ニュース過剰反応（Newsy）・遅延反応（R+3） の非対称性が再現される。

2）x_t の predictability（リターン予測力）

回帰：
$$r_{t+1}=\alpha+\beta x_t + \epsilon_t$$

実証では：
$β > 0$（有意）
$x$ の五分位ソートで 単調に $r_{t+1}$ が上昇
ローリング相関でも関係が安定
→ $x$ が有効な Mispricing Signalであることを補強。

Guo (2025) は、
「四半期第1月（newsy month：1,4,7,10 月など）での市場リターンが、その後の月次リターンを強く予測する」ことを示します。
具体的には、

予測対象月が newsy のとき：過去の newsy 月リターンと強い負の相関

予測対象月が non-newsy（第2・3月） のとき：過去の newsy 月リターンと強い正の相関

という「符号の反転を伴う」予測パターンが存在し、これが従来の「月次市場リターンはほとんど予測不能」という結果をマスクしている、と主張します。
OUP Academic

このパターンは、投資家が

earnings（特に newsy 月の決算発表）を将来に向けて外挿する一方で、

「newsy 月と non-newsy 月で earnings の自己相関構造が異なる（newsy のほうが予測しづらい）」ことを十分に認識していない

という「earnings extrapolation + correlation neglect（相関構造の無視）」モデルから導出されます。

誤認知には主に 2 種類がある：

Correlation Neglect（相関無視）

投資家が「news（情報ショック）を“独立”だと誤解する」

実際には news が持続/分解構造を持つのに、それを無視する。

結果として、newsy → repetitive で momentum や reversal が起きる。

Parameter Compression（パラメータ圧縮）

投資家は複雑なパラメータ序列を「圧縮して単純化」して理解する。

earnings announcement のようなイベントで誤った updating が発生する。

## 仮説
### 決算サイクルの構造
決算発表日は企業ごとにばらつきがあり, 四半期終了後3か月以内に分散して発表される.

統計的には四半期内の発表は概ね以下のように配分される：
- 第1月（newsy month）：約45%
- 第2月（repetitive month）：約48%
- 第3月：残り少数（約8%）

市場では短期的に以下のバイアスが生じる：
- 新しい情報が集中的に反映される月（Newsy）
- その情報が惰性的に繰り返される月（Repetitive）

これにより $E[r_t]$ は３ヶ月周期の構造を持つ
